```python
#### Rock–Paper–Scissors Python project

Beginner Version (straightforward, matches your screenshot logic)
Intermediate Version (functions + input validation)
Advanced Version (OOP + replay + score tracking)

⭐ 1. Beginner Version — Matches Your Screenshot Logic
python
            import random

            print("=== Rock Paper Scissors Game ===")

            options = ["rock", "paper", "scissor"]

            player = input("Enter your choice (rock/paper/scissor): ").lower()
            computer = random.choice(options)

            print(f"Computer chose: {computer}")

            # Game logic
            if player == computer:
                print("It's a tie!")

            elif player == "rock":
                if computer == "paper":
                    print("Paper wins!")
                else:
                    print("Rock wins!")

            elif player == "paper":
                if computer == "scissor":
                    print("Scissor wins!")
                else:
                    print("Paper wins!")

            elif player == "scissor":
                if computer == "rock":
                    print("Rock wins!")
                else:
                    print("Scissor wins!")

            else:
                print("Invalid input. Please choose rock, paper, or scissor.")
==========================================================
⭐ 2. Intermediate Version — Clean Functions + Validation
python
        import random

        OPTIONS = ["rock", "paper", "scissor"]

        def get_winner(player, computer):
            if player == computer:
                return "tie"

            if (player == "rock" and computer == "scissor") or \
              (player == "paper" and computer == "rock") or \
              (player == "scissor" and computer == "paper"):
                return "player"

            return "computer"


        def main():
            print("=== Rock Paper Scissors ===")

            player = input("Enter rock/paper/scissor: ").lower()
            if player not in OPTIONS:
                print("Invalid choice.")
                return

            computer = random.choice(OPTIONS)
            print(f"Computer chose: {computer}")

            result = get_wwinner(player, computer)

            if result == "tie":
                print("It's a tie!")
            elif result == "player":
                print("You win!")
            else:
                print("Computer wins!")


        if __name__ == "__main__":
            main()
==================================
⭐ 3. Advanced Version — OOP + Scoreboard + Replay Loop
python
            import random

            class RPSGame:
                OPTIONS = ["rock", "paper", "scissor"]

                def __init__(self):
                    self.player_score = 0
                    self.computer_score = 0

                def get_computer_choice(self):
                    return random.choice(self.OPTIONS)

                def get_winner(self, player, computer):
                    if player == computer:
                        return "tie"

                    wins = {
                        "rock": "scissor",
                        "paper": "rock",
                        "scissor": "paper"
                    }

                    return "player" if wins[player] == computer else "computer"

                def play_round(self):
                    player = input("Enter rock/paper/scissor: ").lower()
                    if player not in self.OPTIONS:
                        print("Invalid choice.")
                        return

                    computer = self.get_computer_choice()
                    print(f"Computer chose: {computer}")

                    result = self.get_winner(player, computer)

                    if result == "tie":
                        print("It's a tie!")
                    elif result == "player":
                        print("You win!")
                        self.player_score += 1
                    else:
                        print("Computer wins!")
                        self.computer_score += 1

                    print(f"Score → You: {self.player_score} | Computer: {self.computer_score}")

                def start(self):
                    print("=== Rock Paper Scissors Game ===")
                    while True:
                        self.play_round()
                        again = input("Play again? (y/n): ").lower()
                        if again != "y":
                            print("Thanks for playing!")
                            break


            if __name__ == "__main__":
                game = RPSGame()
                game.start()
```

```python
##
def parse_headers(header_line):
    return header_line.strip().split(",")

def parse_values(data_line):
    return [float(item) if item.replace('.','',1).isdigit() else item
            for item in data_line.strip().split(",")]

def create_item_dict(values, headers):
    return dict(zip(headers, values))

def read_csv(path):
    result = []
    with open(path, 'r') as f:
        lines = f.readlines()
        headers = parse_headers(lines[0])
        for data_line in lines[1:]:
            values = parse_values(data_line)
            item_dict = create_item_dict(values, headers)
            result.append(item_dict)
    return result
```

```python
### Handling data using SQLITE3

import sqlite3
import time
import datetime
import random
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib import style

style.use('fivethirtyeight')

# -----------------------------------------
# Database connection
# -----------------------------------------
conn = sqlite3.connect('tutorial.db')
c = conn.cursor()

# -----------------------------------------
# Table creation
# -----------------------------------------
def create_table():
    c.execute('CREATE TABLE IF NOT EXISTS stuffToPlot(unix REAL, datestamp TEXT, keyword TEXT, value REAL)')

# -----------------------------------------
# Static data entry
# -----------------------------------------
def data_entry():
    c.execute("INSERT INTO stuffToPlot VALUES(1451255552, '2016-01-02', 'Python', 8)")
    conn.commit()
    c.close()
    conn.close()

# -----------------------------------------
# Dynamic data entry
# -----------------------------------------
def dynamic_data_entry():
    unix = time.time()
    date = str(datetime.datetime.fromtimestamp(unix).strftime('%Y-%m-%d %H:%M:%S'))
    keyword = 'Python'
    value = random.randrange(0, 10)
    c.execute("INSERT INTO stuffToPlot (unix, datestamp, keyword, value) VALUES (?, ?, ?, ?)",
              (unix, date, keyword, value))
    conn.commit()

# -----------------------------------------
# Read from DB
# -----------------------------------------
def read_from_db():
    c.execute("SELECT keyword, unix FROM stuffToPlot WHERE unix > 1452618731")
    for row in c.fetchall():
        print(row)

# -----------------------------------------
# Graph data
# -----------------------------------------
def graph_data():
    c.execute("SELECT unix, value FROM stuffToPlot")
    dates = []
    values = []
    for row in c.fetchall():
        dates.append(datetime.datetime.fromtimestamp(row[0]))
        values.append(row[1])

    plt.plot_date(dates, values, '-')
    plt.show()

# -----------------------------------------
# Delete and update operations
# (combined from both update/delete images)
# -----------------------------------------
def del_and_update():
    # Initial table print
    c.execute('SELECT * FROM stuffToPlot')
    [print(row) for row in c.fetchall()]

    # Optional update (commented in image)
    # c.execute('UPDATE stuffToPlot SET value = 99 WHERE value = 8')
    # conn.commit()
    # c.execute('SELECT * FROM stuffToPlot')
    # [print(row) for row in c.fetchall()]

    # Delete rows where value = 99
    c.execute('DELETE FROM stuffToPlot WHERE value = 99')
    conn.commit()

    print(50 * '#')

    # Final table print
    c.execute('SELECT * FROM stuffToPlot')
    [print(row) for row in c.fetchall()]
```
