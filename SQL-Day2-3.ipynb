```python
✅ 1. Program that prints “Hello, World!” with comments
Below is a simple Python script that includes:

A single‑line comment

A multi‑line comment

A correct print statement

python
# This is a single-line comment: the program starts here

"""
This is a multi-line comment.
It explains that the program prints 'Hello, World!' to the console.
"""

print("Hello, World!")
=================

📚 Reading List — Explained Clearly
Here’s a crisp, interview‑ready breakdown of your three study topics.
-----------------------------
1. How does Python compile and run the code?
Python uses a hybrid model:

🔹 Step 1: Your code → Bytecode
Python compiles your .py file into bytecode (a low‑level, platform‑independent representation).

🔹 Step 2: Bytecode → Python Virtual Machine (PVM)
The PVM interprets the bytecode line by line.

🔹 Summary

Python = Compiled to bytecode + Interpreted by PVM

This is why Python feels dynamic and flexible.
--------------------------------
2. Difference between Python and Java

Feature	Python	Java
Typing----Dynamically typed	------------- Statically typed
Syntax---	Very concise	 -----------------Verbose
Speed	----Slower (interpreted)------------	Faster (compiled to JVM bytecode)
Use cases--	AI, ML, scripting, automation--	Enterprise apps, Android, large systems
Learning curve---	Beginner‑friendly---------	More strict, steeper

Quick memory trick
============
Python = flexibility + readability

Java = structure + performance
====================================
3. Difference between Compiler and Interpreter

Concept--------------	Compiler	-------------------Interpreter
Execution	-----Translates entire program at once-----	Executes line by line
Output-------	Machine code / bytecode--------------	Immediate execution
Speed--------	Faster at runtime	-----------------Slower at runtime
Error detection----	Shows all errors after compilation---	Stops at first error
Examples------	C, C++, Java (JVM)---------	Python (PVM), JavaScript

Easy way to remember
==================
Compiler = “Translate the whole book first.”

Interpreter = “Read and act one line at a time.”
```

```python
⭐ 1. Find customer_id, first_name, last_name, age, item, country

Solution A — INNER JOIN

            SELECT
                c.customer_id,
                c.first_name,
                c.last_name,
                c.age,
                o.item,
                c.country
            FROM customers c
            JOIN orders o
                ON c.customer_id = o.customer_id;

Solution B — LEFT JOIN (keeps customers even without orders)
            SELECT
                c.customer_id,
                c.first_name,
                c.last_name,
                c.age,
                o.item,
                c.country
            FROM customers c
            LEFT JOIN orders o
                ON c.customer_id = o.customer_id;

Solution C — Using a subquery
          SELECT
              customer_id, first_name, last_name, age, country,
              (SELECT item FROM orders o WHERE o.customer_id = c.customer_id) AS item
          FROM customers c;
===================================================
⭐ 2. Find customers who have not placed any orders

Solution A — LEFT JOIN + NULL filter
            SELECT c.*
            FROM customers c
            LEFT JOIN orders o
                ON c.customer_id = o.customer_id
            WHERE o.customer_id IS NULL;

Solution B — NOT IN
          SELECT *
          FROM customers
          WHERE customer_id NOT IN (SELECT customer_id FROM orders);

Solution C — NOT EXISTS
            SELECT *
            FROM customers c
            WHERE NOT EXISTS (
                SELECT 1
                FROM orders o
                WHERE o.customer_id = c.customer_id
            );
==================================================
⭐ 3. Find customers from USA who placed orders

Solution A — INNER JOIN
            SELECT DISTINCT c.*
            FROM customers c
            JOIN orders o
                ON c.customer_id = o.customer_id
            WHERE c.country = 'USA';

Solution B — EXISTS
          SELECT *
          FROM customers c
          WHERE c.country = 'USA'
            AND EXISTS (
                  SELECT 1
                  FROM orders o
                  WHERE o.customer_id = c.customer_id
            );

Solution C — IN
          SELECT *
          FROM customers
          WHERE country = 'USA'
            AND customer_id IN (SELECT customer_id FROM orders);
=============================================
⭐ 4. Get orders where amount > 500 along with customer names

Solution A — JOIN
          SELECT
              o.order_id,
              o.amount,
              c.first_name,
              c.last_name
          FROM orders o
          JOIN customers c
              ON o.customer_id = c.customer_id
          WHERE o.amount > 500;

Solution B — Subquery
          SELECT *
          FROM orders
          WHERE amount > 500
            AND customer_id IN (SELECT customer_id FROM customers);

Solution C — EXISTS
          SELECT o.*, c.first_name, c.last_name
          FROM orders o
          JOIN customers c
              ON o.customer_id = c.customer_id
          WHERE o.amount > 500;
===============================================
⭐ 5. Find customers aged >25 who placed at least one order

Solution A — JOIN
          SELECT DISTINCT c.*
          FROM customers c
          JOIN orders o
              ON c.customer_id = o.customer_id
          WHERE c.age > 25;

Solution B — EXISTS
        SELECT *
        FROM customers c
        WHERE c.age > 25
          AND EXISTS (
                SELECT 1
                FROM orders o
                WHERE o.customer_id = c.customer_id
          );

Solution C — IN
        SELECT *
        FROM customers
        WHERE age > 25
          AND customer_id IN (SELECT customer_id FROM orders);
=========================================================
⭐ 6. Find total amount spent by each customer

Solution A — GROUP BY

        SELECT
            customer_id,
            SUM(amount) AS total_spent
        FROM orders
        GROUP BY customer_id;

Solution B — JOIN + GROUP BY
        SELECT
            c.customer_id,
            c.first_name,
            c.last_name,
            SUM(o.amount) AS total_spent
        FROM customers c
        JOIN orders o
            ON c.customer_id = o.customer_id
        GROUP BY c.customer_id, c.first_name, c.last_name;

Solution C — Window function
          SELECT DISTINCT
              customer_id,
              SUM(amount) OVER (PARTITION BY customer_id) AS total_spent
          FROM orders;
=====================================================
⭐ 7. Find the customer who spent the highest total amount

Solution A — ORDER BY + LIMIT

          SELECT
              customer_id,
              SUM(amount) AS total_spent
          FROM orders
          GROUP BY customer_id
          ORDER BY total_spent DESC
          LIMIT 1;

Solution B — Window function
            SELECT customer_id, total_spent
            FROM (
                SELECT
                    customer_id,
                    SUM(amount) AS total_spent,
                    RANK() OVER (ORDER BY SUM(amount) DESC) AS rnk
                FROM orders
                GROUP BY customer_id
            ) t
            WHERE rnk = 1;

Solution C — MAX + HAVING
              SELECT customer_id, SUM(amount) AS total_spent
              FROM orders
              GROUP BY customer_id
              HAVING SUM(amount) = (
                  SELECT MAX(total_spent)
                  FROM (
                      SELECT SUM(amount) AS total_spent
                      FROM orders
                      GROUP BY customer_id
                  ) x
              );
=========================================
⭐ 8. Get avg order value per country

Solution A — GROUP BY

          SELECT
              country,
              AVG(amount) AS avg_order_value
          FROM orders
          GROUP BY country;

Solution B — JOIN customers
          SELECT
              c.country,
              AVG(o.amount) AS avg_order_value
          FROM customers c
          JOIN orders o
              ON c.customer_id = o.customer_id
          GROUP BY c.country;

Solution C — Window function
          SELECT DISTINCT
              country,
              AVG(amount) OVER (PARTITION BY country) AS avg_order_value
          FROM orders;
===================================================
⭐ 9. Count number of orders per country

Solution A — GROUP BY

          SELECT
              country,
              COUNT(*) AS order_count
          FROM orders
          GROUP BY country;

Solution B — JOIN customers
          SELECT
              c.country,
              COUNT(o.order_id) AS order_count
          FROM customers c
          JOIN orders o
              ON c.customer_id = o.customer_id
          GROUP BY c.country;

Solution C — Window function
      SELECT DISTINCT
          country,
          COUNT(order_id) OVER (PARTITION BY country) AS order_count
      FROM orders;
```

```python
##⭐ 4. Find customers who have not placed any orders but have a shipping record

Assume tables:

        customers(customer_id, first_name, …)

        orders(order_id, customer_id, …)

        shipments(shipment_id, order_id, status, …)

✅ Solution 1 — LEFT JOIN (classic interview style)
sql

      SELECT DISTINCT c.customer_id, c.first_name, c.last_name
      FROM customers c
      LEFT JOIN orders o ON c.customer_id = o.customer_id
      JOIN shipments s ON s.order_id = o.order_id
      WHERE o.order_id IS NULL;

Why it works
      LEFT JOIN ensures all customers appear

      o.order_id IS NULL → no orders

      But they still appear in shipments → means shipments exist for some order_id linked to them

✅ Solution 2 — NOT EXISTS (cleaner logic)
sql
      SELECT DISTINCT c.customer_id, c.first_name, c.last_name
      FROM customers c
      JOIN shipments s ON c.customer_id = s.customer_id
      WHERE NOT EXISTS (
          SELECT 1 FROM orders o
          WHERE o.customer_id = c.customer_id
      );

✅ Solution 3 — Anti-join using NOT IN
sql
        SELECT DISTINCT c.customer_id, c.first_name, c.last_name
        FROM customers c
        JOIN shipments s ON c.customer_id = s.customer_id
        WHERE c.customer_id NOT IN (
            SELECT customer_id FROM orders
        );

⭐ 5. Find customers from USA who placed orders that are still pending
Assume shipment status = 'Pending'.

✅ Solution 1 — JOIN all three tables
sql
        SELECT DISTINCT c.customer_id, c.first_name, c.last_name
        FROM customers c
        JOIN orders o ON c.customer_id = o.customer_id
        JOIN shipments s ON o.order_id = s.order_id
        WHERE c.country = 'USA'
          AND s.status = 'Pending';

✅ Solution 2 — Using EXISTS
sql
        SELECT c.customer_id, c.first_name, c.last_name
        FROM customers c
        WHERE c.country = 'USA'
          AND EXISTS (
                SELECT 1
                FROM orders o
                JOIN shipments s ON o.order_id = s.order_id
                WHERE o.customer_id = c.customer_id
                  AND s.status = 'Pending'
          );

✅ Solution 3 — Using subquery for pending orders
sql
        SELECT c.customer_id, c.first_name, c.last_name
        FROM customers c
        WHERE c.country = 'USA'
          AND c.customer_id IN (
                SELECT o.customer_id
                FROM orders o
                JOIN shipments s ON o.order_id = s.order_id
                WHERE s.status = 'Pending'
          );

⭐ 6. List all customers with total number of orders and their latest shipping status
Assume “latest” = max(shipment_date).

✅ Solution 1 — GROUP BY + window function
sql
        SELECT
            c.customer_id,
            c.first_name,
            c.last_name,
            COUNT(o.order_id) AS total_orders,
            s.status AS latest_status
        FROM customers c
        LEFT JOIN orders o ON c.customer_id = o.customer_id
        LEFT JOIN (
            SELECT
                order_id,
                status,
                ROW_NUMBER() OVER (PARTITION BY order_id ORDER BY shipment_date DESC) AS rn
            FROM shipments
        ) s ON o.order_id = s.order_id AND s.rn = 1
        GROUP BY c.customer_id, c.first_name, c.last_name, s.status;

✅ Solution 2 — Correlated subquery for latest shipment
sql
          SELECT
              c.customer_id,
              c.first_name,
              c.last_name,
              (SELECT COUNT(*) FROM orders o WHERE o.customer_id = c.customer_id) AS total_orders,
              (
                  SELECT s.status
                  FROM shipments s
                  JOIN orders o ON o.order_id = s.order_id
                  WHERE o.customer_id = c.customer_id
                  ORDER BY s.shipment_date DESC
                  LIMIT 1
              ) AS latest_status
          FROM customers c;

✅ Solution 3 — Two subqueries (clean teaching version)
sql
          SELECT
              c.customer_id,
              c.first_name,
              c.last_name,
              o.total_orders,
              s.latest_status
          FROM customers c
          LEFT JOIN (
              SELECT customer_id, COUNT(*) AS total_orders
              FROM orders
              GROUP BY customer_id
          ) o ON c.customer_id = o.customer_id
          LEFT JOIN (
              SELECT o.customer_id,
                    s.status AS latest_status
              FROM shipments s
              JOIN orders o ON s.order_id = o.order_id
              WHERE (s.order_id, s.shipment_date) IN (
                  SELECT order_id, MAX(shipment_date)
                  FROM shipments
                  GROUP BY order_id
              )
          ) s ON c.customer_id = s.customer_id;
```

```python
##✅ 7. Total amount spent by each customer whose shipments are delivered
sql
        SELECT
            c.customer_id,
            c.customer_name,
            SUM(o.order_amount) AS total_spent
        FROM customers c
        JOIN orders o
            ON c.customer_id = o.customer_id
        JOIN shipments s
            ON o.order_id = s.order_id
        WHERE s.shipment_status = 'Delivered'
        GROUP BY c.customer_id, c.customer_name;

✅ 8. Customer who spent the highest total amount
sql
        SELECT customer_id, customer_name, total_spent
        FROM (
            SELECT
                c.customer_id,
                c.customer_name,
                SUM(o.order_amount) AS total_spent,
                RANK() OVER (ORDER BY SUM(o.order_amount) DESC) AS rnk
            FROM customers c
            JOIN orders o
                ON c.customer_id = o.customer_id
            GROUP BY c.customer_id, c.customer_name
        ) t
        WHERE rnk = 1;

✅ 9. Count number of orders per country
sql
        SELECT
            c.country,
            COUNT(o.order_id) AS order_count
        FROM customers c
        JOIN orders o
            ON c.customer_id = o.customer_id
        GROUP BY c.country;

✅ 10. Total revenue from pending shipments
sql
        SELECT
            SUM(o.order_amount) AS pending_revenue
        FROM orders o
        JOIN shipments s
            ON o.order_id = s.order_id
        WHERE s.shipment_status = 'Pending';

✅ 11. Count of customers per shipping station
This assumes a customer is counted once per station based on their orders shipped from that station.

sql
        SELECT
            s.shipping_station,
            COUNT(DISTINCT o.customer_id) AS customer_count
        FROM shipments s
        JOIN orders o
            ON s.order_id = o.order_id
        GROUP BY s.shipping_station;

✅ 12. Country contributing the highest delivered revenue
sql
        SELECT country, delivered_revenue
        FROM (
            SELECT
                c.country,
                SUM(o.order_amount) AS delivered_revenue,
                RANK() OVER (ORDER BY SUM(o.order_amount) DESC) AS rnk
            FROM customers c
            JOIN orders o
                ON c.customer_id = o.customer_id
            JOIN shipments s
                ON o.order_id = s.order_id
            WHERE s.shipment_status = 'Delivered'
            GROUP BY c.country
        ) t
        WHERE rnk = 1;
```

```python
🚀 Overall Performance Principles

Fastest → Slowest (general rule)
        JOIN + WHERE

        EXISTS / NOT EXISTS

        IN / NOT IN

        LEFT JOIN + NULL filter

        EXCEPT / INTERSECT (depends on engine, often slower)

Why?
        JOINs use indexed paths efficiently.
        EXISTS stops scanning early.
        IN/NOT IN may materialize sets.
        LEFT JOIN + NULL requires scanning both sides.
        EXCEPT sorts/deduplicates → expensive.

⭐ Query 1: Get customers + items + amount + shipping status
✔ Best Performance
JOIN + LEFT JOIN (Solution A/B)
sql
        SELECT ...
        FROM customers c
        JOIN orders o ON c.customer_id = o.customer_id
        LEFT JOIN shippings s ON c.customer_id = s.customer_id;

Why it’s fastest
        Uses direct indexed joins on primary/foreign keys.
        LEFT JOIN is cheap because it’s on customer_id (likely indexed).
        No subqueries, no deduplication.

❌ Less efficient
      Using INNER JOIN on shippings (Solution C) is not slower,
      but it drops customers without shipping, which may not be desired.

⭐ Query 2: Customers with orders AND shipping status = 'Delivered'
✔ Best Performance
JOIN + WHERE (Solution A)
sql
        SELECT DISTINCT ...
        FROM customers c
        JOIN orders o ON c.customer_id = o.customer_id
        JOIN shippings s ON c.customer_id = s.customer_id
        WHERE s.status = 'Delivered';
Why?
      Straight join + filter → uses indexes efficiently.
      DISTINCT is cheap because customer_id is small.

✔ Also excellent
EXISTS (Solution B)
sql
        WHERE EXISTS (
            SELECT 1 ...
        )

Why?
      EXISTS stops scanning as soon as it finds a match.
      Very efficient when customers have many orders.

❌ Less efficient
    IN (Solution C)
    Materializes a list → slower than EXISTS.

⭐ Query 3: Customers with orders but NO shipping record
✔ Best Performance
    NOT EXISTS (Solution B)

sql
      WHERE NOT EXISTS (
          SELECT 1
          FROM shippings s
          WHERE s.customer_id = c.customer_id
      );
Why?
NOT EXISTS is the gold standard for anti‑joins.
Stops early.
Avoids NULL pitfalls.

✔ Also good
LEFT JOIN + NULL (Solution A)

sql
      LEFT JOIN shippings s ...
      WHERE s.customer_id IS NULL;
Why?
Very readable.
Optimizer often rewrites it into NOT EXISTS internally.

❌ Slowest
NOT IN (Solution C)

sql
    c.customer_id NOT IN (SELECT customer_id FROM shippings)

Why?
NOT IN + NULLs = tricky semantics.
Requires full scan of the subquery.
Often forces the engine to materialize the set.

🏆 Final Ranking Summary

Query	Fastest Approach	Why
1. Customers + items + shipping	JOIN + LEFT JOIN	Direct indexed joins, no subqueries
2. Delivered customers	JOIN + WHERE or EXISTS	Both use indexes efficiently; EXISTS short‑circuits
3. Orders but no shipping	NOT EXISTS	Best anti‑join pattern, avoids NULL issues
```
