```python
"""
COMPLETE PYTHON PANDAS SOLUTIONS FOR ALL LEETCODE SQL PROBLEMS
"""

import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# =====================================================
# 1. TOP TRAVELLERS (LeetCode 1407)
# =====================================================
def top_travellers(users: pd.DataFrame, rides: pd.DataFrame) -> pd.DataFrame:
    result = users.merge(rides, left_on='id', right_on='user_id', how='left')
    result = result.groupby(['id', 'name'])['distance'].sum().fillna(0).reset_index()
    result.columns = ['id', 'name', 'travelled_distance']
    return result.sort_values(['travelled_distance', 'name'],
                              ascending=[False, True])[['name', 'travelled_distance']]


# =====================================================
# 2. PATIENTS WITH A CONDITION (LeetCode 1527)
# =====================================================
def patients_with_condition(patients: pd.DataFrame) -> pd.DataFrame:
    # Match DIAB1 at the beginning or after a space
    return patients[patients['conditions'].str.contains(r'\bDIAB1', regex=True, na=False)]


# =====================================================
# 3. FIX NAMES IN A TABLE (LeetCode 1667)
# =====================================================
def fix_names(users: pd.DataFrame) -> pd.DataFrame:
    users['name'] = users['name'].str.capitalize()
    return users.sort_values('user_id')[['user_id', 'name']]


# =====================================================
# 4. REFORMAT DEPARTMENT TABLE (LeetCode 1179)
# =====================================================
def reformat_table(department: pd.DataFrame) -> pd.DataFrame:
    result = department.pivot(index='id', columns='month', values='revenue')
    result.columns = [col + '_Revenue' for col in result.columns]
    return result.reset_index()


# =====================================================
# 5. SALES ANALYSIS III (LeetCode 1084)
# =====================================================
def sales_analysis(product: pd.DataFrame, sales: pd.DataFrame) -> pd.DataFrame:
    # Filter sales in Q1 2019
    sales_filtered = sales[
        (sales['sale_date'] >= '2019-01-01') &
        (sales['sale_date'] <= '2019-03-31')
    ]

    # Get products sold only in Q1 2019
    all_sales = sales.groupby('product_id')['sale_date'].agg(['min', 'max'])
    valid_products = all_sales[
        (all_sales['min'] >= '2019-01-01') &
        (all_sales['max'] <= '2019-03-31')
    ]

    return product[product['product_id'].isin(valid_products.index)][['product_id', 'product_name']]


# =====================================================
# 6. GAME PLAY ANALYSIS I (LeetCode 511)
# =====================================================
def game_analysis(activity: pd.DataFrame) -> pd.DataFrame:
    result = activity.groupby('player_id')['event_date'].min().reset_index()
    result.columns = ['player_id', 'first_login']
    return result


# =====================================================
# 7. REARRANGE PRODUCTS TABLE (LeetCode 1795)
# =====================================================
def rearrange_products(products: pd.DataFrame) -> pd.DataFrame:
    # Unpivot the table
    result = products.melt(
        id_vars='product_id',
        value_vars=['store1', 'store2', 'store3'],
        var_name='store',
        value_name='price'
    )
    return result.dropna().reset_index(drop=True)


# =====================================================
# 8. DAILY LEADS AND PARTNERS (LeetCode 1693)
# =====================================================
def daily_leads_partners(daily_sales: pd.DataFrame) -> pd.DataFrame:
    result = daily_sales.groupby(['date_id', 'make_name']).agg(
        unique_leads=('lead_id', 'nunique'),
        unique_partners=('partner_id', 'nunique')
    ).reset_index()
    return result


# =====================================================
# 9. COMBINE TWO TABLES (LeetCode 175)
# =====================================================
def combine_tables(person: pd.DataFrame, address: pd.DataFrame) -> pd.DataFrame:
    result = person.merge(address, on='personId', how='left')
    return result[['firstName', 'lastName', 'city', 'state']]


# =====================================================
# 10. SECOND HIGHEST SALARY (LeetCode 176)
# =====================================================
def second_highest_salary(employee: pd.DataFrame) -> pd.DataFrame:
    unique_salaries = employee['salary'].drop_duplicates().sort_values(ascending=False)
    if len(unique_salaries) < 2:
        return pd.DataFrame({'SecondHighestSalary': [None]})
    return pd.DataFrame({'SecondHighestSalary': [unique_salaries.iloc[1]]})


# =====================================================
# 11. NTH HIGHEST SALARY (LeetCode 177)
# =====================================================
def nth_highest_salary(employee: pd.DataFrame, N: int) -> pd.DataFrame:
    unique_salaries = employee['salary'].drop_duplicates().sort_values(ascending=False)
    if N < 1 or len(unique_salaries) < N:
        return pd.DataFrame({f'getNthHighestSalary({N})': [None]})
    return pd.DataFrame({f'getNthHighestSalary({N})': [unique_salaries.iloc[N-1]]})


# =====================================================
# 12. RANK SCORES (LeetCode 178)
# =====================================================
def rank_scores(scores: pd.DataFrame) -> pd.DataFrame:
    scores['rank'] = scores['score'].rank(method='dense', ascending=False)
    return scores[['score', 'rank']].sort_values('score', ascending=False)


# =====================================================
# 13. CONSECUTIVE NUMBERS (LeetCode 180)
# =====================================================
def consecutive_numbers(logs: pd.DataFrame) -> pd.DataFrame:
    logs = logs.sort_values('id')
    logs['next_num'] = logs['num'].shift(-1)
    logs['next_next_num'] = logs['num'].shift(-2)

    result = logs[
        (logs['num'] == logs['next_num']) &
        (logs['num'] == logs['next_next_num'])
    ]['num'].drop_duplicates()

    return pd.DataFrame({'ConsecutiveNums': result.values})


# =====================================================
# 14. EMPLOYEES EARNING MORE THAN THEIR MANAGERS (LeetCode 181)
# =====================================================
def employees_earning_more(employee: pd.DataFrame) -> pd.DataFrame:
    merged = employee.merge(
        employee,
        left_on='managerId',
        right_on='id',
        suffixes=('', '_manager')
    )
    result = merged[merged['salary'] > merged['salary_manager']]
    return result[['name']].rename(columns={'name': 'Employee'})


# =====================================================
# 15. DUPLICATE EMAILS (LeetCode 182)
# =====================================================
def duplicate_emails(person: pd.DataFrame) -> pd.DataFrame:
    email_counts = person.groupby('email').size().reset_index(name='count')
    duplicates = email_counts[email_counts['count'] > 1]
    return duplicates[['email']]


# =====================================================
# 16. CUSTOMERS WHO NEVER ORDER (LeetCode 183)
# =====================================================
def customers_never_order(customers: pd.DataFrame, orders: pd.DataFrame) -> pd.DataFrame:
    result = customers[~customers['id'].isin(orders['customerId'])]
    return result[['name']].rename(columns={'name': 'Customers'})


# =====================================================
# 17. DEPARTMENT HIGHEST SALARY (LeetCode 184)
# =====================================================
def department_highest_salary(employee: pd.DataFrame, department: pd.DataFrame) -> pd.DataFrame:
    # Get max salary per department
    max_salaries = employee.groupby('departmentId')['salary'].max().reset_index()

    # Join to get employees with max salary
    result = employee.merge(max_salaries, on=['departmentId', 'salary'])
    result = result.merge(department, left_on='departmentId', right_on='id')

    return result[['name_y', 'name_x', 'salary']].rename(
        columns={'name_y': 'Department', 'name_x': 'Employee', 'salary': 'Salary'}
    )


# =====================================================
# 18. DEPARTMENT TOP THREE SALARIES (LeetCode 185)
# =====================================================
def top_three_salaries(employee: pd.DataFrame, department: pd.DataFrame) -> pd.DataFrame:
    # Rank salaries within each department
    employee['rank'] = employee.groupby('departmentId')['salary'].rank(
        method='dense', ascending=False
    )

    # Filter top 3 and join with department
    top_three = employee[employee['rank'] <= 3]
    result = top_three.merge(department, left_on='departmentId', right_on='id')

    return result[['name_y', 'name_x', 'salary']].rename(
        columns={'name_y': 'Department', 'name_x': 'Employee', 'salary': 'Salary'}
    )


# =====================================================
# 19. CUSTOMERS WHO VISITED BUT DID NOT MAKE ANY TRANSACTIONS (LeetCode 1581)
# =====================================================
def no_transactions(visits: pd.DataFrame, transactions: pd.DataFrame) -> pd.DataFrame:
    merged = visits.merge(transactions, on='visit_id', how='left')
    no_trans = merged[merged['transaction_id'].isna()]
    result = no_trans.groupby('customer_id').size().reset_index(name='count_no_trans')
    return result


# =====================================================
# 20. EMPLOYEES WITH MISSING INFORMATION (LeetCode 1965)
# =====================================================
def missing_information(employees: pd.DataFrame, salaries: pd.DataFrame) -> pd.DataFrame:
    # Full outer join
    merged = employees.merge(salaries, on='employee_id', how='outer')
    # Find missing information
    missing = merged[(merged['name'].isna()) | (merged['salary'].isna())]
    return missing[['employee_id']].sort_values('employee_id')


# =====================================================
# 21. DELETE DUPLICATE EMAILS (LeetCode 196)
# =====================================================
def delete_duplicate_emails(person: pd.DataFrame) -> pd.DataFrame:
    # Keep first occurrence (smallest id) for each email
    person.sort_values('id', inplace=True)
    person.drop_duplicates(subset='email', keep='first', inplace=True)
    return person


# =====================================================
# 22. RISING TEMPERATURE (LeetCode 197)
# =====================================================
def rising_temperature(weather: pd.DataFrame) -> pd.DataFrame:
    weather = weather.sort_values('recordDate').reset_index(drop=True)
    weather['prev_temp'] = weather['temperature'].shift(1)
    weather['prev_date'] = weather['recordDate'].shift(1)

    # Calculate date difference
    weather['date_diff'] = (weather['recordDate'] - weather['prev_date']).dt.days

    result = weather[
        (weather['temperature'] > weather['prev_temp']) &
        (weather['date_diff'] == 1)
    ]
    return result[['id']]


# =====================================================
# 23. BANK ACCOUNT SUMMARY II (LeetCode 1587)
# =====================================================
def bank_account_summary(users: pd.DataFrame, transactions: pd.DataFrame) -> pd.DataFrame:
    # Calculate balance per account
    balances = transactions.groupby('account')['amount'].sum().reset_index()
    balances.columns = ['account', 'balance']

    # Filter accounts with balance > 10000
    high_balance = balances[balances['balance'] > 10000]

    # Join with users
    result = high_balance.merge(users, on='account')
    return result[['name', 'balance']]


# =====================================================
# 24. BIG COUNTRIES (LeetCode 595)
# =====================================================
def big_countries(world: pd.DataFrame) -> pd.DataFrame:
    result = world[(world['area'] >= 3000000) | (world['population'] >= 25000000)]
    return result[['name', 'population', 'area']]


# =====================================================
# 25. CLASSES WITH AT LEAST 5 STUDENTS (LeetCode 596)
# =====================================================
def classes_at_least_5(courses: pd.DataFrame) -> pd.DataFrame:
    class_counts = courses.groupby('class').size().reset_index(name='count')
    result = class_counts[class_counts['count'] >= 5]
    return result[['class']]


# =====================================================
# 26. FIND FOLLOWERS COUNT (LeetCode 1729)
# =====================================================
def followers_count(followers: pd.DataFrame) -> pd.DataFrame:
    result = followers.groupby('user_id')['follower_id'].count().reset_index()
    result.columns = ['user_id', 'followers_count']
    return result.sort_values('user_id')


# =====================================================
# 27. GROUP SOLD PRODUCTS BY THE DATE (LeetCode 1484)
# =====================================================
def group_sold_products(activities: pd.DataFrame) -> pd.DataFrame:
    result = activities.groupby('sell_date').agg(
        num_sold=('product', 'nunique'),
        products=('product', lambda x: ','.join(sorted(x.unique())))
    ).reset_index()
    return result.sort_values('sell_date')


# =====================================================
# 28. USER ACTIVITY FOR THE PAST 30 DAYS I (LeetCode 1141)
# =====================================================
def user_activity_30_days(activity: pd.DataFrame) -> pd.DataFrame:
    end_date = pd.to_datetime('2019-07-27')
    start_date = end_date - pd.Timedelta(days=29)

    filtered = activity[
        (activity['activity_date'] >= start_date) &
        (activity['activity_date'] <= end_date)
    ]

    result = filtered.groupby('activity_date')['user_id'].nunique().reset_index()
    result.columns = ['day', 'active_users']
    return result


# =====================================================
# 29. SALES PERSON (LeetCode 607)
# =====================================================
def sales_person(sales_person: pd.DataFrame, company: pd.DataFrame,
                 orders: pd.DataFrame) -> pd.DataFrame:
    # Find sales_id who sold to RED company
    red_company_id = company[company['name'] == 'RED']['com_id'].values
    red_sales = orders[orders['com_id'].isin(red_company_id)]['sales_id'].unique()

    # Find sales persons who never sold to RED
    result = sales_person[~sales_person['sales_id'].isin(red_sales)]
    return result[['name']]


# =====================================================
# 30. TREE NODE (LeetCode 608)
# =====================================================
def tree_node(tree: pd.DataFrame) -> pd.DataFrame:
    # Find all parent ids
    parent_ids = tree['p_id'].dropna().unique()

    def get_type(row):
        if pd.isna(row['p_id']):
            return 'Root'
        elif row['id'] in parent_ids:
            return 'Inner'
        else:
            return 'Leaf'

    tree['type'] = tree.apply(get_type, axis=1)
    return tree[['id', 'type']]


# =====================================================
# 31. FIND TOTAL TIME SPENT BY EACH EMPLOYEE (LeetCode 1741)
# =====================================================
def total_time(employees: pd.DataFrame) -> pd.DataFrame:
    employees['total_time'] = employees['out_time'] - employees['in_time']
    result = employees.groupby(['event_day', 'emp_id'])['total_time'].sum().reset_index()
    result.columns = ['day', 'emp_id', 'total_time']
    return result


# =====================================================
# 32. CALCULATE SPECIAL BONUS (LeetCode 1873)
# =====================================================
def calculate_bonus(employees: pd.DataFrame) -> pd.DataFrame:
    def get_bonus(row):
        if row['employee_id'] % 2 == 1 and not row['name'].startswith('M'):
            return row['salary']
        return 0

    employees['bonus'] = employees.apply(get_bonus, axis=1)
    return employees[['employee_id', 'bonus']].sort_values('employee_id')


# =====================================================
# 33. ARTICLE VIEWS I (LeetCode 1148)
# =====================================================
def article_views(views: pd.DataFrame) -> pd.DataFrame:
    result = views[views['author_id'] == views['viewer_id']]['author_id'].unique()
    return pd.DataFrame({'id': sorted(result)})


# =====================================================
# 34. NOT BORING MOVIES (LeetCode 620)
# =====================================================
def not_boring_movies(cinema: pd.DataFrame) -> pd.DataFrame:
    result = cinema[(cinema['id'] % 2 == 1) & (cinema['description'] != 'boring')]
    return result.sort_values('rating', ascending=False)


# =====================================================
# 35. ACTORS AND DIRECTORS WHO COOPERATED AT LEAST THREE TIMES (LeetCode 1050)
# =====================================================
def actors_directors(actor_director: pd.DataFrame) -> pd.DataFrame:
    cooperations = actor_director.groupby(['actor_id', 'director_id']).size().reset_index(name='count')
    result = cooperations[cooperations['count'] >= 3]
    return result[['actor_id', 'director_id']]


# =====================================================
# 36. EXCHANGE SEATS (LeetCode 626)
# =====================================================
def exchange_seats(seat: pd.DataFrame) -> pd.DataFrame:
    max_id = seat['id'].max()

    def swap_id(row):
        if row['id'] % 2 == 1 and row['id'] == max_id:
            return row['id']
        elif row['id'] % 2 == 1:
            return row['id'] + 1
        else:
            return row['id'] - 1

    seat['new_id'] = seat.apply(swap_id, axis=1)
    result = seat.sort_values('new_id')[['new_id', 'student']]
    result.columns = ['id', 'student']
    return result


# =====================================================
# 37. SWAP SALARY (LeetCode 627)
# =====================================================
def swap_salary(salary: pd.DataFrame) -> pd.DataFrame:
    salary['sex'] = salary['sex'].map({'m': 'f', 'f': 'm'})
    return salary


# =====================================================
# 38. MARKET ANALYSIS I (LeetCode 1158)
# =====================================================
def market_analysis(users: pd.DataFrame, orders: pd.DataFrame,
                    items: pd.DataFrame) -> pd.DataFrame:
    # Filter orders from 2019
    orders['year'] = pd.to_datetime(orders['order_date']).dt.year
    orders_2019 = orders[orders['year'] == 2019]

    # Count orders per buyer
    order_counts = orders_2019.groupby('buyer_id').size().reset_index(name='orders_in_2019')

    # Left join with users
    result = users.merge(order_counts, left_on='user_id', right_on='buyer_id', how='left')
    result['orders_in_2019'] = result['orders_in_2019'].fillna(0).astype(int)

    return result[['user_id', 'join_date', 'orders_in_2019']].rename(
        columns={'user_id': 'buyer_id'}
    )


# =====================================================
# 39. CAPITAL GAIN/LOSS (LeetCode 1393)
# =====================================================
def capital_gain_loss(stocks: pd.DataFrame) -> pd.DataFrame:
    # Apply negative sign to Buy operations
    stocks['amount'] = stocks.apply(
        lambda row: -row['price'] if row['operation'] == 'Buy' else row['price'],
        axis=1
    )

    result = stocks.groupby('stock_name')['amount'].sum().reset_index()
    result.columns = ['stock_name', 'capital_gain_loss']
    return result


# =====================================================
# 40. RECYCLABLE AND LOW FAT PRODUCTS (LeetCode 1757)
# =====================================================
def recyclable_low_fat(products: pd.DataFrame) -> pd.DataFrame:
    result = products[(products['low_fats'] == 'Y') & (products['recyclable'] == 'Y')]
    return result[['product_id']]


# =====================================================
# 41. THE LATEST LOGIN IN 2020 (LeetCode 1890)
# =====================================================
def latest_login(logins: pd.DataFrame) -> pd.DataFrame:
    # Extract year and filter for 2020
    logins['year'] = pd.to_datetime(logins['time_stamp']).dt.year
    logins_2020 = logins[logins['year'] == 2020]

    # Get max timestamp per user
    result = logins_2020.groupby('user_id')['time_stamp'].max().reset_index()
    result.columns = ['user_id', 'last_stamp']
    return result


# =====================================================
# EXAMPLE USAGE
# =====================================================
if __name__ == "__main__":
    # Example 1: Top Travellers
    users = pd.DataFrame({
        'id': [1, 2, 3],
        'name': ['Alice', 'Bob', 'Charlie']
    })
    rides = pd.DataFrame({
        'id': [1, 2, 3],
        'user_id': [1, 1, 2],
        'distance': [120, 150, 200]
    })
    print("Top Travellers:")
    print(top_travellers(users, rides))
    print()

    # Example 2: Big Countries
    world = pd.DataFrame({
        'name': ['Afghanistan', 'Albania', 'Algeria', 'Russia'],
        'population': [25500100, 2831741, 37100000, 146793744],
        'area': [652230, 28748, 2381741, 17098242]
    })
    print("Big Countries:")
    print(big_countries(world))
    print()

    # Example 3: Second Highest Salary
    employee = pd.DataFrame({
        'id': [1, 2, 3],
        'salary': [100, 200, 300]
    })
    print("Second Highest Salary:")
    print(second_highest_salary(employee))
```

```python
"""
COMPLETE PYTHON PANDAS SOLUTIONS FOR ALL LEETCODE SQL PROBLEMS
"""

import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# =====================================================
# 1. TOP TRAVELLERS (LeetCode 1407)
# =====================================================
def top_travellers(users: pd.DataFrame, rides: pd.DataFrame) -> pd.DataFrame:
    result = users.merge(rides, left_on='id', right_on='user_id', how='left')
    result = result.groupby(['id', 'name'])['distance'].sum().fillna(0).reset_index()
    result.columns = ['id', 'name', 'travelled_distance']
    return result.sort_values(['travelled_distance', 'name'],
                              ascending=[False, True])[['name', 'travelled_distance']]


# =====================================================
# 2. PATIENTS WITH A CONDITION (LeetCode 1527)
# =====================================================
def patients_with_condition(patients: pd.DataFrame) -> pd.DataFrame:
    # Match DIAB1 at the beginning or after a space
    return patients[patients['conditions'].str.contains(r'\bDIAB1', regex=True, na=False)]


# =====================================================
# 3. FIX NAMES IN A TABLE (LeetCode 1667)
# =====================================================
def fix_names(users: pd.DataFrame) -> pd.DataFrame:
    users['name'] = users['name'].str.capitalize()
    return users.sort_values('user_id')[['user_id', 'name']]


# =====================================================
# 4. REFORMAT DEPARTMENT TABLE (LeetCode 1179)
# =====================================================
def reformat_table(department: pd.DataFrame) -> pd.DataFrame:
    result = department.pivot(index='id', columns='month', values='revenue')
    result.columns = [col + '_Revenue' for col in result.columns]
    return result.reset_index()


# =====================================================
# 5. SALES ANALYSIS III (LeetCode 1084)
# =====================================================
def sales_analysis(product: pd.DataFrame, sales: pd.DataFrame) -> pd.DataFrame:
    # Filter sales in Q1 2019
    sales_filtered = sales[
        (sales['sale_date'] >= '2019-01-01') &
        (sales['sale_date'] <= '2019-03-31')
    ]

    # Get products sold only in Q1 2019
    all_sales = sales.groupby('product_id')['sale_date'].agg(['min', 'max'])
    valid_products = all_sales[
        (all_sales['min'] >= '2019-01-01') &
        (all_sales['max'] <= '2019-03-31')
    ]

    return product[product['product_id'].isin(valid_products.index)][['product_id', 'product_name']]


# =====================================================
# 6. GAME PLAY ANALYSIS I (LeetCode 511)
# =====================================================
def game_analysis(activity: pd.DataFrame) -> pd.DataFrame:
    result = activity.groupby('player_id')['event_date'].min().reset_index()
    result.columns = ['player_id', 'first_login']
    return result


# =====================================================
# 7. REARRANGE PRODUCTS TABLE (LeetCode 1795)
# =====================================================
def rearrange_products(products: pd.DataFrame) -> pd.DataFrame:
    # Unpivot the table
    result = products.melt(
        id_vars='product_id',
        value_vars=['store1', 'store2', 'store3'],
        var_name='store',
        value_name='price'
    )
    return result.dropna().reset_index(drop=True)


# =====================================================
# 8. DAILY LEADS AND PARTNERS (LeetCode 1693)
# =====================================================
def daily_leads_partners(daily_sales: pd.DataFrame) -> pd.DataFrame:
    result = daily_sales.groupby(['date_id', 'make_name']).agg(
        unique_leads=('lead_id', 'nunique'),
        unique_partners=('partner_id', 'nunique')
    ).reset_index()
    return result


# =====================================================
# 9. COMBINE TWO TABLES (LeetCode 175)
# =====================================================
def combine_tables(person: pd.DataFrame, address: pd.DataFrame) -> pd.DataFrame:
    result = person.merge(address, on='personId', how='left')
    return result[['firstName', 'lastName', 'city', 'state']]


# =====================================================
# 10. SECOND HIGHEST SALARY (LeetCode 176)
# =====================================================
def second_highest_salary(employee: pd.DataFrame) -> pd.DataFrame:
    unique_salaries = employee['salary'].drop_duplicates().sort_values(ascending=False)
    if len(unique_salaries) < 2:
        return pd.DataFrame({'SecondHighestSalary': [None]})
    return pd.DataFrame({'SecondHighestSalary': [unique_salaries.iloc[1]]})


# =====================================================
# 11. NTH HIGHEST SALARY (LeetCode 177)
# =====================================================
def nth_highest_salary(employee: pd.DataFrame, N: int) -> pd.DataFrame:
    unique_salaries = employee['salary'].drop_duplicates().sort_values(ascending=False)
    if N < 1 or len(unique_salaries) < N:
        return pd.DataFrame({f'getNthHighestSalary({N})': [None]})
    return pd.DataFrame({f'getNthHighestSalary({N})': [unique_salaries.iloc[N-1]]})


# =====================================================
# 12. RANK SCORES (LeetCode 178)
# =====================================================
def rank_scores(scores: pd.DataFrame) -> pd.DataFrame:
    scores['rank'] = scores['score'].rank(method='dense', ascending=False)
    return scores[['score', 'rank']].sort_values('score', ascending=False)


# =====================================================
# 13. CONSECUTIVE NUMBERS (LeetCode 180)
# =====================================================
def consecutive_numbers(logs: pd.DataFrame) -> pd.DataFrame:
    logs = logs.sort_values('id')
    logs['next_num'] = logs['num'].shift(-1)
    logs['next_next_num'] = logs['num'].shift(-2)

    result = logs[
        (logs['num'] == logs['next_num']) &
        (logs['num'] == logs['next_next_num'])
    ]['num'].drop_duplicates()

    return pd.DataFrame({'ConsecutiveNums': result.values})


# =====================================================
# 14. EMPLOYEES EARNING MORE THAN THEIR MANAGERS (LeetCode 181)
# =====================================================
def employees_earning_more(employee: pd.DataFrame) -> pd.DataFrame:
    merged = employee.merge(
        employee,
        left_on='managerId',
        right_on='id',
        suffixes=('', '_manager')
    )
    result = merged[merged['salary'] > merged['salary_manager']]
    return result[['name']].rename(columns={'name': 'Employee'})


# =====================================================
# 15. DUPLICATE EMAILS (LeetCode 182)
# =====================================================
def duplicate_emails(person: pd.DataFrame) -> pd.DataFrame:
    email_counts = person.groupby('email').size().reset_index(name='count')
    duplicates = email_counts[email_counts['count'] > 1]
    return duplicates[['email']]


# =====================================================
# 16. CUSTOMERS WHO NEVER ORDER (LeetCode 183)
# =====================================================
def customers_never_order(customers: pd.DataFrame, orders: pd.DataFrame) -> pd.DataFrame:
    result = customers[~customers['id'].isin(orders['customerId'])]
    return result[['name']].rename(columns={'name': 'Customers'})


# =====================================================
# 17. DEPARTMENT HIGHEST SALARY (LeetCode 184)
# =====================================================
def department_highest_salary(employee: pd.DataFrame, department: pd.DataFrame) -> pd.DataFrame:
    # Get max salary per department
    max_salaries = employee.groupby('departmentId')['salary'].max().reset_index()

    # Join to get employees with max salary
    result = employee.merge(max_salaries, on=['departmentId', 'salary'])
    result = result.merge(department, left_on='departmentId', right_on='id')

    return result[['name_y', 'name_x', 'salary']].rename(
        columns={'name_y': 'Department', 'name_x': 'Employee', 'salary': 'Salary'}
    )


# =====================================================
# 18. DEPARTMENT TOP THREE SALARIES (LeetCode 185)
# =====================================================
def top_three_salaries(employee: pd.DataFrame, department: pd.DataFrame) -> pd.DataFrame:
    # Rank salaries within each department
    employee['rank'] = employee.groupby('departmentId')['salary'].rank(
        method='dense', ascending=False
    )

    # Filter top 3 and join with department
    top_three = employee[employee['rank'] <= 3]
    result = top_three.merge(department, left_on='departmentId', right_on='id')

    return result[['name_y', 'name_x', 'salary']].rename(
        columns={'name_y': 'Department', 'name_x': 'Employee', 'salary': 'Salary'}
    )


# =====================================================
# 19. CUSTOMERS WHO VISITED BUT DID NOT MAKE ANY TRANSACTIONS (LeetCode 1581)
# =====================================================
def no_transactions(visits: pd.DataFrame, transactions: pd.DataFrame) -> pd.DataFrame:
    merged = visits.merge(transactions, on='visit_id', how='left')
    no_trans = merged[merged['transaction_id'].isna()]
    result = no_trans.groupby('customer_id').size().reset_index(name='count_no_trans')
    return result


# =====================================================
# 20. EMPLOYEES WITH MISSING INFORMATION (LeetCode 1965)
# =====================================================
def missing_information(employees: pd.DataFrame, salaries: pd.DataFrame) -> pd.DataFrame:
    # Full outer join
    merged = employees.merge(salaries, on='employee_id', how='outer')
    # Find missing information
    missing = merged[(merged['name'].isna()) | (merged['salary'].isna())]
    return missing[['employee_id']].sort_values('employee_id')


# =====================================================
# 21. DELETE DUPLICATE EMAILS (LeetCode 196)
# =====================================================
def delete_duplicate_emails(person: pd.DataFrame) -> pd.DataFrame:
    # Keep first occurrence (smallest id) for each email
    person.sort_values('id', inplace=True)
    person.drop_duplicates(subset='email', keep='first', inplace=True)
    return person


# =====================================================
# 22. RISING TEMPERATURE (LeetCode 197)
# =====================================================
def rising_temperature(weather: pd.DataFrame) -> pd.DataFrame:
    weather = weather.sort_values('recordDate').reset_index(drop=True)
    weather['prev_temp'] = weather['temperature'].shift(1)
    weather['prev_date'] = weather['recordDate'].shift(1)

    # Calculate date difference
    weather['date_diff'] = (weather['recordDate'] - weather['prev_date']).dt.days

    result = weather[
        (weather['temperature'] > weather['prev_temp']) &
        (weather['date_diff'] == 1)
    ]
    return result[['id']]


# =====================================================
# 23. BANK ACCOUNT SUMMARY II (LeetCode 1587)
# =====================================================
def bank_account_summary(users: pd.DataFrame, transactions: pd.DataFrame) -> pd.DataFrame:
    # Calculate balance per account
    balances = transactions.groupby('account')['amount'].sum().reset_index()
    balances.columns = ['account', 'balance']

    # Filter accounts with balance > 10000
    high_balance = balances[balances['balance'] > 10000]

    # Join with users
    result = high_balance.merge(users, on='account')
    return result[['name', 'balance']]


# =====================================================
# 24. BIG COUNTRIES (LeetCode 595)
# =====================================================
def big_countries(world: pd.DataFrame) -> pd.DataFrame:
    result = world[(world['area'] >= 3000000) | (world['population'] >= 25000000)]
    return result[['name', 'population', 'area']]


# =====================================================
# 25. CLASSES WITH AT LEAST 5 STUDENTS (LeetCode 596)
# =====================================================
def classes_at_least_5(courses: pd.DataFrame) -> pd.DataFrame:
    class_counts = courses.groupby('class').size().reset_index(name='count')
    result = class_counts[class_counts['count'] >= 5]
    return result[['class']]


# =====================================================
# 26. FIND FOLLOWERS COUNT (LeetCode 1729)
# =====================================================
def followers_count(followers: pd.DataFrame) -> pd.DataFrame:
    result = followers.groupby('user_id')['follower_id'].count().reset_index()
    result.columns = ['user_id', 'followers_count']
    return result.sort_values('user_id')


# =====================================================
# 27. GROUP SOLD PRODUCTS BY THE DATE (LeetCode 1484)
# =====================================================
def group_sold_products(activities: pd.DataFrame) -> pd.DataFrame:
    result = activities.groupby('sell_date').agg(
        num_sold=('product', 'nunique'),
        products=('product', lambda x: ','.join(sorted(x.unique())))
    ).reset_index()
    return result.sort_values('sell_date')


# =====================================================
# 28. USER ACTIVITY FOR THE PAST 30 DAYS I (LeetCode 1141)
# =====================================================
def user_activity_30_days(activity: pd.DataFrame) -> pd.DataFrame:
    end_date = pd.to_datetime('2019-07-27')
    start_date = end_date - pd.Timedelta(days=29)

    filtered = activity[
        (activity['activity_date'] >= start_date) &
        (activity['activity_date'] <= end_date)
    ]

    result = filtered.groupby('activity_date')['user_id'].nunique().reset_index()
    result.columns = ['day', 'active_users']
    return result


# =====================================================
# 29. SALES PERSON (LeetCode 607)
# =====================================================
def sales_person(sales_person: pd.DataFrame, company: pd.DataFrame,
                 orders: pd.DataFrame) -> pd.DataFrame:
    # Find sales_id who sold to RED company
    red_company_id = company[company['name'] == 'RED']['com_id'].values
    red_sales = orders[orders['com_id'].isin(red_company_id)]['sales_id'].unique()

    # Find sales persons who never sold to RED
    result = sales_person[~sales_person['sales_id'].isin(red_sales)]
    return result[['name']]


# =====================================================
# 30. TREE NODE (LeetCode 608)
# =====================================================
def tree_node(tree: pd.DataFrame) -> pd.DataFrame:
    # Find all parent ids
    parent_ids = tree['p_id'].dropna().unique()

    def get_type(row):
        if pd.isna(row['p_id']):
            return 'Root'
        elif row['id'] in parent_ids:
            return 'Inner'
        else:
            return 'Leaf'

    tree['type'] = tree.apply(get_type, axis=1)
    return tree[['id', 'type']]


# =====================================================
# 31. FIND TOTAL TIME SPENT BY EACH EMPLOYEE (LeetCode 1741)
# =====================================================
def total_time(employees: pd.DataFrame) -> pd.DataFrame:
    employees['total_time'] = employees['out_time'] - employees['in_time']
    result = employees.groupby(['event_day', 'emp_id'])['total_time'].sum().reset_index()
    result.columns = ['day', 'emp_id', 'total_time']
    return result


# =====================================================
# 32. CALCULATE SPECIAL BONUS (LeetCode 1873)
# =====================================================
def calculate_bonus(employees: pd.DataFrame) -> pd.DataFrame:
    def get_bonus(row):
        if row['employee_id'] % 2 == 1 and not row['name'].startswith('M'):
            return row['salary']
        return 0

    employees['bonus'] = employees.apply(get_bonus, axis=1)
    return employees[['employee_id', 'bonus']].sort_values('employee_id')


# =====================================================
# 33. ARTICLE VIEWS I (LeetCode 1148)
# =====================================================
def article_views(views: pd.DataFrame) -> pd.DataFrame:
    result = views[views['author_id'] == views['viewer_id']]['author_id'].unique()
    return pd.DataFrame({'id': sorted(result)})


# =====================================================
# 34. NOT BORING MOVIES (LeetCode 620)
# =====================================================
def not_boring_movies(cinema: pd.DataFrame) -> pd.DataFrame:
    result = cinema[(cinema['id'] % 2 == 1) & (cinema['description'] != 'boring')]
    return result.sort_values('rating', ascending=False)


# =====================================================
# 35. ACTORS AND DIRECTORS WHO COOPERATED AT LEAST THREE TIMES (LeetCode 1050)
# =====================================================
def actors_directors(actor_director: pd.DataFrame) -> pd.DataFrame:
    cooperations = actor_director.groupby(['actor_id', 'director_id']).size().reset_index(name='count')
    result = cooperations[cooperations['count'] >= 3]
    return result[['actor_id', 'director_id']]


# =====================================================
# 36. EXCHANGE SEATS (LeetCode 626)
# =====================================================
def exchange_seats(seat: pd.DataFrame) -> pd.DataFrame:
    max_id = seat['id'].max()

    def swap_id(row):
        if row['id'] % 2 == 1 and row['id'] == max_id:
            return row['id']
        elif row['id'] % 2 == 1:
            return row['id'] + 1
        else:
            return row['id'] - 1

    seat['new_id'] = seat.apply(swap_id, axis=1)
    result = seat.sort_values('new_id')[['new_id', 'student']]
    result.columns = ['id', 'student']
    return result


# =====================================================
# 37. SWAP SALARY (LeetCode 627)
# =====================================================
def swap_salary(salary: pd.DataFrame) -> pd.DataFrame:
    salary['sex'] = salary['sex'].map({'m': 'f', 'f': 'm'})
    return salary


# =====================================================
# 38. MARKET ANALYSIS I (LeetCode 1158)
# =====================================================
def market_analysis(users: pd.DataFrame, orders: pd.DataFrame,
                    items: pd.DataFrame) -> pd.DataFrame:
    # Filter orders from 2019
    orders['year'] = pd.to_datetime(orders['order_date']).dt.year
    orders_2019 = orders[orders['year'] == 2019]

    # Count orders per buyer
    order_counts = orders_2019.groupby('buyer_id').size().reset_index(name='orders_in_2019')

    # Left join with users
    result = users.merge(order_counts, left_on='user_id', right_on='buyer_id', how='left')
    result['orders_in_2019'] = result['orders_in_2019'].fillna(0).astype(int)

    return result[['user_id', 'join_date', 'orders_in_2019']].rename(
        columns={'user_id': 'buyer_id'}
    )


# =====================================================
# 39. CAPITAL GAIN/LOSS (LeetCode 1393)
# =====================================================
def capital_gain_loss(stocks: pd.DataFrame) -> pd.DataFrame:
    # Apply negative sign to Buy operations
    stocks['amount'] = stocks.apply(
        lambda row: -row['price'] if row['operation'] == 'Buy' else row['price'],
        axis=1
    )

    result = stocks.groupby('stock_name')['amount'].sum().reset_index()
    result.columns = ['stock_name', 'capital_gain_loss']
    return result


# =====================================================
# 40. RECYCLABLE AND LOW FAT PRODUCTS (LeetCode 1757)
# =====================================================
def recyclable_low_fat(products: pd.DataFrame) -> pd.DataFrame:
    result = products[(products['low_fats'] == 'Y') & (products['recyclable'] == 'Y')]
    return result[['product_id']]


# =====================================================
# 41. THE LATEST LOGIN IN 2020 (LeetCode 1890)
# =====================================================
def latest_login(logins: pd.DataFrame) -> pd.DataFrame:
    # Extract year and filter for 2020
    logins['year'] = pd.to_datetime(logins['time_stamp']).dt.year
    logins_2020 = logins[logins['year'] == 2020]

    # Get max timestamp per user
    result = logins_2020.groupby('user_id')['time_stamp'].max().reset_index()
    result.columns = ['user_id', 'last_stamp']
    return result


# =====================================================
# EXAMPLE USAGE
# =====================================================
if __name__ == "__main__":
    # Example 1: Top Travellers
    users = pd.DataFrame({
        'id': [1, 2, 3],
        'name': ['Alice', 'Bob', 'Charlie']
    })
    rides = pd.DataFrame({
        'id': [1, 2, 3],
        'user_id': [1, 1, 2],
        'distance': [120, 150, 200]
    })
    print("Top Travellers:")
    print(top_travellers(users, rides))
    print()

    # Example 2: Big Countries
    world = pd.DataFrame({
        'name': ['Afghanistan', 'Albania', 'Algeria', 'Russia'],
        'population': [25500100, 2831741, 37100000, 146793744],
        'area': [652230, 28748, 2381741, 17098242]
    })
    print("Big Countries:")
    print(big_countries(world))
    print()

    # Example 3: Second Highest Salary
    employee = pd.DataFrame({
        'id': [1, 2, 3],
        'salary': [100, 200, 300]
    })
    print("Second Highest Salary:")
    print(second_highest_salary(employee))


    complete Python solutions file with all 41 LeetCode SQL problems implemented using pandas.
Key Features:
All Functions Include:

✅ Proper function signatures matching LeetCode format
✅ DataFrame manipulation using pandas operations
✅ Efficient implementations with groupby, merge, pivot operations
✅ Proper handling of NULL values and edge cases
✅ Comments explaining the logic

Common Pandas Techniques Used:

Merging/Joining: merge() with different join types (left, right, outer)
Grouping: groupby() with aggregations (sum, count, nunique)
Filtering: Boolean indexing with conditions
Sorting: sort_values() with multiple columns
Window Operations: shift(), rank() for consecutive/ranking problems
Pivoting: pivot() and melt() for table restructuring
String Operations: str.contains(), str.capitalize(), str.startswith()
Date Operations: pd.to_datetime(), .dt.year, .dt.days
Aggregations: Custom lambda functions with agg()
Conditional Logic: apply() with custom functions

Example Usage Included:
The file includes working examples at the bottom showing how to:

Create sample DataFrames
Call the functions
View results
```
