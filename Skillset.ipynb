```python
###Data Engineering for AI-> the core Data Engineering patterns that power AI systems — the exact things that make RAG, LLM apps,
and vector search actually work in production.

          Text preprocessing pipelines
          Document chunking strategies
          Embedding generation pipelines
          Real-time retrieval pipelines

🧱 1. Text Preprocessing Pipelines -> Goal: Convert raw text → clean, normalized, structured text ready for chunking + embeddings.

🔹 What problems this solves

          Removes noise (HTML, emojis, boilerplate)
          Normalizes text for consistent embeddings
          Extracts metadata (titles, headers, timestamps)
          Handles multi‑document ingestion at scale

🔹 Common preprocessing steps

          Lowercasing (optional for LLMs)
          Removing stopwords (optional for embeddings)
          Removing boilerplate (headers, footers)
          Sentence segmentation
          Tokenization
          Lemmatization (rarely needed for embeddings)
          Cleaning HTML, JSON, PDFs

Example 1 — Basic preprocessing pipeline (Python)

            import re
            from bs4 import BeautifulSoup
            import nltk
            nltk.download('punkt')

            def clean_text(raw):
                # Remove HTML
                text = BeautifulSoup(raw, "html.parser").get_text()

                # Remove non-alphanumeric
                text = re.sub(r"[^a-zA-Z0-9\s]", " ", text)

                # Normalize whitespace
                text = re.sub(r"\s+", " ", text).strip()

                return text

            def preprocess_document(doc):
                cleaned = clean_text(doc)
                sentences = nltk.sent_tokenize(cleaned)
                return sentences

            raw_doc = "<h1>Welcome!</h1> This is a test document. 😊"
            print(preprocess_document(raw_doc))
===================================
Example 2 — Preprocessing pipeline for enterprise PDFs

            from pypdf import PdfReader

            def extract_pdf_text(path):
                reader = PdfReader(path)
                text = ""
                for page in reader.pages:
                    text += page.extract_text() + "\n"
                return clean_text(text)
            Example 3 — SQL-based preprocessing (for logs or structured text)
            sql
            SELECT
                LOWER(REGEXP_REPLACE(message, '[^a-zA-Z0-9 ]', '')) AS cleaned_message,
                event_time,
                user_id
            FROM raw_logs;
==============================================
📚 2. Document Chunking Strategies

        Chunking is the most important part of RAG.
        Bad chunking = hallucinations.
        Good chunking = precise retrieval.

🔹 Common chunking strategies

1. Fixed-size chunking
        500–1000 tokens
        Simple, fast, works for most use cases

2. Recursive chunking
        Break by headings → paragraphs → sentences
        Preserves semantic structure

3. Sliding window chunking
        Overlapping chunks
        Best for legal, medical, financial documents

4. Semantic chunking

Use embeddings to detect topic boundaries
====================================
Example 1 — Fixed-size chunking

          def chunk_text(text, max_tokens=300):
              words = text.split()
              chunks = []

              for i in range(0, len(words), max_tokens):
                  chunk = " ".join(words[i:i+max_tokens])
                  chunks.append(chunk)

              return chunks

Example 2 — Sliding window chunking

          def sliding_window_chunks(words, size=300, overlap=50):
              chunks = []
              step = size - overlap

              for i in range(0, len(words), step):
                  chunk = words[i:i+size]
                  chunks.append(" ".join(chunk))

              return chunks

Example 3 — Recursive chunking (LangChain style)

          from langchain.text_splitter import RecursiveCharacterTextSplitter

          splitter = RecursiveCharacterTextSplitter(
              chunk_size=800,
              chunk_overlap=100,
              separators=["\n\n", "\n", ".", " ", ""]
          )

          chunks = splitter.split_text(long_text)
========================================================
🧬 3. Embedding Generation Pipelines -> Embeddings convert text → vectors for search, retrieval, clustering, and RAG.

🔹 What embedding pipelines must handle

          Batch embedding generation
          Deduplication
          Metadata association
          Versioning
          Re-embedding when models change
          Storing vectors in vector DBs

Example 1 — Generate embeddings (OpenAI / Azure OpenAI)

        from openai import OpenAI
        client = OpenAI()

        def embed_text(text):
            response = client.embeddings.create(
                model="text-embedding-3-large",
                input=text
            )
            return response.data[0].embedding
========================
Example 2 — Batch embedding pipeline
        def embed_chunks(chunks):
            vectors = []
            for chunk in chunks:
                vec = embed_text(chunk)
                vectors.append({"text": chunk, "embedding": vec})
            return vectors
=================================
Example 3 — Store embeddings in a vector DB (FAISS)

          import faiss
          import numpy as np

          def build_faiss_index(vectors):
              dim = len(vectors[0])
              index = faiss.IndexFlatL2(dim)
              index.add(np.array(vectors).astype('float32'))
              return index
===========================================
⚡ 4. Real-Time Retrieval Pipelines ->These pipelines power chatbots, search engines, and RAG systems.

🔹 Retrieval pipeline components

          User query → embedding
          Vector search (FAISS, Pinecone, Weaviate, Chroma)
          Hybrid search (keyword + vector)
          Reranking (optional)
          Context assembly
          LLM response generation

Example 1 — Real-time vector search (FAISS)

        def search(query, index, k=5):
            q_vec = embed_text(query)
            q_vec = np.array([q_vec]).astype('float32')
            distances, indices = index.search(q_vec, k)
            return indices

Example 2 — Real-time retrieval with metadata

        def retrieve_context(query, index, chunks):
            idxs = search(query, index)
            return [chunks[i] for i in idxs[0]]

Example 3 — Full RAG pipeline (simplified)

        def rag_answer(query, index, chunks):
            context = retrieve_context(query, index, chunks)
            prompt = f"Context:\n{context}\n\nQuestion: {query}\nAnswer:"

            response = client.chat.completions.create(
                model="gpt-4o-mini",
                messages=[{"role": "user", "content": prompt}]
            )
            return response.choices[0].message["content"]
===============================
🧩 5. Putting It All Together — End-to-End AI Data Pipeline

Pipeline Flow

          Ingest raw documents (PDF, HTML, logs, emails)
          Preprocess text (cleaning, normalization)
          Chunk documents (recursive + sliding window)
          Generate embeddings (batch jobs)
          Store vectors (FAISS / Pinecone / Weaviate)
          Real-time retrieval (vector + keyword)
          RAG answer generation

This is the backbone of every AI system:

          Chatbots
          Enterprise search
          Knowledge assistants
          LLM-powered analytics
          Customer support automation
          Legal/medical document retrieval
```
