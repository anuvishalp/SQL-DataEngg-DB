```python
Set 1 – Basics and SQL Fundamentals

Q1. What is SQL and what are its primary uses?
Answer: SQL (Structured Query Language) is a standard language to interact with relational databases.
Primary uses:

Data Definition: Create/alter/drop tables and other objects.

Data Manipulation: Insert/update/delete/select data.

Data Control: Grant/revoke permissions.

Transaction Control: Commit/rollback units of work.

Q2. Difference between SQL and MySQL.
Answer:

SQL: A language/standard used to query and manage relational data.

MySQL: A relational database management system (RDBMS) that implements SQL (with its own extensions).

Q3. Different sublanguages in SQL.
Answer:

DDL (Data Definition Language): CREATE, ALTER, DROP, TRUNCATE.

DML (Data Manipulation Language): SELECT, INSERT, UPDATE, DELETE, MERGE.

DCL (Data Control Language): GRANT, REVOKE.

TCL (Transaction Control Language): COMMIT, ROLLBACK, SAVEPOINT, SET TRANSACTION.

Q4. Difference between DDL, DML, DCL, TCL with examples.
Answer:

DDL: Defines structure.
\- Example: CREATE TABLE Employees (...);

DML: Manipulates data.
\- Example: INSERT INTO Employees VALUES (...);

DCL: Controls access.
\- Example: GRANT SELECT ON Employees TO UserX;

TCL: Manages transactions.
\- Example: COMMIT; ROLLBACK; SAVEPOINT sp1;.

Q5. What are constraints in SQL? Common types.
Answer: Constraints enforce rules on data to maintain integrity. Common types:

PRIMARY KEY: Uniquely identifies each row, not null.

UNIQUE: Ensures all values in a column (or set) are unique.

FOREIGN KEY: Enforces referential integrity to another table.

CHECK: Enforces a Boolean condition on values.

NOT NULL: Disallows null values.

DEFAULT: Provides default value when none is supplied.

Q6. Difference between PRIMARY KEY and UNIQUE constraints.
Answer:

PRIMARY KEY:

Only one per table (logical).

Implies UNIQUE + NOT NULL.

UNIQUE:

Multiple per table allowed.

Allows one NULL (in most RDBMSs) unless otherwise restricted.

Q7. Concept of NULL and how it differs from zero or empty string.
Answer:

NULL: Represents “unknown” or “missing” value, not a value itself.

Zero: Numeric value 0.

Empty string: A string of length 0 ('').
Comparisons with NULL use IS NULL/IS NOT NULL, not = or <>.

Q8. What is normalization? Different normal forms.
Answer: Normalization is the process of structuring tables to reduce redundancy and improve integrity.

1NF: No repeating groups; atomic columns.

2NF: 1NF + no partial dependency on a composite key.

3NF: 2NF + no transitive dependencies (non-key attributes depend only on key).

BCNF: Stronger 3NF; every determinant is a candidate key.

Higher forms (4NF, 5NF) handle multi‑valued and join dependencies.

Q9. What is denormalization and when to use it?
Answer: Denormalization intentionally introduces redundancy to improve read performance.
Use when:

Heavy read / complex joins hurt performance.

Reporting/analytics need pre‑joined or aggregated data.
Trade‑off: performance vs. storage and update complexity.

Q10. Concept of a transaction and ACID properties.
Answer: A transaction is a logical unit of work that must be fully completed or fully rolled back.
ACID:

Atomicity: All or nothing.

Consistency: Moves DB from one valid state to another.

Isolation: Concurrent transactions don’t interfere logically.

Durability: Once committed, changes persist even after failures.

Q11. Difference between SQL and NoSQL databases.
Answer:

SQL (Relational): Fixed schema, tables/rows, strong ACID, powerful joins, ideal for structured data.

NoSQL: Flexible schema, key‑value/document/column/graph models, often scale‑out, eventual consistency options, ideal for semi‑structured/unstructured or massive scale workloads.

Q12. Difference between logical and physical database design.
Answer:

Logical design: Entities, attributes, relationships, constraints—independent of specific DBMS; ERDs, normalization.

Physical design: Actual tables, indexes, partitions, filegroups, storage layout, access paths tuned for performance.

Q13. Various SQL standards and most widely used.
Answer:

Standards: SQL-86, SQL-89, SQL-92, SQL:1999, SQL:2003, SQL:2008, SQL:2011, SQL:2016, etc.

Most vendors loosely follow SQL-92 and later, but each adds proprietary extensions (T‑SQL, PL/SQL, etc.).

Q14. How does SQL handle data consistency across tables?
Answer:

Constraints: Primary/foreign keys, unique, check.

Transactions: ACID guarantees.

Triggers: Enforce complex rules.

Referential integrity: Ensures related rows remain valid.

Q15. Benefits of SQL over file‑based storage systems.
Answer:

Structured querying: Powerful, declarative queries.

Concurrency control: Multiple users safely access data.

Integrity: Constraints and transactions.

Security: Fine‑grained permissions.

Scalability & indexing: Efficient access paths.

Backup/recovery: Built‑in mechanisms.

Q16. Concept of data independence in SQL.
Answer: Ability to change schema at one level without affecting higher levels.

Logical data independence: Change logical schema (e.g., add columns) without changing applications.

Physical data independence: Change storage structures (indexes, files) without changing logical schema or applications.

Q17. What are database schemas and how are they used?
Answer: A schema is a logical container/grouping of database objects (tables, views, procedures, etc.).
Used for:

Organization: Group related objects.

Security: Grant permissions at schema level.

Namespace: Avoid name collisions.

Q18. Difference between OLTP and OLAP systems.
Answer:

OLTP (Online Transaction Processing):

Many small, concurrent transactions.

Highly normalized schema.

Focus on write performance and integrity.

OLAP (Online Analytical Processing):

Complex, read‑heavy queries.

Denormalized/star schemas.

Focus on aggregations, historical analysis.

Q19. Database instance vs. database schema.
Answer:

Database instance: The running DBMS process plus memory structures and physical files—an operational environment.

Database schema: The logical structure/definition of objects (tables, views, relationships) inside a database.

Q20. How does SQL enforce data integrity at different levels?
Answer:

Entity integrity: Primary keys, NOT NULL.

Referential integrity: Foreign keys with actions (CASCADE, SET NULL, RESTRICT).

Domain integrity: Data types, CHECK, DEFAULT, constraints.

Business rules: Triggers, stored procedures, application logic.
```

```python
Set 2 – Data Retrieval and Querying
1. What is the difference between WHERE and HAVING clauses?
WHERE filters rows before grouping; cannot use aggregates directly. HAVING filters after GROUP BY; typically used with aggregate expressions.

2. Explain the different types of JOINs in SQL with examples.

INNER JOIN: Matching rows in both tables.

LEFT JOIN: All left rows + matching right rows.

RIGHT JOIN: All right rows + matching left rows.

FULL OUTER JOIN: All rows from both, matched where possible.
Example:

sql
SELECT e.Name, d.Name
FROM Employees e
INNER JOIN Departments d ON e.DeptId = d.Id;

3. Difference between INNER JOIN and OUTER JOIN.
INNER JOIN returns only matching rows. OUTER JOIN (LEFT/RIGHT/FULL) returns matching rows plus non‑matching rows from one or both sides, with NULLs for missing columns.

4. How does a CROSS JOIN work and when would you use it?
CROSS JOIN returns the Cartesian product of two tables (every row of A with every row of B). Used for generating combinations, test data, or when logically needed.

5. Difference between UNION and UNION ALL.
UNION removes duplicates (distinct). UNION ALL keeps duplicates and is usually faster because it skips the distinct step.

6. Explain the difference between the LIKE and REGEXP operators.
LIKE uses simple wildcard patterns (%, _). REGEXP (or RLIKE) uses full regular expressions for more complex pattern matching (syntax DB‑specific).

7. How do you use wildcards in SQL queries?
With LIKE:

% = any sequence of characters.

_ = single character.
Example: WHERE Name LIKE 'A%' or WHERE Code LIKE 'A_3'.

8. What is the order of execution in an SQL query?
Roughly:

FROM / JOIN

WHERE

GROUP BY

HAVING

SELECT

DISTINCT

ORDER BY

OFFSET / FETCH / LIMIT.

9. Explain the GROUP BY clause and its purpose.
GROUP BY groups rows sharing common values into summary rows so aggregates (SUM, COUNT, etc.) can be computed per group.

10. What are aggregate functions? Give examples.
Functions that operate on sets of rows and return a single value: COUNT, SUM, AVG, MIN, MAX, STRING_AGG (DB‑specific), etc.

11. What is the purpose of the DISTINCT keyword?
Removes duplicate rows from the result set. Applied to all selected columns (or expression list) in the SELECT.

12. How do you sort query results using ORDER BY?
Use ORDER BY with column names/aliases or expressions, optionally ASC/DESC. Example: ORDER BY LastName ASC, FirstName DESC.

13. Explain the use of the OFFSET FETCH clause.
Used for pagination: skip a number of rows and then fetch a subset. Example (SQL Server/Postgres):

sql
ORDER BY Id
OFFSET 20 ROWS FETCH NEXT 10 ROWS ONLY;
14. What is the difference between INTERSECT and EXCEPT operators?
INTERSECT returns rows common to both queries. EXCEPT (or MINUS in Oracle) returns rows from the first query that are not in the second. Both usually remove duplicates.

15. How do you use the ALL, ANY, and SOME operators with subqueries?

> ALL(subquery): greater than every value in subquery.

> ANY(subquery) / > SOME(subquery): greater than at least one value.
Used with comparison operators to compare a value to a set.

16. What is the purpose of the TOP clause in SQL Server or LIMIT in MySQL?
They restrict the number of rows returned.

SQL Server: SELECT TOP 10 * FROM T;

MySQL/Postgres: SELECT * FROM T LIMIT 10;.

17. How do you perform string manipulation in SQL queries?
Using functions like CONCAT, SUBSTRING/SUBSTR, LEFT, RIGHT, UPPER, LOWER, TRIM, REPLACE, CHARINDEX/INSTR, etc. Exact names vary by DB.

18. Explain how NULL values affect the results of joins and aggregations.
In joins, NULLs don’t match other values (including NULL) unless using IS NULL. In aggregates, COUNT(*) counts all rows; COUNT(col) ignores NULLs; SUM/AVG ignore NULLs.

19. What is the difference between a self-join and a recursive query?
Self‑join: a table joined to itself in a single query (e.g., employees to managers). Recursive query: repeatedly references its own result (usually via recursive CTE) to traverse hierarchies.

20. How do you use pattern matching effectively in SQL queries?
Use LIKE/ILIKE (case‑insensitive in some DBs) for simple patterns, REGEXP for complex ones, and indexes (e.g., functional or full‑text) where possible to avoid full scans.

21. What is the COALESCE function and how is it used?
COALESCE(expr1, expr2, ...) returns the first non‑NULL expression. Commonly used to replace NULLs with defaults: COALESCE(MiddleName, '').

22. Explain the usage of NULLIF and ISNULL functions.

NULLIF(a, b): returns NULL if a = b, else a.

ISNULL(expr, replacement) (SQL Server) or IFNULL (MySQL): returns replacement if expr is NULL, else expr.

23. How do you use conditional logic in SELECT statements?
With CASE expressions:

sql
SELECT
  CASE WHEN Score >= 90 THEN 'A'
       WHEN Score >= 80 THEN 'B'
       ELSE 'C'
  END AS Grade
FROM Exams;
24. What is the purpose of the WITH TIES option in SQL Server?
Used with TOP and ORDER BY to include additional rows that tie on the last ordering value. Example: SELECT TOP 3 WITH TIES ... ORDER BY Score DESC;.

25. Explain how to implement pagination in SQL queries.
Use ORDER BY plus OFFSET/FETCH or LIMIT/OFFSET. Example:

sql
SELECT *
FROM T
ORDER BY Id
OFFSET (page-1)*page_size ROWS
FETCH NEXT page_size ROWS ONLY;
```

```python
Set 3 – Advanced Querying
1. What are subqueries and how do they work?
A subquery is a query nested inside another query. It can return a single value, a row, or a set of rows, and can be used in SELECT, FROM, or WHERE/HAVING.

2. Difference between correlated and non-correlated subqueries.
Non‑correlated: independent; can run alone. Correlated: references columns from the outer query and is evaluated once per outer row.

3. Explain the EXISTS operator with an example.
EXISTS checks if a subquery returns any row (true/false).

sql
SELECT *
FROM Customers c
WHERE EXISTS (
  SELECT 1 FROM Orders o
  WHERE o.CustomerId = c.Id
);
4. How do you use the CASE statement in SQL?
CASE returns different values based on conditions, used in SELECT, ORDER BY, etc. (see example in Set 2, Q23).

5. What are Common Table Expressions (CTEs) and how are they useful?
CTEs are named temporary result sets defined with WITH and used within a single statement. They improve readability, allow recursion, and can replace some derived tables.

6. Explain window functions and their applications.
Window functions perform calculations across a set of rows related to the current row, without collapsing rows. Used for rankings, running totals, moving averages, lag/lead comparisons, etc.

7. Difference between ROW_NUMBER(), RANK(), and DENSE_RANK().

ROW_NUMBER(): unique sequence, no ties.

RANK(): same rank for ties, gaps after ties.

DENSE_RANK(): same rank for ties, no gaps.

8. How would you find the Nth highest salary in a table?
Using ROW_NUMBER or DENSE_RANK:

sql
SELECT Salary
FROM (
  SELECT Salary,
         DENSE_RANK() OVER (ORDER BY Salary DESC) AS r
  FROM Employees
) x
WHERE r = N;
9. Explain how to pivot data in SQL.
Pivot converts rows to columns. Example (SQL Server):

sql
SELECT *
FROM (
  SELECT Year, Quarter, Sales FROM SalesData
) src
PIVOT (
  SUM(Sales) FOR Quarter IN ([Q1],[Q2],[Q3],[Q4])
) p;
10. What are recursive queries and when would you use them?
Recursive queries (recursive CTEs) repeatedly reference themselves to traverse hierarchies (org charts, folder trees, bill of materials).

11. How do you use analytical functions like LAG and LEAD?
LAG/LEAD access previous/next row values in the window:

sql
SELECT
  Date,
  Sales,
  LAG(Sales)  OVER (ORDER BY Date) AS PrevSales,
  LEAD(Sales) OVER (ORDER BY Date) AS NextSales
FROM DailySales;
12. What is the purpose of FIRST_VALUE and LAST_VALUE functions?
They return the first/last value in the window frame for each row, useful for comparisons to starting/ending values.

13. Explain the use of GROUPING SETS, ROLLUP, and CUBE.
They generate multiple groupings in a single query:

ROLLUP: hierarchical subtotals (e.g., by year, then year+month, then grand total).

CUBE: all combinations of grouping columns.

GROUPING SETS: explicit list of groupings.

14. How do you implement running totals and moving averages?
Using window functions with frame clauses:

sql
SUM(Amount) OVER (ORDER BY Date
                  ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS RunningTotal
Moving average uses a sliding frame (e.g., last 3 rows).

15. What is a lateral join and when would you use it?
A lateral join (CROSS APPLY/OUTER APPLY or LATERAL) allows the right side to reference columns from the left side. Useful when the right subquery depends on each left row.

16. How do you handle hierarchical data queries in SQL?
Using recursive CTEs, adjacency lists (ParentId), nested sets, or DB‑specific hierarchy types (CONNECT BY in Oracle, hierarchyid in SQL Server).

17. Explain the concept of window frame specification in window functions.
The frame defines which rows in the partition are considered for each calculation (e.g., ROWS BETWEEN 1 PRECEDING AND 1 FOLLOWING, or RANGE BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW).

18. What are the EXCLUDE and RESPECT NULLS options in window functions?
In some DBs (e.g., Postgres), they control whether certain rows or NULLs are included in the window frame for functions like FIRST_VALUE/LAST_VALUE.

19. How do you implement gap and island problems in SQL?
Typically using window functions and grouping logic: compare row numbers vs. values, or use LAG to detect breaks, then group contiguous ranges.

20. What is the difference between a derived table and a CTE?
Both are temporary result sets. Derived table: inline subquery in FROM. CTE: defined with WITH and referenced by name; often more readable and can be recursive.

21. How do you use the WITH RECURSIVE clause in PostgreSQL?

sql
WITH RECURSIVE cte AS (
  SELECT Id, ParentId, 1 AS Level
  FROM Nodes
  WHERE ParentId IS NULL
  UNION ALL
  SELECT n.Id, n.ParentId, c.Level + 1
  FROM Nodes n
  JOIN cte c ON n.ParentId = c.Id
)
SELECT * FROM cte;
22. Explain how to perform complex aggregations using multiple window functions.
You can define several window functions in the same SELECT, each with its own OVER clause (different partitions/frames) to compute multiple metrics per row (e.g., rank, running total, moving average).

23. How do you use NTILE() and PERCENT_RANK() functions?

NTILE(n): divides ordered rows into n buckets (1..n).

PERCENT_RANK(): relative rank between 0 and 1 based on ordering.

24. What are table-valued functions and how do they differ from scalar functions?
Table‑valued functions return a table that can be used in FROM. Scalar functions return a single value. TVFs integrate into queries like views but can accept parameters.

25. How do you handle complex date-based calculations in SQL?
Use date functions: DATEADD, DATEDIFF, DATE_TRUNC, EXTRACT, DATEPART, plus window functions for period‑over‑period comparisons, and calendar tables for business‑day logic.
```

```python
Set 8 – Transactions and Concurrency
1. What are the different transaction isolation levels in SQL?
Read Uncommitted: Allows dirty reads.

Read Committed: Prevents dirty reads; default in many DBs.

Repeatable Read: Prevents dirty + non‑repeatable reads.

Serializable: Highest isolation; prevents phantoms.

Snapshot (SQL Server): Uses row versioning to avoid blocking.

2. Explain the problems that can occur in concurrent transactions.
Dirty Read: Read uncommitted data.

Non‑Repeatable Read: Same query returns different results within a transaction.

Phantom Read: New rows appear that match a previous search condition.

Lost Update: Two transactions overwrite each other’s changes.

3. What is deadlock and how can it be prevented?
Deadlock occurs when two sessions wait on each other’s locks.
Prevention:

Access objects in consistent order

Keep transactions short

Use proper indexes

Avoid user interaction inside transactions

4. Explain optimistic vs. pessimistic locking.
Optimistic: Assume no conflicts; verify at commit using row versions. Best for low‑contention systems.

Pessimistic: Lock rows before access; prevents conflicts but increases blocking.

5. What is row versioning in SQL?
Technique where old row versions are stored (tempdb in SQL Server) so readers don’t block writers. Used in snapshot isolation.

6. How does READ COMMITTED SNAPSHOT isolation level work?
Readers get a consistent snapshot of committed data using row versions. Eliminates shared locks → reduces blocking.

7. What is lock escalation and when does it occur?
DB converts many fine‑grained locks (row/page) into a table lock to reduce overhead. Happens when lock count exceeds thresholds.

8. How do you handle long‑running transactions?
Break into smaller batches

Use snapshot isolation

Reduce locking footprint

Avoid holding locks during user input

Tune queries to reduce execution time

9. Difference between implicit and explicit transactions.
Implicit: DB auto‑starts a transaction for each statement.

Explicit: Developer uses BEGIN TRAN, COMMIT, ROLLBACK.

10. Explain how to use the WITH NOLOCK hint and its implications.
WITH (NOLOCK) = dirty reads allowed; avoids blocking but risks reading uncommitted, inconsistent, or phantom data. Should be used cautiously.

11. What is lock granularity and how does it affect performance?
Granularity = size of locked resource (row, page, table).
Fine‑grained = less blocking, more overhead.
Coarse‑grained = more blocking, less overhead.

12. How do you monitor locks in a database system?
Use DMVs (SQL Server), pg_locks (Postgres), V$LOCK (Oracle), profiler tools, or monitoring dashboards.

13. What is a transaction log and how is it used in recovery?
Log records every change. Used for rollback, crash recovery, point‑in‑time restore, replication.

14. Explain the concept of two-phase commit.
Protocol for distributed transactions:

Phase 1: Prepare — participants vote commit/abort.

Phase 2: Commit — coordinator instructs commit if all voted yes.

15. How do you handle distributed transactions?
Use transaction coordinators (MSDTC, XA), two‑phase commit, or application‑level compensation logic.

16. What is the impact of isolation levels on concurrency and throughput?
Higher isolation = fewer anomalies but more blocking.
Lower isolation = more concurrency but more anomalies.

17. Explain how row versioning affects tempdb usage.
Snapshot isolation stores row versions in tempdb. Heavy versioning → tempdb growth and potential contention.

18. What are the different lock types (shared, exclusive, update, etc.)?
Shared (S): Read locks.

Exclusive (X): Write locks.

Update (U): Prevents deadlocks during updates.

Intent locks: Indicate future locks at lower granularity.

Schema locks: Protect metadata.

19. How do you handle blocking transactions?
Identify blockers

Kill long blockers (carefully)

Use snapshot isolation

Tune queries

Add indexes

Reduce transaction scope

20. What is a transaction savepoint and how is it used?
A marker inside a transaction allowing partial rollback.

sql
SAVEPOINT sp1;
ROLLBACK TO sp1;
21. Explain the concept of transaction chaining.
In some DBs, after a transaction ends, the next statement automatically starts a new one. Rarely used today.

22. How do you implement retry logic for handling deadlocks?
Catch deadlock error → wait → retry transaction with exponential backoff.

23. What is a transaction schedule and serializability?
Schedule = order of operations from concurrent transactions.
Serializable = schedule equivalent to some serial execution → no anomalies.

24. How do you implement optimistic concurrency in applications?
Use row version/timestamp columns:

sql
UPDATE T
SET ...
WHERE Id = @Id AND RowVersion = @OldVersion;
If 0 rows affected → concurrency conflict.

25. Best practices for transaction management in high‑concurrency environments.
Keep transactions short

Use proper isolation levels

Use snapshot isolation when appropriate

Avoid locking large ranges

Index filtering columns

Avoid user interaction inside transactions
```
