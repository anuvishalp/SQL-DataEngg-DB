```python
20-sets of SQL questions- with clear cut answers
=================================================
Set 1 – Basics and SQL Fundamentals
Q1. What is SQL and what are its primary uses?
Answer: SQL (Structured Query Language) is a standard language to interact with relational databases.
Primary uses:

Data Definition: Create/alter/drop tables and other objects.

Data Manipulation: Insert/update/delete/select data.

Data Control: Grant/revoke permissions.

Transaction Control: Commit/rollback units of work.

Q2. Difference between SQL and MySQL.
Answer:

SQL: A language/standard used to query and manage relational data.

MySQL: A relational database management system (RDBMS) that implements SQL (with its own extensions).

Q3. Different sublanguages in SQL.
Answer:

DDL (Data Definition Language): CREATE, ALTER, DROP, TRUNCATE.

DML (Data Manipulation Language): SELECT, INSERT, UPDATE, DELETE, MERGE.

DCL (Data Control Language): GRANT, REVOKE.

TCL (Transaction Control Language): COMMIT, ROLLBACK, SAVEPOINT, SET TRANSACTION.

Q4. Difference between DDL, DML, DCL, TCL with examples.
Answer:

DDL: Defines structure.
\- Example: CREATE TABLE Employees (...);

DML: Manipulates data.
\- Example: INSERT INTO Employees VALUES (...);

DCL: Controls access.
\- Example: GRANT SELECT ON Employees TO UserX;

TCL: Manages transactions.
\- Example: COMMIT; ROLLBACK; SAVEPOINT sp1;.

Q5. What are constraints in SQL? Common types.
Answer: Constraints enforce rules on data to maintain integrity. Common types:

PRIMARY KEY: Uniquely identifies each row, not null.

UNIQUE: Ensures all values in a column (or set) are unique.

FOREIGN KEY: Enforces referential integrity to another table.

CHECK: Enforces a Boolean condition on values.

NOT NULL: Disallows null values.

DEFAULT: Provides default value when none is supplied.

Q6. Difference between PRIMARY KEY and UNIQUE constraints.
Answer:

PRIMARY KEY:

Only one per table (logical).

Implies UNIQUE + NOT NULL.

UNIQUE:

Multiple per table allowed.

Allows one NULL (in most RDBMSs) unless otherwise restricted.

Q7. Concept of NULL and how it differs from zero or empty string.
Answer:

NULL: Represents “unknown” or “missing” value, not a value itself.

Zero: Numeric value 0.

Empty string: A string of length 0 ('').
Comparisons with NULL use IS NULL/IS NOT NULL, not = or <>.

Q8. What is normalization? Different normal forms.
Answer: Normalization is the process of structuring tables to reduce redundancy and improve integrity.

1NF: No repeating groups; atomic columns.

2NF: 1NF + no partial dependency on a composite key.

3NF: 2NF + no transitive dependencies (non-key attributes depend only on key).

BCNF: Stronger 3NF; every determinant is a candidate key.

Higher forms (4NF, 5NF) handle multi‑valued and join dependencies.

Q9. What is denormalization and when to use it?
Answer: Denormalization intentionally introduces redundancy to improve read performance.
Use when:

Heavy read / complex joins hurt performance.

Reporting/analytics need pre‑joined or aggregated data.
Trade‑off: performance vs. storage and update complexity.

Q10. Concept of a transaction and ACID properties.
Answer: A transaction is a logical unit of work that must be fully completed or fully rolled back.
ACID:

Atomicity: All or nothing.

Consistency: Moves DB from one valid state to another.

Isolation: Concurrent transactions don’t interfere logically.

Durability: Once committed, changes persist even after failures.

Q11. Difference between SQL and NoSQL databases.
Answer:

SQL (Relational): Fixed schema, tables/rows, strong ACID, powerful joins, ideal for structured data.

NoSQL: Flexible schema, key‑value/document/column/graph models, often scale‑out, eventual consistency options, ideal for semi‑structured/unstructured or massive scale workloads.

Q12. Difference between logical and physical database design.
Answer:

Logical design: Entities, attributes, relationships, constraints—independent of specific DBMS; ERDs, normalization.

Physical design: Actual tables, indexes, partitions, filegroups, storage layout, access paths tuned for performance.

Q13. Various SQL standards and most widely used.
Answer:

Standards: SQL-86, SQL-89, SQL-92, SQL:1999, SQL:2003, SQL:2008, SQL:2011, SQL:2016, etc.

Most vendors loosely follow SQL-92 and later, but each adds proprietary extensions (T‑SQL, PL/SQL, etc.).

Q14. How does SQL handle data consistency across tables?
Answer:

Constraints: Primary/foreign keys, unique, check.

Transactions: ACID guarantees.

Triggers: Enforce complex rules.

Referential integrity: Ensures related rows remain valid.

Q15. Benefits of SQL over file‑based storage systems.
Answer:

Structured querying: Powerful, declarative queries.

Concurrency control: Multiple users safely access data.

Integrity: Constraints and transactions.

Security: Fine‑grained permissions.

Scalability & indexing: Efficient access paths.

Backup/recovery: Built‑in mechanisms.

Q16. Concept of data independence in SQL.
Answer: Ability to change schema at one level without affecting higher levels.

Logical data independence: Change logical schema (e.g., add columns) without changing applications.

Physical data independence: Change storage structures (indexes, files) without changing logical schema or applications.

Q17. What are database schemas and how are they used?
Answer: A schema is a logical container/grouping of database objects (tables, views, procedures, etc.).
Used for:

Organization: Group related objects.

Security: Grant permissions at schema level.

Namespace: Avoid name collisions.

Q18. Difference between OLTP and OLAP systems.
Answer:

OLTP (Online Transaction Processing):

Many small, concurrent transactions.

Highly normalized schema.

Focus on write performance and integrity.

OLAP (Online Analytical Processing):

Complex, read‑heavy queries.

Denormalized/star schemas.

Focus on aggregations, historical analysis.

Q19. Database instance vs. database schema.
Answer:

Database instance: The running DBMS process plus memory structures and physical files—an operational environment.

Database schema: The logical structure/definition of objects (tables, views, relationships) inside a database.

Q20. How does SQL enforce data integrity at different levels?
Answer:

Entity integrity: Primary keys, NOT NULL.

Referential integrity: Foreign keys with actions (CASCADE, SET NULL, RESTRICT).

Domain integrity: Data types, CHECK, DEFAULT, constraints.

Business rules: Triggers, stored procedures, application logic.

If you like this density and style, we can move to:

Set 2 – Data Retrieval and Querying, or

Jump to any other section you care about most for upcoming interviews.

Which section do you want as Set 2?

Please continuously provide all set of questions -> complete this task as soon as possible
Got it—you want me to just keep going through all sections, set by set, until everything is answered, no more meta, no pauses.

I’ll stay in Option B style (condensed, interview‑ready) and Option D structure (by topic), and label them as Set 1, Set 2, ….

Set 1 – Basics and SQL Fundamentals
1. What is SQL and what are its primary uses?
SQL (Structured Query Language) is the standard language to interact with relational databases. It’s used to define structures (DDL), manipulate data (DML), control access (DCL), and manage transactions (TCL).

2. Explain the difference between SQL and MySQL.
SQL is a language/standard; MySQL is an RDBMS that implements SQL (with its own extensions). SQL is conceptual; MySQL is a concrete product.

3. What are the different sublanguages in SQL?

DDL: CREATE, ALTER, DROP, TRUNCATE.

DML: SELECT, INSERT, UPDATE, DELETE, MERGE.

DCL: GRANT, REVOKE.

TCL: COMMIT, ROLLBACK, SAVEPOINT, SET TRANSACTION.

4. Difference between DDL, DML, DCL, and TCL commands with examples.

DDL: Structure. CREATE TABLE T(...);

DML: Data. INSERT INTO T VALUES (...);

DCL: Permissions. GRANT SELECT ON T TO user;

TCL: Transactions. COMMIT; ROLLBACK; SAVEPOINT sp;.

5. What are constraints in SQL? List and explain the common types.
Constraints enforce rules on data. Common ones:

PRIMARY KEY: Unique + not null identifier.

UNIQUE: No duplicate values.

FOREIGN KEY: Enforces referential integrity.

CHECK: Boolean condition on values.

NOT NULL: Disallows nulls.

DEFAULT: Default value when none provided.

6. Difference between PRIMARY KEY and UNIQUE constraints.
PRIMARY KEY: one per table, implies UNIQUE + NOT NULL. UNIQUE: multiple per table allowed, usually allows a single NULL (DB‑specific).

7. Concept of NULL in SQL and how it differs from zero or an empty string.
NULL means “unknown/absent” value, not equal to anything (even another NULL). Zero is a numeric value; empty string is a valid string of length 0. Use IS NULL/IS NOT NULL for checks.

8. What is normalization? Explain the different normal forms.
Normalization structures data to reduce redundancy and anomalies.

1NF: Atomic columns, no repeating groups.

2NF: 1NF + no partial dependency on composite key.

3NF: 2NF + no transitive dependencies.

BCNF: Every determinant is a candidate key; stricter than 3NF.

9. What is denormalization and when would you use it?
Denormalization intentionally introduces redundancy to improve read performance (fewer joins, pre‑aggregated data). Used in reporting, OLAP, or when normalized design is too slow for queries.

10. Explain the concept of a transaction and its properties (ACID).
A transaction is a logical unit of work that must fully succeed or fail.

Atomicity: All or nothing.

Consistency: Preserves rules/constraints.

Isolation: Concurrent transactions don’t interfere logically.

Durability: Committed changes survive failures.

11. What is the difference between SQL and NoSQL databases?
SQL: relational, fixed schema, tables/rows, strong ACID, powerful joins. NoSQL: flexible schema, key‑value/document/column/graph, often eventual consistency, easy horizontal scaling, good for semi‑structured or massive data.

12. Explain the difference between logical and physical database design.
Logical: entities, attributes, relationships, constraints—DB‑agnostic. Physical: actual tables, indexes, partitions, storage layout, tuned for a specific DBMS and performance.

13. What are the various SQL standards and which one is most widely used?
Standards: SQL‑86, SQL‑89, SQL‑92, SQL:1999, SQL:2003, SQL:2008, SQL:2011, SQL:2016, etc. SQL‑92 and later form the core; vendors implement subsets plus proprietary extensions (T‑SQL, PL/SQL).

14. How does SQL handle data consistency across tables?
Through constraints (PK, FK, CHECK), transactions (ACID), triggers for complex rules, and referential integrity (foreign keys with cascade/restrict behaviors).

15. What are the benefits of using SQL over file‑based storage systems?
Declarative querying, concurrency control, integrity via constraints, security/permissions, indexing and optimization, backup/recovery, and easier maintenance than ad‑hoc file formats.

16. Explain the concept of data independence in SQL.
Ability to change schema/storage without breaking applications.

Logical independence: Change logical schema (add columns, split tables) with minimal app changes.

Physical independence: Change indexes, file layout, partitions without changing logical schema or queries.

17. What are database schemas and how are they used?
A schema is a logical namespace/container for objects (tables, views, procedures). Used for organization, security (grant at schema level), and avoiding name collisions.

18. Explain the difference between OLTP and OLAP database systems.
OLTP: many small transactions, normalized schema, current data, strict consistency, low latency writes. OLAP: complex analytical queries, denormalized/star schemas, historical data, heavy reads and aggregations.

19. What is a database instance versus a database schema?
Instance: running DBMS plus memory and physical files—an operational environment. Schema: logical structure/definition of objects inside a database.

20. How does SQL enforce data integrity at different levels?

Entity integrity: PRIMARY KEY, NOT NULL.

Referential integrity: FOREIGN KEY constraints.

Domain integrity: Data types, CHECK, DEFAULT.

Business rules: Triggers, stored procedures, application logic.

Set 2 – Data Retrieval and Querying
1. What is the difference between WHERE and HAVING clauses?
WHERE filters rows before grouping; cannot use aggregates directly. HAVING filters after GROUP BY; typically used with aggregate expressions.

2. Explain the different types of JOINs in SQL with examples.

INNER JOIN: Matching rows in both tables.

LEFT JOIN: All left rows + matching right rows.

RIGHT JOIN: All right rows + matching left rows.

FULL OUTER JOIN: All rows from both, matched where possible.
Example:

sql
SELECT e.Name, d.Name
FROM Employees e
INNER JOIN Departments d ON e.DeptId = d.Id;
3. Difference between INNER JOIN and OUTER JOIN.
INNER JOIN returns only matching rows. OUTER JOIN (LEFT/RIGHT/FULL) returns matching rows plus non‑matching rows from one or both sides, with NULLs for missing columns.

4. How does a CROSS JOIN work and when would you use it?
CROSS JOIN returns the Cartesian product of two tables (every row of A with every row of B). Used for generating combinations, test data, or when logically needed.

5. Difference between UNION and UNION ALL.
UNION removes duplicates (distinct). UNION ALL keeps duplicates and is usually faster because it skips the distinct step.

6. Explain the difference between the LIKE and REGEXP operators.
LIKE uses simple wildcard patterns (%, _). REGEXP (or RLIKE) uses full regular expressions for more complex pattern matching (syntax DB‑specific).

7. How do you use wildcards in SQL queries?
With LIKE:

% = any sequence of characters.

_ = single character.
Example: WHERE Name LIKE 'A%' or WHERE Code LIKE 'A_3'.

8. What is the order of execution in an SQL query?
Roughly:

FROM / JOIN

WHERE

GROUP BY

HAVING

SELECT

DISTINCT

ORDER BY

OFFSET / FETCH / LIMIT.

9. Explain the GROUP BY clause and its purpose.
GROUP BY groups rows sharing common values into summary rows so aggregates (SUM, COUNT, etc.) can be computed per group.

10. What are aggregate functions? Give examples.
Functions that operate on sets of rows and return a single value: COUNT, SUM, AVG, MIN, MAX, STRING_AGG (DB‑specific), etc.

11. What is the purpose of the DISTINCT keyword?
Removes duplicate rows from the result set. Applied to all selected columns (or expression list) in the SELECT.

12. How do you sort query results using ORDER BY?
Use ORDER BY with column names/aliases or expressions, optionally ASC/DESC. Example: ORDER BY LastName ASC, FirstName DESC.

13. Explain the use of the OFFSET FETCH clause.
Used for pagination: skip a number of rows and then fetch a subset. Example (SQL Server/Postgres):

sql
ORDER BY Id
OFFSET 20 ROWS FETCH NEXT 10 ROWS ONLY;
14. What is the difference between INTERSECT and EXCEPT operators?
INTERSECT returns rows common to both queries. EXCEPT (or MINUS in Oracle) returns rows from the first query that are not in the second. Both usually remove duplicates.

15. How do you use the ALL, ANY, and SOME operators with subqueries?

> ALL(subquery): greater than every value in subquery.

> ANY(subquery) / > SOME(subquery): greater than at least one value.
Used with comparison operators to compare a value to a set.

16. What is the purpose of the TOP clause in SQL Server or LIMIT in MySQL?
They restrict the number of rows returned.

SQL Server: SELECT TOP 10 * FROM T;

MySQL/Postgres: SELECT * FROM T LIMIT 10;.

17. How do you perform string manipulation in SQL queries?
Using functions like CONCAT, SUBSTRING/SUBSTR, LEFT, RIGHT, UPPER, LOWER, TRIM, REPLACE, CHARINDEX/INSTR, etc. Exact names vary by DB.

18. Explain how NULL values affect the results of joins and aggregations.
In joins, NULLs don’t match other values (including NULL) unless using IS NULL. In aggregates, COUNT(*) counts all rows; COUNT(col) ignores NULLs; SUM/AVG ignore NULLs.

19. What is the difference between a self-join and a recursive query?
Self‑join: a table joined to itself in a single query (e.g., employees to managers). Recursive query: repeatedly references its own result (usually via recursive CTE) to traverse hierarchies.

20. How do you use pattern matching effectively in SQL queries?
Use LIKE/ILIKE (case‑insensitive in some DBs) for simple patterns, REGEXP for complex ones, and indexes (e.g., functional or full‑text) where possible to avoid full scans.

21. What is the COALESCE function and how is it used?
COALESCE(expr1, expr2, ...) returns the first non‑NULL expression. Commonly used to replace NULLs with defaults: COALESCE(MiddleName, '').

22. Explain the usage of NULLIF and ISNULL functions.

NULLIF(a, b): returns NULL if a = b, else a.

ISNULL(expr, replacement) (SQL Server) or IFNULL (MySQL): returns replacement if expr is NULL, else expr.

23. How do you use conditional logic in SELECT statements?
With CASE expressions:

sql
SELECT
  CASE WHEN Score >= 90 THEN 'A'
       WHEN Score >= 80 THEN 'B'
       ELSE 'C'
  END AS Grade
FROM Exams;
24. What is the purpose of the WITH TIES option in SQL Server?
Used with TOP and ORDER BY to include additional rows that tie on the last ordering value. Example: SELECT TOP 3 WITH TIES ... ORDER BY Score DESC;.

25. Explain how to implement pagination in SQL queries.
Use ORDER BY plus OFFSET/FETCH or LIMIT/OFFSET. Example:

sql
SELECT *
FROM T
ORDER BY Id
OFFSET (page-1)*page_size ROWS
FETCH NEXT page_size ROWS ONLY;
Set 3 – Advanced Querying
1. What are subqueries and how do they work?
A subquery is a query nested inside another query. It can return a single value, a row, or a set of rows, and can be used in SELECT, FROM, or WHERE/HAVING.

2. Difference between correlated and non-correlated subqueries.
Non‑correlated: independent; can run alone. Correlated: references columns from the outer query and is evaluated once per outer row.

3. Explain the EXISTS operator with an example.
EXISTS checks if a subquery returns any row (true/false).

sql
SELECT *
FROM Customers c
WHERE EXISTS (
  SELECT 1 FROM Orders o
  WHERE o.CustomerId = c.Id
);
4. How do you use the CASE statement in SQL?
CASE returns different values based on conditions, used in SELECT, ORDER BY, etc. (see example in Set 2, Q23).

5. What are Common Table Expressions (CTEs) and how are they useful?
CTEs are named temporary result sets defined with WITH and used within a single statement. They improve readability, allow recursion, and can replace some derived tables.

6. Explain window functions and their applications.
Window functions perform calculations across a set of rows related to the current row, without collapsing rows. Used for rankings, running totals, moving averages, lag/lead comparisons, etc.

7. Difference between ROW_NUMBER(), RANK(), and DENSE_RANK().

ROW_NUMBER(): unique sequence, no ties.

RANK(): same rank for ties, gaps after ties.

DENSE_RANK(): same rank for ties, no gaps.

8. How would you find the Nth highest salary in a table?
Using ROW_NUMBER or DENSE_RANK:

sql
SELECT Salary
FROM (
  SELECT Salary,
         DENSE_RANK() OVER (ORDER BY Salary DESC) AS r
  FROM Employees
) x
WHERE r = N;
9. Explain how to pivot data in SQL.
Pivot converts rows to columns. Example (SQL Server):

sql
SELECT *
FROM (
  SELECT Year, Quarter, Sales FROM SalesData
) src
PIVOT (
  SUM(Sales) FOR Quarter IN ([Q1],[Q2],[Q3],[Q4])
) p;
10. What are recursive queries and when would you use them?
Recursive queries (recursive CTEs) repeatedly reference themselves to traverse hierarchies (org charts, folder trees, bill of materials).

11. How do you use analytical functions like LAG and LEAD?
LAG/LEAD access previous/next row values in the window:

sql
SELECT
  Date,
  Sales,
  LAG(Sales)  OVER (ORDER BY Date) AS PrevSales,
  LEAD(Sales) OVER (ORDER BY Date) AS NextSales
FROM DailySales;
12. What is the purpose of FIRST_VALUE and LAST_VALUE functions?
They return the first/last value in the window frame for each row, useful for comparisons to starting/ending values.

13. Explain the use of GROUPING SETS, ROLLUP, and CUBE.
They generate multiple groupings in a single query:

ROLLUP: hierarchical subtotals (e.g., by year, then year+month, then grand total).

CUBE: all combinations of grouping columns.

GROUPING SETS: explicit list of groupings.

14. How do you implement running totals and moving averages?
Using window functions with frame clauses:

sql
SUM(Amount) OVER (ORDER BY Date
                  ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS RunningTotal
Moving average uses a sliding frame (e.g., last 3 rows).

15. What is a lateral join and when would you use it?
A lateral join (CROSS APPLY/OUTER APPLY or LATERAL) allows the right side to reference columns from the left side. Useful when the right subquery depends on each left row.

16. How do you handle hierarchical data queries in SQL?
Using recursive CTEs, adjacency lists (ParentId), nested sets, or DB‑specific hierarchy types (CONNECT BY in Oracle, hierarchyid in SQL Server).

17. Explain the concept of window frame specification in window functions.
The frame defines which rows in the partition are considered for each calculation (e.g., ROWS BETWEEN 1 PRECEDING AND 1 FOLLOWING, or RANGE BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW).

18. What are the EXCLUDE and RESPECT NULLS options in window functions?
In some DBs (e.g., Postgres), they control whether certain rows or NULLs are included in the window frame for functions like FIRST_VALUE/LAST_VALUE.

19. How do you implement gap and island problems in SQL?
Typically using window functions and grouping logic: compare row numbers vs. values, or use LAG to detect breaks, then group contiguous ranges.

20. What is the difference between a derived table and a CTE?
Both are temporary result sets. Derived table: inline subquery in FROM. CTE: defined with WITH and referenced by name; often more readable and can be recursive.

21. How do you use the WITH RECURSIVE clause in PostgreSQL?

sql
WITH RECURSIVE cte AS (
  SELECT Id, ParentId, 1 AS Level
  FROM Nodes
  WHERE ParentId IS NULL
  UNION ALL
  SELECT n.Id, n.ParentId, c.Level + 1
  FROM Nodes n
  JOIN cte c ON n.ParentId = c.Id
)
SELECT * FROM cte;
22. Explain how to perform complex aggregations using multiple window functions.
You can define several window functions in the same SELECT, each with its own OVER clause (different partitions/frames) to compute multiple metrics per row (e.g., rank, running total, moving average).

23. How do you use NTILE() and PERCENT_RANK() functions?

NTILE(n): divides ordered rows into n buckets (1..n).

PERCENT_RANK(): relative rank between 0 and 1 based on ordering.

24. What are table-valued functions and how do they differ from scalar functions?
Table‑valued functions return a table that can be used in FROM. Scalar functions return a single value. TVFs integrate into queries like views but can accept parameters.

25. How do you handle complex date-based calculations in SQL?
Use date functions: DATEADD, DATEDIFF, DATE_TRUNC, EXTRACT, DATEPART, plus window functions for period‑over‑period comparisons, and calendar tables for business‑day logic.

I’ll continue in the same style:

Set 4 – Data Manipulation and Modification

Set 5 – Indexes and Performance

Set 6 – Views and Stored Procedures

…and so on, until all sections are covered.

I’ll go straight into Set 4 next without waiting for confirmation.

yes please
Perfect, Anu. I’ll keep going continuously, set by set, in the same high‑density, interview‑ready style until the entire SQL interview bank is complete.

Here comes Set 4.

Set 4 – Data Manipulation and Modification
1. Difference between DELETE, TRUNCATE, and DROP.
DELETE: Removes rows; logged; can use WHERE; slower; can be rolled back.

TRUNCATE: Removes all rows; minimal logging; cannot use WHERE; resets identity; faster.

DROP: Removes entire table structure + data; cannot be rolled back in some DBs.

2. How do you insert data into a table from another table?
sql
INSERT INTO TargetTable (col1, col2)
SELECT colA, colB
FROM SourceTable
WHERE ...
3. What is the MERGE statement and how is it used?
MERGE performs INSERT/UPDATE/DELETE in one statement based on matching conditions.
Used for upserts, syncing tables, CDC patterns.

4. How do you update multiple columns in a single statement?
sql
UPDATE Employees
SET Salary = Salary * 1.1,
    Title = 'Senior ' + Title
WHERE DeptId = 10;
5. Explain the concept of transactions in SQL.
A transaction is a logical unit of work that must commit or roll back as a whole. Ensures ACID properties.

6. What are savepoints in a transaction?
Savepoints allow partial rollbacks inside a transaction.

sql
SAVEPOINT sp1;
ROLLBACK TO sp1;
7. How do you handle error conditions in SQL transactions?
Use TRY…CATCH (SQL Server), EXCEPTION blocks (PL/pgSQL), or BEGIN…EXCEPTION (Oracle). Roll back on error, commit on success.

8. Difference between implicit and explicit transactions.
Implicit: DB auto‑starts a transaction for each statement.

Explicit: Developer uses BEGIN TRAN, COMMIT, ROLLBACK.

9. How to use the OUTPUT clause with DML statements?
Returns affected rows or values during INSERT, UPDATE, DELETE.

sql
UPDATE T
SET Salary = Salary + 1000
OUTPUT inserted.Id, inserted.Salary;
10. What happens if a transaction is not explicitly committed or rolled back?
Depends on DB:

SQL Server: auto‑rollback on session end.

Oracle: auto‑commit on disconnect.

Postgres: aborts transaction and requires explicit commit/rollback.

11. How do you perform bulk insert operations?
Use DB‑specific tools:

SQL Server: BULK INSERT, bcp, SSIS.

MySQL: LOAD DATA INFILE.

Postgres: COPY.

12. What are the different isolation levels in transactions?
Read Uncommitted

Read Committed

Repeatable Read

Serializable

(SQL Server) Snapshot

13. Explain the concept of transaction logs.
Transaction logs record all changes for recovery, rollback, and replication. Critical for durability.

14. How do you implement optimistic concurrency control?
Use row versioning or timestamps; check if row changed before update:

sql
WHERE RowVersion = @OldVersion
15. What are deadlocks and how can they be minimized?
Deadlock = two sessions waiting on each other’s locks.
Minimize by:

Access tables in same order

Keep transactions short

Use proper indexes

Avoid long‑running locks

16. Usage of INSTEAD OF triggers.
Override default DML behavior on views or tables. Often used to make non‑updatable views updatable.

17. How do you handle constraint violations during data modifications?
Validate before DML

Use TRY/CATCH

Use deferred constraints (Postgres)

Use staging tables

18. Impact of referential integrity on DELETE operations.
Foreign keys may block deletes unless using ON DELETE CASCADE or SET NULL.

19. How do you perform conditional updates based on multiple criteria?
Use CASE or multiple conditions:

sql
UPDATE T
SET Status =
  CASE WHEN Score >= 90 THEN 'A'
       WHEN Score >= 80 THEN 'B'
       ELSE 'C'
  END;
20. Explain deferred constraints.
Constraints checked at transaction commit instead of per statement. Supported in Postgres and Oracle.

21. Difference between physical and logical deletion of data.
Logical delete: Mark row as deleted (IsDeleted = 1).

Physical delete: Remove row from table.

22. How do you implement soft deletes in SQL?
Add IsDeleted column; filter queries with WHERE IsDeleted = 0.

23. How to use the RETURNING clause in PostgreSQL?
Returns inserted/updated/deleted rows:

sql
INSERT INTO T (Name) VALUES ('A')
RETURNING Id;
24. Best practices for large data modification operations.
Batch updates

Disable indexes/constraints temporarily (if safe)

Use partition switching

Avoid long transactions

Use bulk tools

25. How do you log data changes in SQL tables?
Triggers

Change Data Capture (CDC)

Change Tracking

Temporal tables

Custom audit tables

Set 5 – Indexes and Performance
1. What are indexes and why are they important?
Indexes speed up data retrieval by creating ordered structures. They reduce I/O and improve query performance.

2. Difference between clustered and non-clustered indexes.
Clustered: Sorts and stores actual data rows; one per table.

Non‑clustered: Separate structure with pointers to data; many allowed.

3. What is a composite index and when would you use one?
Index on multiple columns. Use when queries filter/sort on those columns together.

4. How do indexes impact INSERT, UPDATE, and DELETE operations?
Indexes must be updated, so DML becomes slower. More indexes = more overhead.

5. What are statistics in SQL and how do they affect query performance?
Statistics describe data distribution. Optimizer uses them to choose best plan. Outdated stats = poor performance.

6. Explain query optimization techniques in SQL.
Proper indexing

Avoid SELECT *

Use sargable predicates

Rewrite subqueries/joins

Update statistics

Partition large tables

7. What is a query execution plan and how do you analyze it?
Plan shows how DB executes a query (scans, seeks, joins). Analyze for bottlenecks: table scans, key lookups, missing indexes.

8. How do you identify and resolve performance bottlenecks in SQL queries?
Check execution plan, missing index warnings, high I/O operators, long-running queries, blocking, fragmentation.

9. What is index fragmentation and how do you address it?
Fragmentation = logical order mismatch. Fix via REORGANIZE (light) or REBUILD (heavy).

10. Explain the concept of query hints and when to use them.
Hints override optimizer decisions. Use sparingly when optimizer consistently chooses poor plans.

11. What is a covering index and when would you use it?
Index that includes all columns needed for a query → avoids lookups. Great for performance-critical queries.

12. Explain filtered indexes in SQL Server.
Index on a subset of rows using a filter predicate. Smaller, faster, ideal for sparse data.

13. How do you choose which columns to index?
Index columns used in:

WHERE

JOIN

ORDER BY

GROUP BY
Avoid indexing low‑selectivity columns.

14. What is index selectivity and why is it important?
Selectivity = fraction of unique values. High selectivity = better index performance.

15. How do you measure the performance impact of an index?
Compare execution plans, I/O stats, CPU time, and query duration before/after index creation.

16. What are bitmap indexes and when are they useful?
Bitmap indexes use bitmaps for values. Great for low‑cardinality columns in data warehouses (Oracle).

17. Explain the concept of index tuning.
Process of analyzing workload and designing optimal indexes. Includes adding, removing, or modifying indexes.

18. Impact of data types on index performance.
Smaller, fixed-length types index faster. Large types (VARCHAR(MAX), NVARCHAR(MAX)) perform poorly.

19. How do you handle index maintenance?
Regular rebuild/reorganize, update statistics, monitor fragmentation.

20. What are spatial indexes and when would you use them?
Indexes for geographic/spatial data types. Used for location-based queries (distance, intersections).

21. Explain included columns in indexes.
Non-key columns stored in index leaf nodes. Improve covering without affecting key order.

22. How do indexes affect sorting operations?
If ORDER BY matches index order, sorting is avoided → faster queries.

23. What is cardinality and how does it affect index performance?
Cardinality = number of distinct values. High cardinality → good index performance.

24. Explain the different index scan types (Index Seek vs. Index Scan).
Seek: Efficient, uses index to find specific rows.

Scan: Reads entire index/table; slower.

25. How do you diagnose index-related performance issues?
Look for:

Missing index warnings

High scans

Key lookups

Fragmentation

Outdated statistics

Anu, I’ll continue straight into Set 6 – Views and Stored Procedures next.

yes please
Absolutely, Anu — continuing smoothly with Set 6, keeping the same crisp, interview‑ready momentum. I’ll keep delivering sets back‑to‑back until the entire SQL bank is complete.

Set 6 – Views and Stored Procedures
1. What is a view in SQL and what are its benefits?
A view is a virtual table based on a query.
Benefits: abstraction, security (expose limited columns), simplified complex queries, consistency for reporting.

2. Difference between a simple view and a complex view.
Simple view: Based on one table, no aggregates, usually updatable.

Complex view: Joins, aggregates, subqueries; often not updatable.

3. What are materialized views and when would you use them?
Materialized views store query results physically.
Used for performance in OLAP/reporting, expensive joins, pre‑aggregated data.

4. What are stored procedures and their advantages?
Stored procedures are precompiled routines stored in the DB.
Advantages: performance, reusability, encapsulation, security, reduced network traffic.

5. Difference between stored procedures and functions.
Procedure: Can return multiple values, can modify data, supports transactions.

Function: Must return a value, cannot modify data (in most DBs), used in expressions.

6. Types of parameters in stored procedures.
IN: Input only.

OUT: Output only.

INOUT: Both input and output.

SQL Server: parameters are IN by default; OUTPUT keyword for OUT.

7. How do you handle errors in stored procedures?
Use TRY…CATCH (SQL Server), EXCEPTION blocks (Postgres/Oracle), log errors, rollback transactions.

8. Difference between EXEC and sp_executesql.
EXEC: Executes dynamic SQL as a string.

sp_executesql: Supports parameterized queries → safer, avoids SQL injection, better plan reuse.

9. Explain the concept of dynamic SQL.
SQL constructed and executed at runtime. Useful for dynamic filtering, table names, pivoting. Must be parameterized to avoid injection.

10. What are triggers and when would you use them?
Triggers are automatic procedures fired on DML/DDL events.
Use for auditing, enforcing complex rules, cascading logic beyond FK constraints.

11. How do you pass multiple values to a stored procedure?
Table‑valued parameters (SQL Server)

JSON/XML input

Comma‑separated strings + string‑split

Temporary tables

12. What is a table-valued parameter and how is it used?
A parameter that accepts a table as input (SQL Server).
Used for bulk inserts, passing lists, batch operations.

13. Difference between AFTER and INSTEAD OF triggers.
AFTER: Executes after DML; cannot modify the base action.

INSTEAD OF: Replaces the DML; used for updatable views.

14. How do you create indexed views in SQL Server?
Create a view with SCHEMABINDING, then create a unique clustered index on it. Used for performance on aggregated/complex queries.

15. Difference between a view and a derived table.
View: Named, reusable, stored definition.

Derived table: Inline subquery in FROM, temporary to the query.

16. How do you implement error handling in stored procedures?
Use TRY…CATCH, log errors, rollback transactions, return error codes/messages.

17. Security implications of using stored procedures.
Reduce direct table access

Allow permission delegation

Prevent SQL injection when parameterized

Must avoid dynamic SQL without parameters

18. Explain nested stored procedures.
A stored procedure calling another stored procedure.
Useful for modular design; must manage transactions carefully.

19. How do you debug stored procedures?
Use DB‑specific tools:

SQL Server: SSMS debugger, PRINT statements

Postgres: RAISE NOTICE

Oracle: DBMS_OUTPUT, debugging tools

20. Maximum number of parameters a stored procedure can have?
DB‑specific:

SQL Server: 2100 parameters

Oracle/Postgres: practically unlimited

21. How do you handle transaction management within stored procedures?
Use explicit BEGIN TRAN, COMMIT, ROLLBACK.
Avoid nested transactions; use savepoints if needed.

22. What are temporary stored procedures and when would you use them?
Procedures created in tempdb (SQL Server) with #procname.
Used for session‑specific logic or temporary automation.

23. Explain the scope of variables in stored procedures.
Local to the procedure unless passed as OUT parameters.
Variables inside dynamic SQL require separate declaration.

24. How do you implement paging in stored procedures?
Use OFFSET/FETCH or ROW_NUMBER() inside the procedure.
Return page results + total count if needed.

25. What are the limitations of views in SQL?
Often not updatable

Cannot include ORDER BY (unless TOP)

Performance issues if nested deeply

Cannot accept parameters (use inline TVFs instead)

Set 7 – Data Integrity and Constraints
1. How do you enforce data integrity in SQL?
Using constraints (PK, FK, CHECK, UNIQUE), triggers, data types, and application logic.

2. Explain referential integrity and how it's maintained.
Ensures foreign key values match primary key values in parent tables. Maintained via FK constraints with actions like CASCADE, SET NULL, RESTRICT.

3. What is cascading in foreign key constraints?
Automatic actions on child rows when parent rows change:

ON DELETE CASCADE

ON UPDATE CASCADE

4. How do you implement check constraints?
sql
ALTER TABLE Employees
ADD CONSTRAINT chk_age CHECK (Age >= 18);
5. What are DEFAULT constraints and how are they used?
Provide default values when none supplied.

sql
ALTER TABLE T ADD CONSTRAINT df_status DEFAULT 'Active' FOR Status;
6. Explain the concept of table partitioning.
Splits a table into smaller physical segments (partitions) based on a key (e.g., date). Improves performance and manageability.

7. Difference between horizontal and vertical partitioning.
Horizontal: Split rows (e.g., by date).

Vertical: Split columns (e.g., frequently used vs. rarely used).

8. How do you implement unique constraints across multiple columns?
sql
ALTER TABLE T
ADD CONSTRAINT uq_name UNIQUE (FirstName, LastName);
9. What are the limitations of constraints in SQL?
Cannot enforce complex business rules requiring procedural logic; limited to Boolean expressions.

10. How do constraints affect performance?
Improve read performance (integrity guarantees) but slow down writes due to validation.

11. What is domain integrity and how is it enforced?
Ensures values fall within valid domain: enforced via data types, CHECK constraints, and lookup tables.

12. Explain assertion constraints.
Assertions enforce conditions across multiple tables. Supported in SQL standard but rarely implemented in major DBs.

13. How do you implement conditional constraints?
Use CHECK with logical conditions:

sql
CHECK (Type <> 'Employee' OR Salary > 0)
14. Impact of constraints on DML operations.
Constraints validate data → may block invalid inserts/updates/deletes.

15. Explain how to use computed columns with constraints.
Computed columns can be persisted and indexed; constraints can be applied to them if deterministic.

16. What are filtered unique constraints in SQL Server?
Unique constraints applied only to rows matching a filter.
Useful for sparse data.

17. How do you implement complex business rules using constraints?
Use CHECK constraints, triggers, or indexed views for computed validations.

18. Difference between an enforced and a trusted constraint.
Enforced: Actively validated.

Trusted: DB assumes data is valid but may not check existing rows.

19. How do you handle constraint violations programmatically?
Use TRY/CATCH, return custom error codes, validate before DML.

20. Explain how to implement cross-database constraints.
Not directly supported; use triggers, application logic, or replication.

21. What are temporal tables and how do they maintain data integrity?
System‑versioned tables that automatically track historical changes. Maintain integrity by storing valid‑from/to timestamps.

22. How do you implement data validation beyond basic constraints?
Use triggers, stored procedures, application logic, or validation frameworks.

23. Difference between primary key and unique key constraints?
PK = unique + not null + identifies row.
Unique = ensures uniqueness but allows nulls.

24. How do you handle foreign key constraints in data warehousing?
Often disabled for load performance; integrity enforced via ETL processes.

25. Explain the concept of deferrable constraints in PostgreSQL.
Constraints checked at transaction commit instead of per statement. Useful for circular references.

Anu, I’ll continue immediately with Set 8 – Transactions and Concurrency next.

yes plz
Absolutely, Anu — continuing with the same momentum.
Here comes Set 8, crisp, complete, and interview‑ready.

Set 8 – Transactions and Concurrency
1. What are the different transaction isolation levels in SQL?
Read Uncommitted: Allows dirty reads.

Read Committed: Prevents dirty reads; default in many DBs.

Repeatable Read: Prevents dirty + non‑repeatable reads.

Serializable: Highest isolation; prevents phantoms.

Snapshot (SQL Server): Uses row versioning to avoid blocking.

2. Explain the problems that can occur in concurrent transactions.
Dirty Read: Read uncommitted data.

Non‑Repeatable Read: Same query returns different results within a transaction.

Phantom Read: New rows appear that match a previous search condition.

Lost Update: Two transactions overwrite each other’s changes.

3. What is deadlock and how can it be prevented?
Deadlock occurs when two sessions wait on each other’s locks.
Prevention:

Access objects in consistent order

Keep transactions short

Use proper indexes

Avoid user interaction inside transactions

4. Explain optimistic vs. pessimistic locking.
Optimistic: Assume no conflicts; verify at commit using row versions. Best for low‑contention systems.

Pessimistic: Lock rows before access; prevents conflicts but increases blocking.

5. What is row versioning in SQL?
Technique where old row versions are stored (tempdb in SQL Server) so readers don’t block writers. Used in snapshot isolation.

6. How does READ COMMITTED SNAPSHOT isolation level work?
Readers get a consistent snapshot of committed data using row versions. Eliminates shared locks → reduces blocking.

7. What is lock escalation and when does it occur?
DB converts many fine‑grained locks (row/page) into a table lock to reduce overhead. Happens when lock count exceeds thresholds.

8. How do you handle long‑running transactions?
Break into smaller batches

Use snapshot isolation

Reduce locking footprint

Avoid holding locks during user input

Tune queries to reduce execution time

9. Difference between implicit and explicit transactions.
Implicit: DB auto‑starts a transaction for each statement.

Explicit: Developer uses BEGIN TRAN, COMMIT, ROLLBACK.

10. Explain how to use the WITH NOLOCK hint and its implications.
WITH (NOLOCK) = dirty reads allowed; avoids blocking but risks reading uncommitted, inconsistent, or phantom data. Should be used cautiously.

11. What is lock granularity and how does it affect performance?
Granularity = size of locked resource (row, page, table).
Fine‑grained = less blocking, more overhead.
Coarse‑grained = more blocking, less overhead.

12. How do you monitor locks in a database system?
Use DMVs (SQL Server), pg_locks (Postgres), V$LOCK (Oracle), profiler tools, or monitoring dashboards.

13. What is a transaction log and how is it used in recovery?
Log records every change. Used for rollback, crash recovery, point‑in‑time restore, replication.

14. Explain the concept of two-phase commit.
Protocol for distributed transactions:

Phase 1: Prepare — participants vote commit/abort.

Phase 2: Commit — coordinator instructs commit if all voted yes.

15. How do you handle distributed transactions?
Use transaction coordinators (MSDTC, XA), two‑phase commit, or application‑level compensation logic.

16. What is the impact of isolation levels on concurrency and throughput?
Higher isolation = fewer anomalies but more blocking.
Lower isolation = more concurrency but more anomalies.

17. Explain how row versioning affects tempdb usage.
Snapshot isolation stores row versions in tempdb. Heavy versioning → tempdb growth and potential contention.

18. What are the different lock types (shared, exclusive, update, etc.)?
Shared (S): Read locks.

Exclusive (X): Write locks.

Update (U): Prevents deadlocks during updates.

Intent locks: Indicate future locks at lower granularity.

Schema locks: Protect metadata.

19. How do you handle blocking transactions?
Identify blockers

Kill long blockers (carefully)

Use snapshot isolation

Tune queries

Add indexes

Reduce transaction scope

20. What is a transaction savepoint and how is it used?
A marker inside a transaction allowing partial rollback.

sql
SAVEPOINT sp1;
ROLLBACK TO sp1;
21. Explain the concept of transaction chaining.
In some DBs, after a transaction ends, the next statement automatically starts a new one. Rarely used today.

22. How do you implement retry logic for handling deadlocks?
Catch deadlock error → wait → retry transaction with exponential backoff.

23. What is a transaction schedule and serializability?
Schedule = order of operations from concurrent transactions.
Serializable = schedule equivalent to some serial execution → no anomalies.

24. How do you implement optimistic concurrency in applications?
Use row version/timestamp columns:

sql
UPDATE T
SET ...
WHERE Id = @Id AND RowVersion = @OldVersion;
If 0 rows affected → concurrency conflict.

25. Best practices for transaction management in high‑concurrency environments.
Keep transactions short

Use proper isolation levels

Use snapshot isolation when appropriate

Avoid locking large ranges

Index filtering columns

Avoid user interaction inside transactions

Anu, I’ll continue immediately with Set 9 – Database Design.

yes - continue till end
Absolutely, Anu — I’m continuing straight through to the end, no pauses, no meta, just clean, dense, interview‑ready content.

Here comes Set 9, and I’ll keep going set after set until the entire SQL bank is complete.

Set 9 – Database Design
1. What are entity-relationship diagrams (ERDs) and how are they used?
ERDs visually model entities, attributes, and relationships. Used for conceptual and logical database design.

2. Explain the different types of relationships (1‑1, 1‑many, many‑many).
1:1 — One row relates to one row.

1:M — One row relates to many rows.

M:N — Many rows relate to many rows; implemented via junction table.

3. How do you implement a many-to-many relationship in SQL?
Use a junction table with two foreign keys referencing parent tables.

4. What is cardinality in database design?
Defines the numerical relationship between entities (1:1, 1:M, M:N).

5. Explain database normalization and its levels.
Process to reduce redundancy.
1NF → atomic columns
2NF → no partial dependencies
3NF → no transitive dependencies
BCNF → every determinant is a candidate key

6. Advantages and disadvantages of normalization.
Advantages: integrity, reduced redundancy, easier updates.
Disadvantages: more joins, slower reads for analytics.

7. When would you denormalize a database schema?
When read performance is critical, joins are expensive, or OLAP/reporting workloads dominate.

8. Design for performance vs. design for maintainability.
Performance → denormalization, indexing, partitioning.
Maintainability → normalization, modular schema, clear relationships.

9. Best practices for naming conventions.
Use consistent, descriptive names; avoid spaces; singular table names; prefix constraints (PK_, FK_, UQ_).

10. How do you handle temporal data in database design?
Use valid‑from/to columns, temporal tables, or history tables.

11. What is a star schema vs. snowflake schema?
Star: Fact table + denormalized dimensions.

Snowflake: Dimensions normalized into sub‑tables.

12. How do you model hierarchical data?
Adjacency list (ParentId), nested sets, closure tables, or DB‑specific types (hierarchyid).

13. What are slowly changing dimensions (SCDs) and their types?
Techniques for tracking dimension changes:

Type 1: overwrite

Type 2: add new row with history

Type 3: store previous value

14. Explain database sharding.
Horizontal partitioning across servers for scalability.

15. What is indexing strategy in database design?
Choose indexes based on workload: WHERE, JOIN, ORDER BY columns; avoid over‑indexing.

16. How do you handle historical vs. current data?
Use history tables, temporal tables, partitions, or archive databases.

17. Surrogate keys vs. natural keys.
Surrogate: artificial (identity, GUID).
Natural: real‑world attribute.
Surrogates simplify joins and stability.

18. How do you design databases for scalability?
Partitioning, sharding, caching, denormalization, read replicas.

19. Impact of normalization on query performance.
More joins → slower reads; faster writes and updates.

20. How to model polymorphic associations?
Use:

Single table with type + id

Separate tables + union view

Bridge tables

21. Multi-tenant database design patterns.
Shared DB, shared schema

Shared DB, separate schema

Separate DB per tenant

22. Design for data warehousing vs. operational databases.
OLTP → normalized, transactional.
OLAP → denormalized, star/snowflake, historical.

23. Considerations for international data.
Unicode, time zones, locale formats, currency, collation.

24. How to handle soft deletes at the database level?
Add IsDeleted column; filter queries; optionally use views.

25. Best practices for designing audit trails.
Use audit tables, triggers, CDC, temporal tables; store who/when/what changed.

Set 10 – Advanced SQL Concepts
1. What are CTEs and how are they different from temporary tables?
CTEs are inline, scoped to one query; temp tables persist for session and can be indexed.

2. Explain PIVOT and UNPIVOT.
PIVOT turns rows → columns.
UNPIVOT turns columns → rows.

3. What are JSON functions in SQL and how are they used?
Functions to parse, query, and modify JSON (JSON_VALUE, JSON_QUERY, JSON_TABLE). Used for semi‑structured data.

4. How do you handle hierarchical data in SQL?
Recursive CTEs, adjacency lists, nested sets, closure tables.

5. Deterministic vs. non-deterministic functions.
Deterministic: same input → same output.
Non‑deterministic: depends on state/time (GETDATE()).

6. How to use ROLLUP, CUBE, GROUPING SETS.
Generate multiple aggregations in one query: subtotals, grand totals, all combinations.

7. Uses of APPLY operator (CROSS APPLY, OUTER APPLY).
Apply table‑valued functions or correlated subqueries to each row of outer table.

8. How to implement paging in SQL queries?
Use OFFSET/FETCH, LIMIT, or ROW_NUMBER().

9. Differences between SQL in MySQL, PostgreSQL, SQL Server, Oracle.
Syntax differences in:

Window functions

MERGE support

JSON functions

Procedural languages

Sequences/identity

Limit/offset syntax

10. Explain SQL injection and how to prevent it.
Injection = malicious SQL via input.
Prevent via parameterized queries, stored procedures, input validation.

11. What are XML functions in SQL?
Functions to query/modify XML (nodes, value, exist, query). Used for structured documents.

12. Explain OLAP cubes and MDX.
OLAP cubes store multidimensional aggregated data. MDX is query language for cubes.

13. How to implement full-text search in SQL?
Use full‑text indexes; query with CONTAINS, FREETEXT, MATCH AGAINST.

14. What are spatial data types and functions?
Types for geometry/geography; functions for distance, intersection, containment.

15. Explain MERGE for upserting data.
MERGE matches source/target and performs INSERT/UPDATE/DELETE in one statement.

16. Table partitioning strategies.
Range, list, hash, composite partitioning.

17. How to implement change data capture (CDC)?
Enable CDC; DB logs changes to tracking tables for ETL/replication.

18. What is columnstore indexing and when to use it?
Column‑oriented storage for analytics; great for large scans and aggregations.

19. Memory-optimized tables and natively compiled procedures.
In‑memory OLTP (SQL Server) for ultra‑low latency; reduces locking/latching.

20. How to implement data masking for sensitive information?
Dynamic data masking, encryption, views, or custom masking functions.

21. Advantages and limitations of SQL for ETL.
Advantages: set‑based, scalable, close to data.
Limitations: complex transformations harder than in ETL tools.

22. How to use SEQUENCE objects instead of identity columns?
Create sequence → use NEXT VALUE FOR seq in inserts.

23. What are temporal tables and how are they implemented?
System‑versioned tables with history table tracking changes automatically.

24. How do you handle versioning in database schemas?
Use migration tools (Flyway, Liquibase), version tables, semantic versioning.

25. Benefits of using Graph tables in SQL Server.
Model relationships as nodes/edges; easier traversal for graph‑like data.

Set 11 – Data Types and Functions
1. Common data types in SQL.
INT, BIGINT, DECIMAL, FLOAT, VARCHAR, NVARCHAR, DATE, DATETIME, BOOLEAN, BLOB, JSON.

2. Difference between VARCHAR and CHAR.
VARCHAR = variable length; CHAR = fixed length. CHAR wastes space but is faster for fixed‑size data.

3. Handling date and time data.
Use DATE, TIME, DATETIME, TIMESTAMP, INTERVAL; use functions like DATEADD, DATEDIFF, EXTRACT.

4. String manipulation functions.
SUBSTRING, LEFT, RIGHT, TRIM, REPLACE, CONCAT, UPPER, LOWER.

5. Numeric and mathematical functions.
ABS, ROUND, CEILING, FLOOR, POWER, SQRT, RAND.

6. Converting between data types.
Use CAST, CONVERT, TRY_CAST, TO_CHAR, TO_DATE.

7. What are collations?
Rules for sorting/comparing strings; affect case sensitivity and language rules.

8. Handling NULL values in functions.
Use COALESCE, ISNULL, NULLIF, default values.

9. User-defined functions (UDFs).
Custom functions returning scalar or table values; used for reusable logic.

10. Aggregate functions and limitations.
SUM, AVG, COUNT, MIN, MAX.
Cannot reference non‑aggregated columns without GROUP BY.

11. Difference between VARCHAR and NVARCHAR.
NVARCHAR stores Unicode; uses 2 bytes per character.

12. Handling Unicode data.
Use NVARCHAR, UTF‑8 collations, proper encoding.

13. CAST vs. CONVERT.
CAST = ANSI standard.
CONVERT = DB‑specific formatting options.

14. DATETIME vs. DATETIME2.
DATETIME2 has larger range, higher precision, smaller storage.

15. Handling time zone conversions.
Use AT TIME ZONE, CONVERT_TZ, or store UTC + offset.

16. Limitations of numeric data types.
Precision/scale limits; overflow; rounding errors in floating types.

17. FORMAT function.
Converts values to formatted strings; slower than CONVERT.

18. Binary data types.
VARBINARY, BLOB; used for files, images, encrypted data.

19. Complex mathematical calculations.
Use built‑in math functions or CLR extensions.

20. Scalar vs. table-valued functions.
Scalar returns one value; TVF returns a table.

21. CLR integration for custom data types.
Allows writing .NET code for custom types/functions (SQL Server).

22. hierarchyid and spatial data types.
hierarchyid = hierarchical paths.
Spatial = geometry/geography.

23. Storage optimization with data compression.
Row/page compression reduces storage and I/O.

24. Differences between ISNULL, COALESCE, NULLIF.
ISNULL = 2 args, DB‑specific.
COALESCE = ANSI, multiple args.
NULLIF = returns NULL if equal.

25. Window functions with different data types.
Work with numeric, date, string types; frame definitions vary by type.

Set 12 – Security and Administration
1. How do you manage user permissions in SQL?
Use roles, GRANT/REVOKE, schema‑level permissions, least privilege.

2. Authentication vs. authorization.
Authentication = identity verification.
Authorization = permission to access resources.

3. Concept of roles in SQL security.
Roles group permissions; assign users to roles for easier management.

4. Encrypting sensitive data.
Use TDE, column‑level encryption, Always Encrypted, hashing, key management.

5. Best practices for backup and recovery.
Full, differential, log backups; test restores; offsite storage; retention policies.

6. Monitoring database performance.
Use DMVs, performance counters, query store, monitoring tools.

7. Tools for SQL query optimization.
Execution plans, profiler, tuning advisor, query store.

8. Implementing row-level security.
Use security predicates, inline functions, policies (SQL Server), RLS in Postgres.

9. Database auditing techniques.
Triggers, audit logs, CDC, extended events, server audits.

10. Handling database migrations and schema changes.
Use versioning tools (Flyway, Liquibase), scripts, CI/CD pipelines.

11. Principle of least privilege.
Give minimum permissions required; avoid sysadmin roles.

12. Transparent Data Encryption (TDE).
Encrypts data at rest (files, backups).

13. Login security vs. user security.
Login = server‑level.
User = database‑level.

14. Types of database backups and recovery models.
Full, differential, log.
Recovery models: simple, full, bulk‑logged.

15. Always-on availability groups.
High availability + read scale; synchronous/asynchronous replicas.

16. Database mirroring vs. replication.
Mirroring = HA.
Replication = data distribution (snapshot, transactional, merge).

17. Handling database corruption.
Use DBCC CHECKDB, restore from backup, repair options.

18. Best practices for SQL Server log management.
Regular log backups, monitor size, avoid long transactions.

19. Disaster recovery for SQL databases.
Backups, log shipping, geo‑replication, failover clusters.

20. Security considerations for stored procedures and views.
Use parameterization, avoid dynamic SQL, restrict permissions.

21. Data masking and column-level encryption.
Mask sensitive fields; encrypt columns with keys/certificates.

22. Contained vs. traditional databases.
Contained DBs store users inside DB; reduce dependency on server logins.

23. Extended events for monitoring.
Lightweight event tracing for performance and troubleshooting.

24. Resource governor features.
Control CPU/memory for workloads; useful in multi‑tenant systems.

25. Managing large-scale database deployments.
Automation, CI/CD, partitioning, sharding, monitoring, version control.
```
