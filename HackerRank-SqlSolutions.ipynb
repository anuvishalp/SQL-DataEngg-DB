```python
CREATE OR REPLACE PROCEDURE calc(a NUMBER, b NUMBER, op VARCHAR2) AS
  result NUMBER;
BEGIN
  CASE op
    WHEN '+' THEN result := a + b;
    WHEN '-' THEN result := a - b;
    WHEN '*' THEN result := a * b;
    WHEN '/' THEN result := a / b;
  END CASE;

  DBMS_OUTPUT.PUT_LINE('Result: ' || result);
END;
/

BEGIN
  calc(20, 4, '/');
END;
/
```

```python
###High-level explanation of a Mario-style game -> A classic Mario-style game is a 2D side-scrolling platformer with:
"""
Core mechanics:

Run & jump: Player moves left/right and jumps with gravity pulling them down.

Platforms & collisions: Solid tiles (ground, bricks, pipes) that the player can stand on, bump into, or break.

Enemies: Simple AI (walk back and forth, hurt on contact, can be defeated by jumping on them).

Collectibles: Coins, power-ups (like mushrooms) that modify player state (bigger, extra life, etc.).

Level progression: Reach a goal (flag, door, pipe) to finish the level.

Below are two separate projects:

A Python mini-platformer (using pygame)

An SQL project modeling Mario-style game data (players, levels, enemies, coins, scores)
"""
Project 1 – Python mini Mario-style platformer (pygame)
Minimal but structured; you can expand physics, animations, enemies, etc.

Folder structure
text
        mario_python/
        ├── main.py
        ├── settings.py
        └── entities.py
        settings.py

python
# settings.py

              # Window
              WIDTH = 800
              HEIGHT = 600
              TITLE = "Mini Mario Clone"

              # Colors
              WHITE = (255, 255, 255)
              BLUE = (0, 0, 255)
              GREEN = (0, 200, 0)
              BLACK = (0, 0, 0)

              # Physics
              GRAVITY = 0.5
              PLAYER_SPEED = 5
              JUMP_STRENGTH = -12

              # Level layout (simple tile map)
              # 'X' = solid block, '.' = empty
              LEVEL_MAP = [
                  "................................",
                  "................................",
                  "................................",
                  "................................",
                  "......................XXX.......",
                  "......................XXX.......",
                  ".............XXX................",
                  "............................XXXX",
                  "XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX",
              ]
              TILE_SIZE = 32

------------------------------
entities.py
python
# entities.py
              import pygame
              from settings import GRAVITY, PLAYER_SPEED, JUMP_STRENGTH, TILE_SIZE, BLUE, GREEN

              class Player(pygame.sprite.Sprite):
                  def __init__(self, x, y, platforms):
                      super().__init__()
                      self.image = pygame.Surface((TILE_SIZE, TILE_SIZE * 1.5))
                      self.image.fill(BLUE)
                      self.rect = self.image.get_rect(topleft=(x, y))

                      self.vel_x = 0
                      self.vel_y = 0
                      self.on_ground = False
                      self.platforms = platforms

                  def handle_input(self, keys):
                      self.vel_x = 0
                      if keys[pygame.K_LEFT]:
                          self.vel_x = -PLAYER_SPEED
                      if keys[pygame.K_RIGHT]:
                          self.vel_x = PLAYER_SPEED
                      if keys[pygame.K_SPACE] and self.on_ground:
                          self.vel_y = JUMP_STRENGTH
                          self.on_ground = False

                  def apply_gravity(self):
                      self.vel_y += GRAVITY
                      if self.vel_y > 20:
                          self.vel_y = 20

                  def horizontal_collisions(self):
                      self.rect.x += self.vel_x
                      for platform in self.platforms:
                          if self.rect.colliderect(platform.rect):
                              if self.vel_x > 0:
                                  self.rect.right = platform.rect.left
                              elif self.vel_x < 0:
                                  self.rect.left = platform.rect.right

                  def vertical_collisions(self):
                      self.rect.y += self.vel_y
                      self.on_ground = False
                      for platform in self.platforms:
                          if self.rect.colliderect(platform.rect):
                              if self.vel_y > 0:
                                  self.rect.bottom = platform.rect.top
                                  self.vel_y = 0
                                  self.on_ground = True
                              elif self.vel_y < 0:
                                  self.rect.top = platform.rect.bottom
                                  self.vel_y = 0

                  def update(self, keys):
                      self.handle_input(keys)
                      self.apply_gravity()
                      self.horizontal_collisions()
                      self.vertical_collisions()


              class Platform(pygame.sprite.Sprite):
                  def __init__(self, x, y, width=TILE_SIZE, height=TILE_SIZE):
                      super().__init__()
                      self.image = pygame.Surface((width, height))
                      self.image.fill(GREEN)
                      self.rect = self.image.get_rect(topleft=(x, y))
-------------------------------------
main.py
python
# main.py

              import pygame
              from settings import (
                  WIDTH, HEIGHT, TITLE, WHITE, BLACK,
                  LEVEL_MAP, TILE_SIZE
              )
              from entities import Player, Platform

              def build_level(level_map):
                  platforms = pygame.sprite.Group()
                  for row_idx, row in enumerate(level_map):
                      for col_idx, cell in enumerate(row):
                          if cell == "X":
                              x = col_idx * TILE_SIZE
                              y = row_idx * TILE_SIZE
                              platforms.add(Platform(x, y))
                  return platforms

              def main():
                  pygame.init()
                  screen = pygame.display.set_mode((WIDTH, HEIGHT))
                  pygame.display.set_caption(TITLE)
                  clock = pygame.time.Clock()

                  platforms = build_level(LEVEL_MAP)
                  player = Player(100, 100, platforms)
                  all_sprites = pygame.sprite.Group(platforms, player)

                  camera_x = 0

                  running = True
                  while running:
                      dt = clock.tick(60) / 1000  # not used yet, but handy for future

                      for event in pygame.event.get():
                          if event.type == pygame.QUIT:
                              running = False

                      keys = pygame.key.get_pressed()
                      player.update(keys)

                      # Simple side-scrolling camera following player
                      camera_x = -player.rect.centerx + WIDTH // 2

                      screen.fill(WHITE)

                      for sprite in all_sprites:
                          offset_rect = sprite.rect.move(camera_x, 0)
                          screen.blit(sprite.image, offset_rect)

                      # Simple HUD
                      font = pygame.font.SysFont(None, 24)
                      text = font.render("Mini Mario Clone - Arrows + Space to jump", True, BLACK)
                      screen.blit(text, (10, 10))

                      pygame.display.flip()

                  pygame.quit()

              if __name__ == "__main__":
                  main()
------------------
To run: pip install pygame, then python main.py.

You can extend this with enemies, coins, power-ups, multiple levels, and a proper tile editor.
------------------------------
#### Project 2 – SQL project for a Mario-style game backend ->This models players, levels, enemies, coins, and scores—useful for analytics, leaderboards, or saving progress.

Folder structure
text
      mario_sql/
      ├── schema.sql
      ├── sample_data.sql
      └── queries.sql

schema.sql
sql
              -- schema.sql

              -- Players of the game
              CREATE TABLE players (
                  player_id      SERIAL PRIMARY KEY,
                  username       VARCHAR(50) UNIQUE NOT NULL,
                  created_at     TIMESTAMP NOT NULL DEFAULT NOW()
              );

              -- Levels in the game
              CREATE TABLE levels (
                  level_id       SERIAL PRIMARY KEY,
                  world_number   INT NOT NULL,
                  level_number   INT NOT NULL,
                  name           VARCHAR(100),
                  difficulty     VARCHAR(20),
                  UNIQUE (world_number, level_number)
              );

              -- Enemies that can appear in levels
              CREATE TABLE enemy_types (
                  enemy_type_id  SERIAL PRIMARY KEY,
                  name           VARCHAR(50) NOT NULL,
                  description    TEXT,
                  base_points    INT NOT NULL
              );

              -- Coins / collectibles
              CREATE TABLE collectible_types (
                  collectible_type_id SERIAL PRIMARY KEY,
                  name                VARCHAR(50) NOT NULL,
                  description         TEXT,
                  base_points         INT NOT NULL
              );

              -- Instances of enemies in a level (for analytics / design)
              CREATE TABLE level_enemies (
                  level_enemy_id  SERIAL PRIMARY KEY,
                  level_id        INT NOT NULL REFERENCES levels(level_id),
                  enemy_type_id   INT NOT NULL REFERENCES enemy_types(enemy_type_id),
                  position_x      INT,
                  position_y      INT
              );

              -- Instances of collectibles in a level
              CREATE TABLE level_collectibles (
                  level_collectible_id SERIAL PRIMARY KEY,
                  level_id             INT NOT NULL REFERENCES levels(level_id),
                  collectible_type_id  INT NOT NULL REFERENCES collectible_types(collectible_type_id),
                  position_x           INT,
                  position_y           INT
              );

              -- Player runs (attempts) of a level
              CREATE TABLE level_runs (
                  run_id          SERIAL PRIMARY KEY,
                  player_id       INT NOT NULL REFERENCES players(player_id),
                  level_id        INT NOT NULL REFERENCES levels(level_id),
                  started_at      TIMESTAMP NOT NULL DEFAULT NOW(),
                  finished_at     TIMESTAMP,
                  completed       BOOLEAN NOT NULL DEFAULT FALSE,
                  total_score     INT NOT NULL DEFAULT 0,
                  total_coins     INT NOT NULL DEFAULT 0,
                  total_deaths    INT NOT NULL DEFAULT 0
              );

              -- Detailed events during a run (defeating enemies, collecting coins, dying, etc.)
              CREATE TABLE run_events (
                  event_id        SERIAL PRIMARY KEY,
                  run_id          INT NOT NULL REFERENCES level_runs(run_id),
                  event_time      TIMESTAMP NOT NULL DEFAULT NOW(),
                  event_type      VARCHAR(50) NOT NULL, -- 'ENEMY_DEFEATED', 'COIN_COLLECTED', 'DEATH', etc.
                  enemy_type_id   INT REFERENCES enemy_types(enemy_type_id),
                  collectible_type_id INT REFERENCES collectible_types(collectible_type_id),
                  points_earned   INT NOT NULL DEFAULT 0,
                  extra_info      JSONB
              );
---------------------------------
sample_data.sql
sql
-- sample_data.sql

              INSERT INTO players (username) VALUES
              ('mario_fan'),
              ('speedrunner'),
              ('casual_player');

              INSERT INTO levels (world_number, level_number, name, difficulty) VALUES
              (1, 1, 'World 1-1', 'Easy'),
              (1, 2, 'World 1-2', 'Easy'),
              (2, 1, 'World 2-1', 'Medium');

              INSERT INTO enemy_types (name, description, base_points) VALUES
              ('Goomba', 'Walks back and forth, defeated by jumping on it', 100),
              ('Koopa', 'Turtle-like enemy, shell can be kicked', 200);

              INSERT INTO collectible_types (name, description, base_points) VALUES
              ('Coin', 'Standard coin', 10),
              ('1-Up Mushroom', 'Grants extra life', 0);

              -- Example level content
              INSERT INTO level_enemies (level_id, enemy_type_id, position_x, position_y) VALUES
              (1, 1, 200, 400),
              (1, 1, 400, 400),
              (1, 2, 600, 400);

              INSERT INTO level_collectibles (level_id, collectible_type_id, position_x, position_y) VALUES
              (1, 1, 150, 350),
              (1, 1, 250, 350),
              (1, 2, 800, 350);

              -- Example run
              INSERT INTO level_runs (player_id, level_id, completed, total_score, total_coins, total_deaths)
              VALUES (1, 1, TRUE, 230, 3, 0);

              INSERT INTO run_events (run_id, event_type, enemy_type_id, points_earned)
              VALUES
              (1, 'ENEMY_DEFEATED', 1, 100),
              (1, 'ENEMY_DEFEATED', 1, 100),
              (1, 'COIN_COLLECTED', NULL, 10),
              (1, 'COIN_COLLECTED', NULL, 10),
              (1, 'COIN_COLLECTED', NULL, 10);
----------------------------
queries.sql – example analytics/queries
sql
-- queries.sql

-- 1. Leaderboard: top players by total score across all runs
            SELECT
                p.username,
                SUM(r.total_score) AS total_score
            FROM players p
            JOIN level_runs r ON r.player_id = p.player_id
            GROUP BY p.username
            ORDER BY total_score DESC
            LIMIT 10;

-- 2. Average completion rate per level
          SELECT
              l.world_number,
              l.level_number,
              l.name,
              COUNT(*) AS attempts,
              SUM(CASE WHEN completed THEN 1 ELSE 0 END) AS completions,
              ROUND(
                  100.0 * SUM(CASE WHEN completed THEN 1 ELSE 0 END) / NULLIF(COUNT(*), 0),
                  2
              ) AS completion_rate_pct
          FROM levels l
          LEFT JOIN level_runs r ON r.level_id = l.level_id
          GROUP BY l.level_id
          ORDER BY l.world_number, l.level_number;

-- 3. Average score per completed run for each player
            SELECT
                p.username,
                AVG(r.total_score) AS avg_score_completed_runs
            FROM players p
            JOIN level_runs r ON r.player_id = p.player_id
            WHERE r.completed = TRUE
            GROUP BY p.username
            ORDER BY avg_score_completed_runs DESC;

-- 4. Most common enemy type defeated
            SELECT
                e.name,
                COUNT(*) AS times_defeated
            FROM run_events ev
            JOIN enemy_types e ON ev.enemy_type_id = e.enemy_type_id
            WHERE ev.event_type = 'ENEMY_DEFEATED'
            GROUP BY e.name
            ORDER BY times_defeated DESC;

-- 5. Total coins collected per player
            SELECT
                p.username,
                SUM(r.total_coins) AS coins_collected
            FROM players p
            JOIN level_runs r ON r.player_id = p.player_id
            GROUP BY p.username
            ORDER BY coins_collected DESC;
```

```python
## MEDIUM-> data engg Interview question
Problem Statement
You are given a table toll_log that records every vehicle crossing a toll plaza.

Each record represents one crossing.

Toll Rules
Motorcycle → Free
Car → ₹40
Bus → ₹70
Truck → ₹80
If the same vehicle returns within 4 hours, apply a reduced toll:

Car → ₹20
Bus → ₹30
Truck → ₹40
Otherwise, charge the full toll again.

Goal
👉 Find the total toll collected per day.

Key Insights (What Interviewers Expect)

To solve this correctly and efficiently, we must:

          Identify the previous crossing per vehicle
          Calculate the time difference
          Decide full vs reduced toll
          Aggregate by day
This is a window-function-driven pricing problem, not a join problem.

SQL Solution
WITH base AS (
    SELECT
        txn_id,
        vehicle_no,
        vehicle_type,
        crossing_time,
        DATE(crossing_time) AS crossing_date,
        LAG(crossing_time) OVER (
            PARTITION BY vehicle_no
            ORDER BY crossing_time
        ) AS prev_crossing_time
    FROM toll_log
),

toll_calc AS (
    SELECT *,
        CASE
            WHEN vehicle_type = 'motorcycle' THEN 0

            WHEN prev_crossing_time IS NOT NULL
             AND TIMESTAMPDIFF(HOUR, prev_crossing_time, crossing_time) <= 4
            THEN CASE
                    WHEN vehicle_type = 'car' THEN 20
                    WHEN vehicle_type = 'bus' THEN 30
                    WHEN vehicle_type = 'truck' THEN 40
                 END

            ELSE CASE
                    WHEN vehicle_type = 'car' THEN 40
                    WHEN vehicle_type = 'bus' THEN 70
                    WHEN vehicle_type = 'truck' THEN 80
                 END
        END AS toll_amount
    FROM base
)

SELECT
    crossing_date,
    SUM(toll_amount) AS total_toll
FROM toll_calc
GROUP BY crossing_date;
----------------------------
Why this SQL is optimal

          Avoids self-joins
          Single linear scan
          Window function gives O(n) complexity
          Works on Spark SQL, Snowflake, BigQuery

PySpark Solution

                  from pyspark.sql import functions as F
                  from pyspark.sql.window import Window

                  w = Window.partitionBy("vehicle_no").orderBy("crossing_time")

                  df2 = df.withColumn(
                      "prev_crossing_time",
                      F.lag("crossing_time").over(w)
                  )

                  df3 = df2.withColumn(
                      "hours_diff",
                      (F.col("crossing_time").cast("long") -
                      F.col("prev_crossing_time").cast("long")) / 3600
                  )

                  df4 = df3.withColumn(
                      "toll_amount",
                      F.when(F.col("vehicle_type") == "motorcycle", 0)
                      .when(
                          (F.col("hours_diff") <= 4) & F.col("prev_crossing_time").isNotNull(),
                          F.when(F.col("vehicle_type") == "car", 20)
                          .when(F.col("vehicle_type") == "bus", 30)
                          .when(F.col("vehicle_type") == "truck", 40)
                      )
                      .otherwise(
                          F.when(F.col("vehicle_type") == "car", 40)
                          .when(F.col("vehicle_type") == "bus", 70)
                          .when(F.col("vehicle_type") == "truck", 80)
                      )
                  )

                  result = df4.groupBy(
                      F.to_date("crossing_time").alias("crossing_date")
                  ).agg(
                      F.sum("toll_amount").alias("total_toll")
                  )

                  result.show()
```

```python
1. JOIN‑based challenges (with solutions)
Challenge J1: List countries with their capital city name
sql
SELECT c.name AS country, ci.name AS capital
FROM country c
JOIN city ci
  ON c.id = ci.country_id
 AND ci.is_capital = TRUE;
Challenge J2: Countries and number of cities
sql
SELECT c.name AS country, COUNT(ci.id) AS city_count
FROM country c
LEFT JOIN city ci
  ON c.id = ci.country_id
GROUP BY c.name;
Challenge J3: Countries with at least one city over 1M population
sql
SELECT DISTINCT c.name
FROM country c
JOIN city ci
  ON c.id = ci.country_id
WHERE ci.population > 1000000;
Challenge J4: Countries and their continents
sql
SELECT c.name AS country, ct.name AS continent
FROM country c
JOIN continent ct
  ON c.continent = ct.name;
Challenge J5: Countries and their official languages
sql
SELECT c.name AS country, l.language_name
FROM country c
JOIN language l
  ON c.id = l.country_id
WHERE l.is_official = TRUE;
Challenge J6: City, its country, and continent
sql
SELECT ci.name AS city, c.name AS country, c.continent
FROM city ci
JOIN country c
  ON ci.country_id = c.id;
Challenge J7: Countries with no cities (anti‑join pattern)
sql
SELECT c.name
FROM country c
LEFT JOIN city ci
  ON c.id = ci.country_id
WHERE ci.id IS NULL;
Challenge J8: Countries with more than 5 cities
sql
SELECT c.name, COUNT(ci.id) AS city_count
FROM country c
JOIN city ci
  ON c.id = ci.country_id
GROUP BY c.name
HAVING COUNT(ci.id) > 5;
Challenge J9: Countries and count of official vs non‑official languages
sql
SELECT c.name,
       SUM(CASE WHEN l.is_official THEN 1 ELSE 0 END) AS official_langs,
       SUM(CASE WHEN l.is_official THEN 0 ELSE 1 END) AS non_official_langs
FROM country c
JOIN language l
  ON c.id = l.country_id
GROUP BY c.name;
Challenge J10: Top city per country by population (join + window)
sql
SELECT country_name, city_name, population
FROM (
    SELECT c.name AS country_name,
           ci.name AS city_name,
           ci.population,
           ROW_NUMBER() OVER (PARTITION BY c.id ORDER BY ci.population DESC) AS rn
    FROM country c
    JOIN city ci
      ON c.id = ci.country_id
) t
WHERE rn = 1;
2. Subquery challenges (with solutions)
Challenge S1: Countries with population above global average
sql
SELECT name, population
FROM country
WHERE population >
      (SELECT AVG(population) FROM country);
Challenge S2: Countries whose population is greater than any city in them
sql
SELECT c.name
FROM country c
WHERE c.population >
      (SELECT MAX(ci.population)
       FROM city ci
       WHERE ci.country_id = c.id);
Challenge S3: Countries that have at least one official language
sql
SELECT name
FROM country
WHERE id IN (
    SELECT DISTINCT country_id
    FROM language
    WHERE is_official = TRUE
);
Challenge S4: Countries with more than 3 cities (subquery version)
sql
SELECT name
FROM country
WHERE id IN (
    SELECT country_id
    FROM city
    GROUP BY country_id
    HAVING COUNT(*) > 3
);
Challenge S5: Cities in the top 5 most populated countries
sql
SELECT ci.name, ci.population
FROM city ci
WHERE ci.country_id IN (
    SELECT id
    FROM country
    ORDER BY population DESC
    LIMIT 5
);
Challenge S6: Continents with total population > 1B
sql
SELECT DISTINCT continent
FROM country
WHERE continent IN (
    SELECT continent
    FROM country
    GROUP BY continent
    HAVING SUM(population) > 1000000000
);
Challenge S7: Countries whose capital city has population > 1M
sql
SELECT c.name
FROM country c
WHERE c.id IN (
    SELECT country_id
    FROM city
    WHERE is_capital = TRUE
      AND population > 1000000
);
Challenge S8: Countries with no official language (NOT IN)
sql
SELECT name
FROM country
WHERE id NOT IN (
    SELECT DISTINCT country_id
    FROM language
    WHERE is_official = TRUE
);
Challenge S9: Rank continents by total population using subquery
sql
SELECT continent, total_pop
FROM (
    SELECT continent,
           SUM(population) AS total_pop,
           RANK() OVER (ORDER BY SUM(population) DESC) AS rnk
    FROM country
    GROUP BY continent
) t;
Challenge S10: Countries whose population is in the top 10% (percentile via subquery)
sql
SELECT name, population
FROM country
WHERE population >= (
    SELECT PERCENTILE_CONT(0.9) WITHIN GROUP (ORDER BY population)
    FROM country
);
(Use dialect‑specific percentile function.)

3. Window function challenge pack (with solutions)
Challenge W1: Rank countries by population within each continent
sql
SELECT name, continent, population,
       RANK() OVER (PARTITION BY continent ORDER BY population DESC) AS pop_rank
FROM country;
Challenge W2: Running total of population ordered by population
sql
SELECT name, population,
       SUM(population) OVER (ORDER BY population DESC) AS running_total
FROM country;
Challenge W3: Percentage of world population per country
sql
SELECT name, population,
       100.0 * population /
       SUM(population) OVER () AS pct_of_world
FROM country;
Challenge W4: City population share within its country
sql
SELECT ci.name AS city,
       c.name AS country,
       ci.population,
       100.0 * ci.population /
       SUM(ci.population) OVER (PARTITION BY c.id) AS pct_of_country
FROM city ci
JOIN country c
  ON ci.country_id = c.id;
Challenge W5: Moving average of population (top 10 by population)
sql
SELECT name, population,
       AVG(population) OVER (ORDER BY population DESC
                             ROWS BETWEEN 2 PRECEDING AND CURRENT ROW) AS moving_avg
FROM country;
Challenge W6: Dense rank of cities by population per country
sql
SELECT c.name AS country, ci.name AS city, ci.population,
       DENSE_RANK() OVER (PARTITION BY c.id ORDER BY ci.population DESC) AS city_rank
FROM country c
JOIN city ci
  ON c.id = ci.country_id;
Challenge W7: Difference from continent average population
sql
SELECT name, continent, population,
       population - AVG(population) OVER (PARTITION BY continent) AS diff_from_cont_avg
FROM country;
Challenge W8: Top 3 cities per continent by population
sql
SELECT *
FROM (
    SELECT ct.name AS continent,
           c.name AS country,
           ci.name AS city,
           ci.population,
           ROW_NUMBER() OVER (
               PARTITION BY ct.name
               ORDER BY ci.population DESC
           ) AS rn
    FROM continent ct
    JOIN country c
      ON c.continent = ct.name
    JOIN city ci
      ON ci.country_id = c.id
) t
WHERE rn <= 3;
4. FULL 100‑question SQL exam (questions only)
4.1 Basics (SELECT, WHERE, ORDER BY, LIMIT) — 20 questions
Select all columns from country.

Select only name and population from country.

Select distinct region values from country.

Find all countries in the Europe region.

Find all countries with population greater than 10 million.

Find all countries with population between 1M and 5M.

Find all countries whose name starts with A.

Find all countries whose name contains land.

Sort countries by population descending.

Sort countries by continent then name.

Return the top 10 most populated countries.

Return the 5 smallest countries by surface area.

Find countries where capital is NULL.

Find countries where capital is NOT NULL.

Find countries in Asia or Europe.

Find countries not in Africa.

Find countries whose population is not between 1M and 10M.

Find countries whose name length is greater than 10 characters.

Find countries whose name is exactly 5 characters long.

Find countries whose name is in a given list ('India','China','USA').

4.2 Joins — 15 questions
List all cities with their country name.

List all capitals with their country name (using city.is_capital).

List all countries and their number of cities.

List countries that have no cities.

List cities and their continent (via country).

List countries and their official languages.

List countries and count of official languages.

List countries that have more than 2 official languages.

List countries that share the same capital name (if any).

List each continent and total number of cities in it.

List countries and their largest city by population.

List countries and their smallest city by population.

List countries where at least one city has population > 5M.

List cities that belong to countries in Europe.

List continents that have at least one country with more than 50M population.

4.3 Aggregates & GROUP BY — 15 questions
Find total world population.

Find average country population.

Find maximum country population.

Find minimum country population.

Find total population per continent.

Find average surface area per continent.

Find number of countries per continent.

Find number of countries per region.

Find continents with total population > 1B.

Find regions with average population > 20M.

Find continents where the smallest country has population > 5M.

Find continents where the largest country has surface area > 1M.

Find the number of cities per country.

Find the average city population per country.

Find countries where the sum of city populations exceeds the country population.

4.4 Subqueries — 15 questions
Find countries with population above global average.

Find countries with surface area above global average.

Find countries whose population is greater than any city in them.

Find countries whose population is less than the sum of their cities.

Find countries that have at least one official language.

Find countries that have no official language.

Find cities in the top 5 most populated countries.

Find countries in the top 3 continents by total population.

Find countries whose capital city population is above the average city population.

Find continents whose total population is in the top 50% of continents.

Find countries whose population is in the top 10% of all countries.

Find countries that have more than 5 cities (subquery version).

Find countries that share at least one language with India.

Find countries that speak all languages spoken in Switzerland (set‑based).

Find countries whose population is greater than the population of Germany.

4.5 Window functions — 15 questions
Rank countries by population globally.

Rank countries by population within each continent.

Compute running total of population ordered by population descending.

Compute percentage of world population per country.

Compute average population per continent and show it alongside each country.

Compute difference between each country’s population and its continent average.

Rank cities by population within each country.

Find top 3 cities per country by population.

Find top 3 cities per continent by population.

Compute moving average of population over 3 countries ordered by population.

Assign dense rank to continents by total population.

Compute cumulative number of countries per continent ordered by population.

Compute row number of each country ordered by surface area.

Compute lag of population (previous country in population order).

Compute lead of population (next country in population order).

4.6 Advanced filtering, CASE, and HAVING — 20 questions
Find countries where population > 10M AND surface_area > 500000.

Find countries where population > 10M OR surface_area > 500000.

Find countries NOT in Asia and NOT in Europe.

Classify countries as Small, Medium, Large by population using CASE.

Count how many Small, Medium, Large countries exist per continent.

Find continents where more than half of countries are Large.

Find countries where city count > 10 using HAVING.

Find countries where average city population > 1M.

Find continents where max country population > 200M.

Find regions where sum of population between 50M and 500M.

Find countries with at least one city named like their capital.

Find countries where no city has population > 500k.

Find countries where all cities have population > 100k (set logic).

Find countries where at least one language is official and at least one is not.

Find continents where number of countries > 10.

Find continents where average surface area < 200k.

Find countries whose name and capital start with the same letter.

Find countries whose name is alphabetically after their capital.

Find countries where population density (population/surface_area) > 500.

Bucket countries into density bands and count per band.

5. Python (pandas) equivalents — core patterns
Assume:

python
import pandas as pd

countries = pd.DataFrame(...)  # country
cities = pd.DataFrame(...)     # city
languages = pd.DataFrame(...)  # language
continents = pd.DataFrame(...) # continent
5.1 Basic SELECT / WHERE / ORDER BY
python
# All columns
countries

# Select columns
countries[['name', 'population']]

# Filter
countries[countries['region'] == 'Europe']

# Population > 10M
countries[countries['population'] > 10_000_000]

# Sort
countries.sort_values('population', ascending=False)

# Top 10
countries.nlargest(10, 'population')
5.2 Joins
python
# Cities with country name
cities.merge(countries, left_on='country_id', right_on='id',
             suffixes=('_city', '_country'))

# Capitals with country
capitals = cities[cities['is_capital']]
capitals.merge(countries, left_on='country_id', right_on='id')
5.3 GroupBy + Aggregates
python
# Total world population
countries['population'].sum()

# Population by continent
countries.groupby('continent')['population'].sum()

# City count per country
cities.groupby('country_id')['id'].count()
5.4 Subquery‑like patterns
python
# Countries above global average population
avg_pop = countries['population'].mean()
countries[countries['population'] > avg_pop]

# Countries with > 3 cities
city_counts = cities.groupby('country_id')['id'].count()
countries[countries['id'].isin(city_counts[city_counts > 3].index)]
5.5 Window functions (via transform / expanding / rolling)
python
# Rank by population within continent
countries['pop_rank_cont'] = (
    countries
    .groupby('continent')['population']
    .rank(method='dense', ascending=False)
)

# Percentage of world population
total_pop = countries['population'].sum()
countries['pct_world'] = 100 * countries['population'] / total_pop

# Running total by population
countries_sorted = countries.sort_values('population', ascending=False)
countries_sorted['running_total'] = countries_sorted['population'].cumsum()
```
