```python
CREATE OR REPLACE PROCEDURE calc(a NUMBER, b NUMBER, op VARCHAR2) AS
  result NUMBER;
BEGIN
  CASE op
    WHEN '+' THEN result := a + b;
    WHEN '-' THEN result := a - b;
    WHEN '*' THEN result := a * b;
    WHEN '/' THEN result := a / b;
  END CASE;

  DBMS_OUTPUT.PUT_LINE('Result: ' || result);
END;
/

BEGIN
  calc(20, 4, '/');
END;
/
```

```python
###High-level explanation of a Mario-style game -> A classic Mario-style game is a 2D side-scrolling platformer with:
"""
Core mechanics:

Run & jump: Player moves left/right and jumps with gravity pulling them down.

Platforms & collisions: Solid tiles (ground, bricks, pipes) that the player can stand on, bump into, or break.

Enemies: Simple AI (walk back and forth, hurt on contact, can be defeated by jumping on them).

Collectibles: Coins, power-ups (like mushrooms) that modify player state (bigger, extra life, etc.).

Level progression: Reach a goal (flag, door, pipe) to finish the level.

Below are two separate projects:

A Python mini-platformer (using pygame)

An SQL project modeling Mario-style game data (players, levels, enemies, coins, scores)
"""
Project 1 – Python mini Mario-style platformer (pygame)
Minimal but structured; you can expand physics, animations, enemies, etc.

Folder structure
text
        mario_python/
        ├── main.py
        ├── settings.py
        └── entities.py
        settings.py

python
# settings.py

              # Window
              WIDTH = 800
              HEIGHT = 600
              TITLE = "Mini Mario Clone"

              # Colors
              WHITE = (255, 255, 255)
              BLUE = (0, 0, 255)
              GREEN = (0, 200, 0)
              BLACK = (0, 0, 0)

              # Physics
              GRAVITY = 0.5
              PLAYER_SPEED = 5
              JUMP_STRENGTH = -12

              # Level layout (simple tile map)
              # 'X' = solid block, '.' = empty
              LEVEL_MAP = [
                  "................................",
                  "................................",
                  "................................",
                  "................................",
                  "......................XXX.......",
                  "......................XXX.......",
                  ".............XXX................",
                  "............................XXXX",
                  "XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX",
              ]
              TILE_SIZE = 32

------------------------------
entities.py
python
# entities.py
              import pygame
              from settings import GRAVITY, PLAYER_SPEED, JUMP_STRENGTH, TILE_SIZE, BLUE, GREEN

              class Player(pygame.sprite.Sprite):
                  def __init__(self, x, y, platforms):
                      super().__init__()
                      self.image = pygame.Surface((TILE_SIZE, TILE_SIZE * 1.5))
                      self.image.fill(BLUE)
                      self.rect = self.image.get_rect(topleft=(x, y))

                      self.vel_x = 0
                      self.vel_y = 0
                      self.on_ground = False
                      self.platforms = platforms

                  def handle_input(self, keys):
                      self.vel_x = 0
                      if keys[pygame.K_LEFT]:
                          self.vel_x = -PLAYER_SPEED
                      if keys[pygame.K_RIGHT]:
                          self.vel_x = PLAYER_SPEED
                      if keys[pygame.K_SPACE] and self.on_ground:
                          self.vel_y = JUMP_STRENGTH
                          self.on_ground = False

                  def apply_gravity(self):
                      self.vel_y += GRAVITY
                      if self.vel_y > 20:
                          self.vel_y = 20

                  def horizontal_collisions(self):
                      self.rect.x += self.vel_x
                      for platform in self.platforms:
                          if self.rect.colliderect(platform.rect):
                              if self.vel_x > 0:
                                  self.rect.right = platform.rect.left
                              elif self.vel_x < 0:
                                  self.rect.left = platform.rect.right

                  def vertical_collisions(self):
                      self.rect.y += self.vel_y
                      self.on_ground = False
                      for platform in self.platforms:
                          if self.rect.colliderect(platform.rect):
                              if self.vel_y > 0:
                                  self.rect.bottom = platform.rect.top
                                  self.vel_y = 0
                                  self.on_ground = True
                              elif self.vel_y < 0:
                                  self.rect.top = platform.rect.bottom
                                  self.vel_y = 0

                  def update(self, keys):
                      self.handle_input(keys)
                      self.apply_gravity()
                      self.horizontal_collisions()
                      self.vertical_collisions()


              class Platform(pygame.sprite.Sprite):
                  def __init__(self, x, y, width=TILE_SIZE, height=TILE_SIZE):
                      super().__init__()
                      self.image = pygame.Surface((width, height))
                      self.image.fill(GREEN)
                      self.rect = self.image.get_rect(topleft=(x, y))
-------------------------------------
main.py
python
# main.py

              import pygame
              from settings import (
                  WIDTH, HEIGHT, TITLE, WHITE, BLACK,
                  LEVEL_MAP, TILE_SIZE
              )
              from entities import Player, Platform

              def build_level(level_map):
                  platforms = pygame.sprite.Group()
                  for row_idx, row in enumerate(level_map):
                      for col_idx, cell in enumerate(row):
                          if cell == "X":
                              x = col_idx * TILE_SIZE
                              y = row_idx * TILE_SIZE
                              platforms.add(Platform(x, y))
                  return platforms

              def main():
                  pygame.init()
                  screen = pygame.display.set_mode((WIDTH, HEIGHT))
                  pygame.display.set_caption(TITLE)
                  clock = pygame.time.Clock()

                  platforms = build_level(LEVEL_MAP)
                  player = Player(100, 100, platforms)
                  all_sprites = pygame.sprite.Group(platforms, player)

                  camera_x = 0

                  running = True
                  while running:
                      dt = clock.tick(60) / 1000  # not used yet, but handy for future

                      for event in pygame.event.get():
                          if event.type == pygame.QUIT:
                              running = False

                      keys = pygame.key.get_pressed()
                      player.update(keys)

                      # Simple side-scrolling camera following player
                      camera_x = -player.rect.centerx + WIDTH // 2

                      screen.fill(WHITE)

                      for sprite in all_sprites:
                          offset_rect = sprite.rect.move(camera_x, 0)
                          screen.blit(sprite.image, offset_rect)

                      # Simple HUD
                      font = pygame.font.SysFont(None, 24)
                      text = font.render("Mini Mario Clone - Arrows + Space to jump", True, BLACK)
                      screen.blit(text, (10, 10))

                      pygame.display.flip()

                  pygame.quit()

              if __name__ == "__main__":
                  main()
------------------
To run: pip install pygame, then python main.py.

You can extend this with enemies, coins, power-ups, multiple levels, and a proper tile editor.
------------------------------
#### Project 2 – SQL project for a Mario-style game backend ->This models players, levels, enemies, coins, and scores—useful for analytics, leaderboards, or saving progress.

Folder structure
text
      mario_sql/
      ├── schema.sql
      ├── sample_data.sql
      └── queries.sql

schema.sql
sql
              -- schema.sql

              -- Players of the game
              CREATE TABLE players (
                  player_id      SERIAL PRIMARY KEY,
                  username       VARCHAR(50) UNIQUE NOT NULL,
                  created_at     TIMESTAMP NOT NULL DEFAULT NOW()
              );

              -- Levels in the game
              CREATE TABLE levels (
                  level_id       SERIAL PRIMARY KEY,
                  world_number   INT NOT NULL,
                  level_number   INT NOT NULL,
                  name           VARCHAR(100),
                  difficulty     VARCHAR(20),
                  UNIQUE (world_number, level_number)
              );

              -- Enemies that can appear in levels
              CREATE TABLE enemy_types (
                  enemy_type_id  SERIAL PRIMARY KEY,
                  name           VARCHAR(50) NOT NULL,
                  description    TEXT,
                  base_points    INT NOT NULL
              );

              -- Coins / collectibles
              CREATE TABLE collectible_types (
                  collectible_type_id SERIAL PRIMARY KEY,
                  name                VARCHAR(50) NOT NULL,
                  description         TEXT,
                  base_points         INT NOT NULL
              );

              -- Instances of enemies in a level (for analytics / design)
              CREATE TABLE level_enemies (
                  level_enemy_id  SERIAL PRIMARY KEY,
                  level_id        INT NOT NULL REFERENCES levels(level_id),
                  enemy_type_id   INT NOT NULL REFERENCES enemy_types(enemy_type_id),
                  position_x      INT,
                  position_y      INT
              );

              -- Instances of collectibles in a level
              CREATE TABLE level_collectibles (
                  level_collectible_id SERIAL PRIMARY KEY,
                  level_id             INT NOT NULL REFERENCES levels(level_id),
                  collectible_type_id  INT NOT NULL REFERENCES collectible_types(collectible_type_id),
                  position_x           INT,
                  position_y           INT
              );

              -- Player runs (attempts) of a level
              CREATE TABLE level_runs (
                  run_id          SERIAL PRIMARY KEY,
                  player_id       INT NOT NULL REFERENCES players(player_id),
                  level_id        INT NOT NULL REFERENCES levels(level_id),
                  started_at      TIMESTAMP NOT NULL DEFAULT NOW(),
                  finished_at     TIMESTAMP,
                  completed       BOOLEAN NOT NULL DEFAULT FALSE,
                  total_score     INT NOT NULL DEFAULT 0,
                  total_coins     INT NOT NULL DEFAULT 0,
                  total_deaths    INT NOT NULL DEFAULT 0
              );

              -- Detailed events during a run (defeating enemies, collecting coins, dying, etc.)
              CREATE TABLE run_events (
                  event_id        SERIAL PRIMARY KEY,
                  run_id          INT NOT NULL REFERENCES level_runs(run_id),
                  event_time      TIMESTAMP NOT NULL DEFAULT NOW(),
                  event_type      VARCHAR(50) NOT NULL, -- 'ENEMY_DEFEATED', 'COIN_COLLECTED', 'DEATH', etc.
                  enemy_type_id   INT REFERENCES enemy_types(enemy_type_id),
                  collectible_type_id INT REFERENCES collectible_types(collectible_type_id),
                  points_earned   INT NOT NULL DEFAULT 0,
                  extra_info      JSONB
              );
---------------------------------
sample_data.sql
sql
-- sample_data.sql

              INSERT INTO players (username) VALUES
              ('mario_fan'),
              ('speedrunner'),
              ('casual_player');

              INSERT INTO levels (world_number, level_number, name, difficulty) VALUES
              (1, 1, 'World 1-1', 'Easy'),
              (1, 2, 'World 1-2', 'Easy'),
              (2, 1, 'World 2-1', 'Medium');

              INSERT INTO enemy_types (name, description, base_points) VALUES
              ('Goomba', 'Walks back and forth, defeated by jumping on it', 100),
              ('Koopa', 'Turtle-like enemy, shell can be kicked', 200);

              INSERT INTO collectible_types (name, description, base_points) VALUES
              ('Coin', 'Standard coin', 10),
              ('1-Up Mushroom', 'Grants extra life', 0);

              -- Example level content
              INSERT INTO level_enemies (level_id, enemy_type_id, position_x, position_y) VALUES
              (1, 1, 200, 400),
              (1, 1, 400, 400),
              (1, 2, 600, 400);

              INSERT INTO level_collectibles (level_id, collectible_type_id, position_x, position_y) VALUES
              (1, 1, 150, 350),
              (1, 1, 250, 350),
              (1, 2, 800, 350);

              -- Example run
              INSERT INTO level_runs (player_id, level_id, completed, total_score, total_coins, total_deaths)
              VALUES (1, 1, TRUE, 230, 3, 0);

              INSERT INTO run_events (run_id, event_type, enemy_type_id, points_earned)
              VALUES
              (1, 'ENEMY_DEFEATED', 1, 100),
              (1, 'ENEMY_DEFEATED', 1, 100),
              (1, 'COIN_COLLECTED', NULL, 10),
              (1, 'COIN_COLLECTED', NULL, 10),
              (1, 'COIN_COLLECTED', NULL, 10);
----------------------------
queries.sql – example analytics/queries
sql
-- queries.sql

-- 1. Leaderboard: top players by total score across all runs
            SELECT
                p.username,
                SUM(r.total_score) AS total_score
            FROM players p
            JOIN level_runs r ON r.player_id = p.player_id
            GROUP BY p.username
            ORDER BY total_score DESC
            LIMIT 10;

-- 2. Average completion rate per level
          SELECT
              l.world_number,
              l.level_number,
              l.name,
              COUNT(*) AS attempts,
              SUM(CASE WHEN completed THEN 1 ELSE 0 END) AS completions,
              ROUND(
                  100.0 * SUM(CASE WHEN completed THEN 1 ELSE 0 END) / NULLIF(COUNT(*), 0),
                  2
              ) AS completion_rate_pct
          FROM levels l
          LEFT JOIN level_runs r ON r.level_id = l.level_id
          GROUP BY l.level_id
          ORDER BY l.world_number, l.level_number;

-- 3. Average score per completed run for each player
            SELECT
                p.username,
                AVG(r.total_score) AS avg_score_completed_runs
            FROM players p
            JOIN level_runs r ON r.player_id = p.player_id
            WHERE r.completed = TRUE
            GROUP BY p.username
            ORDER BY avg_score_completed_runs DESC;

-- 4. Most common enemy type defeated
            SELECT
                e.name,
                COUNT(*) AS times_defeated
            FROM run_events ev
            JOIN enemy_types e ON ev.enemy_type_id = e.enemy_type_id
            WHERE ev.event_type = 'ENEMY_DEFEATED'
            GROUP BY e.name
            ORDER BY times_defeated DESC;

-- 5. Total coins collected per player
            SELECT
                p.username,
                SUM(r.total_coins) AS coins_collected
            FROM players p
            JOIN level_runs r ON r.player_id = p.player_id
            GROUP BY p.username
            ORDER BY coins_collected DESC;
```
