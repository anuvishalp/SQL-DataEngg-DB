```python
###PySpark Complete Cheatsheet

1. What Spark is and Why Distributed Processing Exists
1A.What is Spark?

        Apache Spark: Open-source unified analytics engine for large-scale data processing
        PySpark: Python API for Spark
        Core Concept: Distributed computing framework that processes data across multiple machines

        Spark is a distributed computing engine. Instead of processing data on one machine,
        it splits data across many machines and processes them in parallel.

        PySpark is simply Spark + Python.

1B.Why Distributed Processing?

        Volume: Handle datasets too large for single machine memory (TB-PB scale)
        Speed: Parallel processing across cluster nodes
        Fault Tolerance: Automatic recovery from node failures
        Scalability: Add more nodes to handle more data

A single machine cannot handle terabytes or petabytes of data.

Spark solves this by:

      Splitting data across multiple machines
      Processing in parallel
      Recovering automatically if a machine fails
      Scaling horizontally (add more machines → more power)

2. Spark Architecture

        Driver Program (Master)
            ↓
        SparkContext
            ↓
        Cluster Manager (YARN/Mesos/Standalone)
            ↓
        Worker Nodes (Executors) → Tasks

Key Components

          Driver: Orchestrates the job, maintains cluster info
          Executors: Run tasks and store data
          Cluster Manager: Allocates resources
          RDD (Resilient Distributed Dataset): Immutable distributed collection
          DataFrame: Distributed collection with named columns (like SQL table)


Component	->What It Does

        Driver->Runs your main program; coordinates everything
        SparkContext->	Connects your program to the cluster
        Cluster Manager->	Allocates machines (YARN, Mesos, Standalone)
        Executors	->Worker processes that run your code
        Tasks	->Smallest unit of work executed on executors

        RDD → low-level, unstructured, functional transformations
        DataFrame → high-level, optimized, SQL-like, column-based

        DataFrames are preferred because Spark can optimize them using Catalyst optimizer

2. How DataFrames Work and How They Differ from Pandas

PySpark DataFrame vs Pandas DataFrame

FeaturePySpark DataFramePandas DataFrameData L
ocationDistributed across clusterIn-memory on single machine
Size LimitPetabytesLimited by RAM (~GBs)
ExecutionLazy evaluationEager
evaluationMutabilityImmutableMutableAPI
Style Functional transformations Object-oriented
IndexingNo indexRow/column
indexesSpeed (small data)Slower (overhead)FasterSpeed (big data)Much faster (parallel)Cannot handle

Creating DataFrames

from pyspark.sql import SparkSession

# Initialize Spark Session
spark = SparkSession.builder \
    .appName("MyApp") \
    .master("local[*]") \
    .getOrCreate()

Explanation:

        SparkSession is the entry point to PySpark.
        .appName() names your job.
        .master("local[*]") runs Spark locally using all CPU cores.
        .getOrCreate() returns an existing session or creates a new one.

# From list of tuples
data = [("Alice", 25), ("Bob", 30), ("Charlie", 35)]
df = spark.createDataFrame(data, ["name", "age"])

        Each tuple becomes a row.
        Column names are provided as a list.
        Spark infers schema automatically.

# From list of dictionaries
data = [{"name": "Alice", "age": 25}, {"name": "Bob", "age": 30}]
df = spark.createDataFrame(data)

        Keys become column names.
        Values become row values.

# From pandas DataFrame
import pandas as pd
pandas_df = pd.DataFrame({"name": ["Alice", "Bob"], "age": [25, 30]})
spark_df = spark.createDataFrame(pandas_df)

        Useful for small data or testing.
        Spark converts Pandas types to Spark types.

# From RDD
rdd = spark.sparkContext.parallelize([(1, "Alice"), (2, "Bob")])
df = rdd.toDF(["id", "name"])

        parallelize() distributes Python list across cluster.
        .toDF() converts RDD → DataFrame.

Basic DataFrame Operations
python# Show data
df.show()  # Display first 20 rows
df.show(5, truncate=False)  # Show 5 rows without truncating

        show() prints first 20 rows.
        truncate=False prevents cutting long strings.

# Schema and structure
df.printSchema()  # Show column types
df.columns  # List column names
df.dtypes  # List columns with types
df.count()  # Number of rows
df.describe().show()  # Statistical summary

        printSchema() → tree view of column names & types
        columns → list of column names
        dtypes → list of (column, datatype)
        count() → number of rows
        describe() → summary stats (count, mean, std, min, max)

# Convert to Pandas (for small data only!)
pandas_df = df.toPandas()

        Converts Spark DataFrame → Pandas DataFrame
        Only safe for small datasets (fits in driver memory)
===============================================
3. Reading Data from Files and Tables
===============================================
A.Reading CSV Files
python# Basic read
          df = spark.read.csv("path/to/file.csv")

# With options
          df = spark.read.csv(
              "path/to/file.csv",
              header=True,  # First row as column names
              inferSchema=True,  # Automatically infer data types
              sep=",",  # Delimiter
              quote='"',  # Quote character
              escape="\\",  # Escape character
              nullValue="NA",  # String to treat as null
              dateFormat="yyyy-MM-dd"
          )

# Alternative syntax
            df = spark.read \
                .option("header", "true") \
                .option("inferSchema", "true") \
                .csv("path/to/file.csv")

# Read multiple files
              df = spark.read.csv("path/to/*.csv")
              df = spark.read.csv(["file1.csv", "file2.csv"])

===============================================
Reading JSON Files
===============================================
python# Simple JSON (one object per line)
        df = spark.read.json("path/to/file.json")

# Multi-line JSON
        df = spark.read.option("multiLine", "true").json("path/to/file.json")

# With schema
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

schema = StructType([
    StructField("name", StringType(), True),
    StructField("age", IntegerType(), True)
])
df = spark.read.schema(schema).json("path/to/file.json")
===============================================
Reading Parquet Files
===============================================
python# Parquet (columnar format, efficient for Spark)
df = spark.read.parquet("path/to/file.parquet")

# Multiple parquet files
df = spark.read.parquet("path/to/parquet_dir")
===============================================
Reading from Databases
===============================================
python# JDBC connection
df = spark.read.format("jdbc") \
    .option("url", "jdbc:postgresql://localhost:5432/mydb") \
    .option("dbtable", "users") \
    .option("user", "username") \
    .option("password", "password") \
    .option("driver", "org.postgresql.Driver") \
    .load()

# Or with query
df = spark.read.format("jdbc") \
    .option("url", "jdbc:postgresql://localhost:5432/mydb") \
    .option("query", "SELECT * FROM users WHERE age > 25") \
    .option("user", "username") \
    .option("password", "password") \
    .load()
===============================================
Reading Hive Tables
===============================================
python# Enable Hive support
spark = SparkSession.builder.enableHiveSupport().getOrCreate()

# Read from Hive
df = spark.sql("SELECT * FROM database.table")
df = spark.table("database.table")
===============================================
Other Formats
===============================================
python# Text file (each line as a row)

          df = spark.read.text("path/to/file.txt")

# ORC
          df = spark.read.orc("path/to/file.orc")

# Avro
          df = spark.read.format("avro").load("path/to/file.avro")
===============================================
4. Filtering, Transforming, and Aggregating Data
===============================================
Selecting Columns

from pyspark.sql.functions import col, lit

# Select specific columns
          df.select("name", "age").show() -> Select only the name and age columns.
          df.select(col("name"), col("age")).show() ->Same as above, but using col() for explicit column objects.
          df.select(df.name, df.age).show()

# Select all columns
          df.select("*").show()

# Select with expressions
          df.select(col("name"), (col("age") + 1).alias("next_age")).show()

        Select name
        Create a new computed column next_age = age + 1

# Add new columns
    df.withColumn("age_plus_10", col("age") + 10).show()
    df.withColumn("constant", lit("value")).show()

      Adds a new column without modifying the original DataFrame (Spark DataFrames are immutable)

# Rename columns
  df.withColumnRenamed("name", "full_name").show() -> Renames a column.

# Drop columns
      df.drop("age").show() -> Removes the age column.
      df.drop(col("age")).show() -> Same as above, but using col().
      df.drop("age", "salary").show()
===============================================
##Filtering Rows
===============================================
from pyspark.sql.functions import col

# Filter with condition
      df.filter(col("age") > 25).show() ->Keep only rows where age > 25.
      df.filter(df.age > 25).show()
      df.where(col("age") > 25).show()  # Same as filter

# Multiple conditions -> using AND, OR condition
      df.filter((col("age") > 25) & (col("salary") > 50000)).show()
      df.filter((col("age") < 20) | (col("age") > 60)).show()

# NOT condition
      df.filter(~(col("age") > 25)).show()

# String operations ->Names starting with A.
      df.filter(col("name").startswith("A")).show() ->Names starting with A.
      df.filter(col("name").endswith("e")).show() ->Names ending with A.
      df.filter(col("name").contains("li")).show()
      df.filter(col("name").like("A%")).show()  # SQL-like pattern
      df.filter(col("name").rlike("^A.*e$")).show()  # Regex

# NULL checks
      df.filter(col("age").isNull()).show() ->Rows where age is NULL.
      df.filter(col("age").isNotNull()).show() ->Rows where age is NOT NULL.

# IN clause
      df.filter(col("name").isin(["Alice", "Bob"])).show()

# Between
      df.filter(col("age").between(20, 30)).show()
===============================================
Transformations
===============================================
from pyspark.sql.functions import *

# String functions
        df.withColumn("upper_name", upper(col("name"))).show() ->Convert name to uppercase.
        df.withColumn("lower_name", lower(col("name"))).show()->Convert name to lower case.
        df.withColumn("name_length", length(col("name"))).show()
        df.withColumn("trimmed", trim(col("name"))).show()
        df.withColumn("substring", substring(col("name"), 1, 3)).show()
        df.withColumn("concat", concat(col("first_name"), lit(" "), col("last_name"))).show()
        df.withColumn("split", split(col("name"), " ")).show()

# Numeric functions
        df.withColumn("rounded_age", round(col("age"), 0)).show() ->Round age to nearest integer.
        df.withColumn("absolute", abs(col("value"))).show()
        df.withColumn("power", pow(col("value"), 2)).show()
        df.withColumn("sqrt", sqrt(col("value"))).show()

# Date functions
        df.withColumn("current_date", current_date()).show() -> Shows current date
        df.withColumn("current_timestamp", current_timestamp()).show() ->current time
        df.withColumn("year", year(col("date_column"))).show() ->Extract year from a date.
        df.withColumn("month", month(col("date_column"))).show()
        df.withColumn("day", dayofmonth(col("date_column"))).show()
        df.withColumn("date_add", date_add(col("date_column"), 7)).show()
        df.withColumn("date_diff", datediff(col("end_date"), col("start_date"))).show()
        df.withColumn("formatted", date_format(col("date_column"), "yyyy-MM-dd")).show()
===============================================
# Type casting
        df.withColumn("age_string", col("age").cast("string")).show()
        df.withColumn("age_int", col("age_string").cast("int")).show()

# Conditional logic -> same as SQL-When
        df.withColumn("age_group",
            when(col("age") < 18, "Minor")
            .when((col("age") >= 18) & (col("age") < 65), "Adult")
            .otherwise("Senior")
        ).show()

# Replace values ->Replace nulls in specific columns.
    df.na.replace(["old_value"], ["new_value"], subset=["column_name"]).show()

# Fill null values
      df.na.fill(0).show()  # Fill all numeric nulls with 0
      df.na.fill({"age": 0, "name": "Unknown"}).show()  # Column-specific fills

# Drop null values
      df.na.drop().show()  # Drop rows with any null
      df.na.drop(how="all").show()  # Drop only if all values are null
      df.na.drop(subset=["age", "name"]).show()  # Drop if null in specific columns
======================================
Aggregations
======================================
from pyspark.sql.functions import *

# Basic aggregations ->Compute multiple aggregations in one call.
        df.agg(
            count("*").alias("total_count"),
            sum("salary").alias("total_salary"),
            avg("age").alias("avg_age"),
            min("age").alias("min_age"),
            max("age").alias("max_age"),
            stddev("salary").alias("std_salary")
        ).show()

# Group by ->Group rows by department and compute average salary.

        df.groupBy("department").agg(
                                      count("*").alias("count"),
                                      avg("salary").alias("avg_salary")
                                  ).show()

# Multiple group by columns
        df.groupBy("department", "location").agg(
            avg("salary").alias("avg_salary")
        ).show()

##Having clause:
        df.groupBy("department")
          .agg(avg("salary").alias("avg_salary"))
          .filter(col("avg_salary") > 60000)
          .show()

# Group by with multiple aggregations
          df.groupBy("department").agg(
              count("*").alias("employee_count"),
              sum("salary").alias("total_salary"),
              avg("age").alias("avg_age"),
              min("hire_date").alias("earliest_hire"),
              max("salary").alias("max_salary")
          ).show()

# Having clause (filter after grouping)
          df.groupBy("department") \
              .agg(avg("salary").alias("avg_salary")) \
              .filter(col("avg_salary") > 60000) \
              .show()

# Collect aggregated values as list -> collect_list()->Creates an array of names per department.
          df.groupBy("department") \
              .agg(collect_list("name").alias("employee_names")) \
              .show()

# Count distinct
        df.agg(countDistinct("department").alias("unique_departments")).show()
============================================
Sorting
============================================
python# Sort ascending
          df.orderBy("age").show()
          df.sort("age").show()

# Sort descending
          df.orderBy(col("age").desc()).show()
          df.orderBy(desc("age")).show()

# Multiple columns
          df.orderBy("department", col("salary").desc()).show()
==========================================
Distinct and Duplicates
==========================================
python# Get distinct rows
        df.distinct().show() ->Remove duplicate rows

# Get distinct values in a column
        df.select("department").distinct().show()

# Drop duplicates
      df.dropDuplicates().show()
      df.dropDuplicates(["name"]).show()  # Based on specific columns
============================
Joins
============================
python# Inner join ->join on column id

        df1.join(df2, df1.id == df2.id, "inner").show()
        df1.join(df2, "id", "inner").show()  # If same column name

# Left join
        df1.join(df2, "id", "left").show()

# Right join
        df1.join(df2, "id", "right").show()

# Outer join
        df1.join(df2, "id", "outer").show()

# Cross join
        df1.crossJoin(df2).show()

# Join on multiple columns
        df1.join(df2, ["id", "department"], "inner").show()

Union and Set Operations -># Union (combine rows)
        df1.union(df2).show()
        df1.unionByName(df2).show()  # Match by column names

# Intersection
        -> df1.intersect(df2).show()

# Except (difference)
        -> df1.subtract(df2).show()  # Rows in df1 but not in df2
==============================================
5. Writing Transformed Data Back to Storage
==============================================
Writing CSV ->python# Basic write
        df.write.csv("output/path")

# With options

        df.write \
            .mode("overwrite") \
            .option("header", "true") \
            .csv("output/path")

# Write modes
          df.write.mode("overwrite").csv("output/path")  # Replace existing
          df.write.mode("append").csv("output/path")  # Add to existing
          df.write.mode("ignore").csv("output/path")  # Do nothing if exists
          df.write.mode("error").csv("output/path")  # Throw error if exists (default)

# Single file output-> Reduce to 1 partition → 1 output file.(Use carefully; can cause bottlenecks)
        df.coalesce(1).write.option("header", "true").csv("output/path")
=====================================
# Partitioned write (creates subdirectories)
          df.write.partitionBy("year", "month").csv("output/path")
          df.write.mode("overwrite").partitionBy("year", "month").csv("output/path")

=====================================
### Writing Parquet-Parquet is columnar, compressed, and fastest for Spark
=====================================
python# Parquet (recommended for Spark)

      df.write.parquet("output/path")

# With compression
df.write \
    .mode("overwrite") \
    .option("compression", "snappy") \
    .parquet("output/path")

# Compression options: none, snappy, gzip, lzo, brotli, lz4, zstd
========================================
Writing JSON -> python# JSON
========================================
df.write.json("output/path")

# Pretty print (not recommended for large data)
df.write.option("pretty", "true").json("output/path")
========================================
Writing to Databases
========================================

## python# JDBC write -> Writes DataFrame to a database table.
df.write.format("jdbc") \
    .mode("overwrite") \
    .option("url", "jdbc:postgresql://localhost:5432/mydb") \
    .option("dbtable", "users") \
    .option("user", "username") \
    .option("password", "password") \
    .option("driver", "org.postgresql.Driver") \
    .save()

# Batch insert for better performance
df.write.format("jdbc") \
    .mode("append") \
    .option("url", "jdbc:postgresql://localhost:5432/mydb") \
    .option("dbtable", "users") \
    .option("user", "username") \
    .option("password", "password") \
    .option("batchsize", "10000") \
    .save()
=============================================
Writing to Hive -> python# Save as Hive table
=============================================
#Creates or replaces a Hive table
      df.write.mode("overwrite").saveAsTable("database.table_name")

# With partitioning
df.write \
    .mode("overwrite") \
    .partitionBy("year", "month") \
    .saveAsTable("database.table_name")

# External table
df.write \
    .mode("overwrite") \
    .option("path", "/path/to/data") \
    .saveAsTable("database.table_name")
================================
Other Formats
python# ORC
================================
        df.write.orc("output/path")

# Avro ->
      df.write.format("avro").save("output/path")

# Text (single column only)
      df.select("text_column").write.text("output/path")
=======================================
Advanced Write Options
=======================================
# Write to multiple files
python# Control number of output files

        df.repartition(10).write.parquet("output/path")  # 10 files
        df.coalesce(1).write.parquet("output/path")  # 1 file (use carefully)

# Bucket by column (for optimized joins later)
df.write \
    .bucketBy(10, "id") \
    .sortBy("age") \
    .saveAsTable("bucketed_table")

# Save with custom partition structure
df.write \
    .partitionBy("year", "month", "day") \
    .mode("overwrite") \
    .parquet("output/path")
================================================
6. Understanding Lazy Execution at a High Level
================================================
What is Lazy Evaluation?
    Lazy Evaluation: Spark doesn't execute transformations immediately.
    Instead, it builds a logical plan and executes only when an action is called.

Transformations vs Actions
Transformations (Lazy - not executed immediately):

            select(), filter(), where()
            groupBy(), join(), union()
            withColumn(), drop(), distinct()
            orderBy(), sort()
            Returns a new DataFrame

Actions (Eager - trigger execution):

            show(), count(), collect()
            take(), first(), head()
            write(), save(), saveAsTable()
            Returns results to driver or writes to storage
============================================================
Example of Lazy Execution
python# Nothing executes yet - just building logical plan
============================================================
            df1 = spark.read.csv("input.csv")  # Transformation
            df2 = df1.filter(col("age") > 25)  # Transformation
            df3 = df2.select("name", "age")  # Transformation
            df4 = df3.orderBy("age")  # Transformation

            # NOW everything executes
            df4.show()  # Action - triggers the entire pipeline

Benefits of Lazy Evaluation

            Optimization: Spark can optimize the entire query plan before execution
            Efficiency: Avoids unnecessary computations
            Pipelining: Combines multiple operations
            Fault Tolerance: Can recompute lost partitions
=======================
Query Plans
=======================
python# See logical plan
        df.explain()

# See physical plan
        df.explain(True)

# See optimized plan
        df.explain("formatted")

Caching and Persistence
python# Cache in memory (use for DataFrames accessed multiple times)
          df.cache()  # Same as df.persist()
          df.persist()  # Default: MEMORY_AND_DISK

# Specific storage levels
          from pyspark import StorageLevel
          df.persist(StorageLevel.MEMORY_ONLY)
          df.persist(StorageLevel.DISK_ONLY)
          df.persist(StorageLevel.MEMORY_AND_DISK)
          df.persist(StorageLevel.MEMORY_AND_DISK_SER)  # Serialized

# Unpersist to free memory
          df.unpersist()

# Check if cached
          df.is_cached

# Example use case
from pyspark.sql.functions import col

          cached_df = df.filter(col("age") > 25).cache()
          cached_df.count()  # First action - caches the result
          cached_df.select("name").show()  # Uses cached data (faster)
          cached_df.groupBy("department").count().show()  # Uses cached data

Partitioning
python# Check current partitions
          df.rdd.getNumPartitions()

# Repartition (increases or decreases, full shuffle)
          df_repartitioned = df.repartition(10)
          df_repartitioned = df.repartition(10, "department")  # Partition by column

# Coalesce (only decreases, no shuffle - more efficient)
          df_coalesced = df.coalesce(5)

# Why partitioning matters:
# - More partitions = more parallelism
# - Too many partitions = overhead
# - Too few partitions = underutilized cluster
# Rule of thumb: 2-4 partitions per CPU core

7. Knowing When to Use PySpark Instead of SQL
When to Use PySpark

1.          Complex Data Transformations

    1.A              Advanced logic with Python functions
     1.B             Machine learning pipelines
      1.C            Custom UDFs (User Defined Functions)



                from pyspark.sql.functions import udf
                from pyspark.sql.types import StringType

                # Complex Python logic
                def complex_transformation(value):
                    # Your custom logic here
                    return value.upper() if value else "DEFAULT"

                transform_udf = udf(complex_transformation, StringType())
                df.withColumn("transformed", transform_udf(col("name"))).show()

2.Integration with Python Libraries

          Pandas interoperability
          NumPy, SciPy for calculations
          ML libraries (scikit-learn, TensorFlow)


Procedural Logic

          Iterative processing
          Dynamic query building
          Complex control flow


Large-Scale Data Processing

          Files in distributed storage (HDFS, S3)
          Data too large for traditional databases
          Need for horizontal scaling

When to Use SQL

Familiar Syntax

          Team knows SQL better
          Standard reporting queries
          Quick ad-hoc analysis

python# Register DataFrame as temp view

            df.createOrReplaceTempView("users")

            # Use SQL
            result = spark.sql("""
                SELECT department,
                      AVG(salary) as avg_salary,
                      COUNT(*) as employee_count
                FROM users
                WHERE age > 25
                GROUP BY department
                HAVING AVG(salary) > 60000
                ORDER BY avg_salary DESC
            """)
            result.show()

Simple Queries

            Straightforward SELECT, WHERE, GROUP BY
            Standard aggregations
            Basic joins

Query Reusability

            Existing SQL codebase
            Standard business logic
            Database migrations

Working with Hive/Database

            Querying existing tables
            Maintaining compatibility
            Using database-specific features

Combining Both Approaches
            python# Read data with PySpark
            df = spark.read.parquet("data.parquet")

            # Use PySpark for complex transformations
            df_transformed = df.filter(col("age") > 25) \
                .withColumn("age_group", when(col("age") < 40, "Young").otherwise("Senior"))

            # Register as temp view
            df_transformed.createOrReplaceTempView("transformed_users")

            # Use SQL for familiar aggregations
            result = spark.sql("""
                SELECT age_group,
                      department,
                      AVG(salary) as avg_salary
                FROM transformed_users
                GROUP BY age_group, department
            """)

            # Continue with PySpark
            final_result = result.filter(col("avg_salary") > 50000)
            final_result.show()
=======================================
Performance Considerations
1.Both PySpark DataFrame API and SQL:

              Use the same Catalyst optimizer
              Generate the same execution plan
              Have identical performance

2.Choose based on:

              Team expertise
              Code readability
              Maintenance requirements
              Integration needs

Quick Reference: Common Patterns
==============================
Pattern 1: ETL Pipeline
==============================
python# Extract

        df = spark.read.csv("input.csv", header=True, inferSchema=True)

# Transform

        df_clean = df.filter(col("age").isNotNull()) \
            .withColumn("age_group", when(col("age") < 30, "Young").otherwise("Senior")) \
            .groupBy("department", "age_group") \
            .agg(avg("salary").alias("avg_salary"))

# Load
        df_clean.write.mode("overwrite").parquet("output.parquet")

==============================
Pattern 2: Data Quality Check
==============================
python# Check for nulls
          null_counts = df.select([
              count(when(col(c).isNull(), c)).alias(c)
              for c in df.columns
          ])
          null_counts.show()

# Check duplicates
            duplicate_count = df.count() - df.dropDuplicates().count()
            print(f"Duplicates: {duplicate_count}")

# Value range checks
            df.select(min("age"), max("age"), avg("age")).show()
==============================
Pattern 3: Window Functions
==============================
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number, rank, dense_rank, lag, lead

# Partition window
        window_spec = Window.partitionBy("department").orderBy(col("salary").desc())

        df.withColumn("rank", rank().over(window_spec)) \
            .withColumn("row_num", row_number().over(window_spec)) \
            .withColumn("prev_salary", lag("salary", 1).over(window_spec)) \
            .show()

# Running totals
        window_running = Window.partitionBy("department").orderBy("date") \
            .rowsBetween(Window.unboundedPreceding, Window.currentRow)

        df.withColumn("running_total", sum("amount").over(window_running)).show()
==============================
Pattern 4: Pivot Tables
==============================
# Pivot
        df.groupBy("department") \
            .pivot("year") \
            .agg(sum("revenue")) \
            .show()

# Unpivot (use stack)
        df.selectExpr(
            "department",
            "stack(3, '2021', year_2021, '2022', year_2022, '2023', year_2023) as (year, revenue)"
        ).show()


Performance Tips

          Avoid collect() on large DataFrames (brings all data to driver)
          Use parquet format for better performance
          Partition data appropriately (by commonly filtered columns)
          Cache DataFrames used multiple times
          Use broadcast joins for small tables
          Avoid UDFs when possible (use built-in functions)
          Filter early in the transformation chain
          Use coalesce instead of repartition when reducing partitions

python# Broadcast join (for small tables)
from pyspark.sql.functions import broadcast
large_df.join(broadcast(small_df), "key").show()

# Monitor execution
        spark.sparkContext.setLogLevel("INFO")
=====================================
Common Functions Reference
=====================================
String Functions

          concat(), concat_ws(), split(), substring()
          upper(), lower(), initcap(), trim(), ltrim(), rtrim()
          regexp_replace(), regexp_extract()
          length(), reverse()

Numeric Functions

          abs(), ceil(), floor(), round()
          sqrt(), pow(), exp(), log()
          sin(), cos(), tan()
          rand(), randn()

Date Functions

          current_date(), current_timestamp()
          year(), month(), dayofmonth(), hour(), minute(), second()
          date_add(), date_sub(), datediff(), months_between()
          to_date(), to_timestamp(), date_format()
          last_day(), next_day(), weekofyear()

Aggregate Functions

          count(), sum(), avg(), min(), max()
          stddev(), variance()
          collect_list(), collect_set()
          first(), last()

Array Functions

          array(), array_contains(), explode(), size()
          sort_array(), array_distinct(), array_union()


Practice these patterns with real datasets to build proficiency.

Remember:

          Lazy evaluation means plan your transformations carefully
          Partitioning and caching are key to performance
          Choose the right tool: PySpark for complex logic, SQL for familiarity
          Test at scale: What works on small data may fail at scale
```
